```python
import pygame
import sys

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Pygame
pygame.init()

# –ö–æ–Ω—Å—Ç–∞–Ω—Ç—ã
WIDTH, HEIGHT = 800, 600
CELL_WIDTH, CELL_HEIGHT = 20, 20
WHITE = (255, 255, 255)
GRAY = (200, 200, 200)
DARK_GRAY = (100, 100, 100)

# –°–æ–∑–¥–∞–Ω–∏–µ –æ–∫–Ω–∞
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Grid Test")
clock = pygame.time.Clock()

# –ö–ª–∞—Å—Å Grid
class Grid:
    def __init__(self, width, height, cell_width, cell_height):
        self.width = width
        self.height = height
        self.cell_width = cell_width
        self.cell_height = cell_height
        self.cols = width // cell_width
        self.rows = height // cell_height
        self.filled_cells = set()
    
    def fill_cell(self, col, row):
        """–ó–∞–ø–æ–ª–Ω–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if 0 <= col < self.cols and 0 <= row < self.rows:
            self.filled_cells.add((col, row))
    
    def clear_cell(self, col, row):
        """–û—á–∏—Å—Ç–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if (col, row) in self.filled_cells:
            self.filled_cells.remove((col, row))

def draw_grid(screen, grid):
    """
    –û—Ç—Ä–∏—Å–æ–≤–∞—Ç—å —Å–µ—Ç–∫—É –Ω–∞ —ç–∫—Ä–∞–Ω–µ

    Args:
        screen: –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å pygame –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è
        grid: –æ–±—ä–µ–∫—Ç Grid —Å –¥–∞–Ω–Ω—ã–º–∏ —Å–µ—Ç–∫–∏
    """
    # –®–∞–≥ 1: –û—á–∏—Å—Ç–∏—Ç—å —ç–∫—Ä–∞–Ω –±–µ–ª—ã–º —Ü–≤–µ—Ç–æ–º
    screen.fill(WHITE)

    # –®–∞–≥ 2: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≤—Å–µ –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã–µ –∫–ª–µ—Ç–∫–∏
    for col, row in grid.filled_cells:
        # –í—ã—á–∏—Å–ª—è–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –ª–µ–≤–æ–≥–æ –≤–µ—Ä—Ö–Ω–µ–≥–æ —É–≥–ª–∞ –∫–ª–µ—Ç–∫–∏
        x = col * grid.cell_width
        y = row * grid.cell_height

        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ–π –∫–ª–µ—Ç–∫–∏
        pygame.draw.rect(screen, DARK_GRAY, (x, y, grid.cell_width, grid.cell_height))

    # –®–∞–≥ 3: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏ —Å–µ—Ç–∫–∏
    for col in range(grid.cols + 1):
        x = col * grid.cell_width
        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—É—é –ª–∏–Ω–∏—é
        pygame.draw.line(screen, GRAY, (x, 0), (x, grid.height), 1)

    # –®–∞–≥ 4: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏ —Å–µ—Ç–∫–∏
    for row in range(grid.rows + 1):
        # –í—ã—á–∏—Å–ª–∏—Ç–µ y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É
        y = row * grid.cell_height
        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—É—é –ª–∏–Ω–∏—é
        pygame.draw.line(screen, GRAY, (0, y), (grid.width, y), 1)

# –°–æ–∑–¥–∞—ë–º —Å–µ—Ç–∫—É –∏ –∑–∞–ø–æ–ª–Ω—è–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –∫–ª–µ—Ç–æ–∫ –¥–ª—è —Ç–µ—Å—Ç–∞
grid = Grid(WIDTH, HEIGHT, CELL_WIDTH, CELL_HEIGHT)

# –ó–∞–ø–æ–ª–Ω—è–µ–º –∫–ª–µ—Ç–∫–∏ –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö —Å–µ—Ç–∫–∏ (–∫–æ–ª–æ–Ω–∫–∏ 0-39, —Å—Ç—Ä–æ–∫–∏ 0-29)
grid.fill_cell(5, 5)
grid.fill_cell(6, 5)
grid.fill_cell(7, 5)
grid.fill_cell(5, 6)
grid.fill_cell(5, 7)

# –ü—Ä–æ—Å—Ç–æ–π –∏–≥—Ä–æ–≤–æ–π —Ü–∏–∫–ª –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # –†–∏—Å—É–µ–º —Å–µ—Ç–∫—É
    draw_grid(screen, grid)

    # –û–±–Ω–æ–≤–ª—è–µ–º —ç–∫—Ä–∞–Ω
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
sys.exit()
```

```python
import pygame
import sys

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Pygame
pygame.init()

# –ö–æ–Ω—Å—Ç–∞–Ω—Ç—ã
WIDTH, HEIGHT = 800, 600
CELL_WIDTH, CELL_HEIGHT = 20, 20
WHITE = (255, 255, 255)
GRAY = (200, 200, 200)
DARK_GRAY = (100, 100, 100)

# –°–æ–∑–¥–∞–Ω–∏–µ –æ–∫–Ω–∞
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Grid Test")
clock = pygame.time.Clock()

# –ö–ª–∞—Å—Å Grid
class Grid:
    def __init__(self, width, height, cell_width, cell_height):
        self.width = width
        self.height = height
        self.cell_width = cell_width
        self.cell_height = cell_height
        self.cols = width // cell_width
        self.rows = height // cell_height
        self.filled_cells = set()
    
    def get_cell_at_position(self, x, y):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω–¥–µ–∫—Å—ã –∫–ª–µ—Ç–∫–∏ –ø–æ —ç–∫—Ä–∞–Ω–Ω—ã–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º"""
        if x < 0 or x >= self.width or y < 0 or y >= self.height:
            return None
        col = int(x // self.cell_width)
        row = int(y // self.cell_height)
        return (col, row)
    
    def fill_cell(self, col, row):
        """–ó–∞–ø–æ–ª–Ω–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if 0 <= col < self.cols and 0 <= row < self.rows:
            self.filled_cells.add((col, row))
    
    def clear_cell(self, col, row):
        """–û—á–∏—Å—Ç–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if (col, row) in self.filled_cells:
            self.filled_cells.remove((col, row))

def draw_grid(screen, grid):
    """
    –û—Ç—Ä–∏—Å–æ–≤–∞—Ç—å —Å–µ—Ç–∫—É –Ω–∞ —ç–∫—Ä–∞–Ω–µ

    Args:
        screen: –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å pygame –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è
        grid: –æ–±—ä–µ–∫—Ç Grid —Å –¥–∞–Ω–Ω—ã–º–∏ —Å–µ—Ç–∫–∏
    """
    # –®–∞–≥ 1: –û—á–∏—Å—Ç–∏—Ç—å —ç–∫—Ä–∞–Ω –±–µ–ª—ã–º —Ü–≤–µ—Ç–æ–º
    screen.fill(WHITE)

    # –®–∞–≥ 2: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≤—Å–µ –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã–µ –∫–ª–µ—Ç–∫–∏
    for col, row in grid.filled_cells:
        # –í—ã—á–∏—Å–ª—è–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –ª–µ–≤–æ–≥–æ –≤–µ—Ä—Ö–Ω–µ–≥–æ —É–≥–ª–∞ –∫–ª–µ—Ç–∫–∏
        x = col * grid.cell_width
        y = row * grid.cell_height

        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ–π –∫–ª–µ—Ç–∫–∏
        pygame.draw.rect(screen, DARK_GRAY, (x, y, grid.cell_width, grid.cell_height))

    # –®–∞–≥ 3: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏ —Å–µ—Ç–∫–∏
    for col in range(grid.cols + 1):
        x = col * grid.cell_width
        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—É—é –ª–∏–Ω–∏—é
        pygame.draw.line(screen, GRAY, (x, 0), (x, grid.height), 1)

    # –®–∞–≥ 4: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏ —Å–µ—Ç–∫–∏
    for row in range(grid.rows + 1):
        # –í—ã—á–∏—Å–ª–∏—Ç–µ y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É
        y = row * grid.cell_height
        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—É—é –ª–∏–Ω–∏—é
        pygame.draw.line(screen, GRAY, (0, y), (grid.width, y), 1)

# –°–æ–∑–¥–∞—ë–º –Ω–æ–≤—É—é —Å–µ—Ç–∫—É (–ø—É—Å—Ç—É—é)
grid = Grid(WIDTH, HEIGHT, CELL_WIDTH, CELL_HEIGHT)

# –ò–≥—Ä–æ–≤–æ–π —Ü–∏–∫–ª —Å –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –º—ã—à–∏
running = True
while running:
    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–æ–±—ã—Ç–∏–π
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–ª–∏–∫–∞ –º—ã—à–∏
        elif event.type == pygame.MOUSEBUTTONDOWN:
            # –ü–æ–ª—É—á–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∫–ª–∏–∫–∞
            mouse_x, mouse_y = event.pos

            # –ü–æ–ª—É—á–∞–µ–º –∫–ª–µ—Ç–∫—É –ø–æ–¥ –∫—É—Ä—Å–æ—Ä–æ–º
            cell = grid.get_cell_at_position(mouse_x, mouse_y)

            # –ï—Å–ª–∏ –∫–ª–∏–∫ –≤–Ω—É—Ç—Ä–∏ —Å–µ—Ç–∫–∏
            if cell is not None:
                col, row = cell

                # –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, –∫–∞–∫–∞—è –∫–Ω–æ–ø–∫–∞ –º—ã—à–∏ –Ω–∞–∂–∞—Ç–∞
                # event.button == 1 - –ª–µ–≤–∞—è –∫–Ω–æ–ø–∫–∞
                if event.button == 1:
                    # –ó–∞–ø–æ–ª–Ω–∏—Ç–µ –∫–ª–µ—Ç–∫—É
                    grid.fill_cell(col, row)
                    print(f"–ó–∞–ø–æ–ª–Ω–µ–Ω–∞ –∫–ª–µ—Ç–∫–∞ ({col}, {row})")

    # –†–∏—Å—É–µ–º —Å–µ—Ç–∫—É
    draw_grid(screen, grid)

    # –û–±–Ω–æ–≤–ª—è–µ–º —ç–∫—Ä–∞–Ω
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```

```python
import pygame
import sys

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Pygame
pygame.init()

# –ö–æ–Ω—Å—Ç–∞–Ω—Ç—ã
WIDTH, HEIGHT = 800, 600
CELL_WIDTH, CELL_HEIGHT = 20, 20
WHITE = (255, 255, 255)
GRAY = (200, 200, 200)
DARK_GRAY = (100, 100, 100)

# –°–æ–∑–¥–∞–Ω–∏–µ –æ–∫–Ω–∞
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Grid Test")
clock = pygame.time.Clock()

# –ö–ª–∞—Å—Å Grid
class Grid:
    def __init__(self, width, height, cell_width, cell_height):
        self.width = width
        self.height = height
        self.cell_width = cell_width
        self.cell_height = cell_height
        self.cols = width // cell_width
        self.rows = height // cell_height
        self.filled_cells = set()
    
    def get_cell_at_position(self, x, y):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω–¥–µ–∫—Å—ã –∫–ª–µ—Ç–∫–∏ –ø–æ —ç–∫—Ä–∞–Ω–Ω—ã–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º"""
        if x < 0 or x >= self.width or y < 0 or y >= self.height:
            return None
        col = int(x // self.cell_width)
        row = int(y // self.cell_height)
        return (col, row)
    
    def fill_cell(self, col, row):
        """–ó–∞–ø–æ–ª–Ω–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if 0 <= col < self.cols and 0 <= row < self.rows:
            self.filled_cells.add((col, row))
    
    def clear_cell(self, col, row):
        """–û—á–∏—Å—Ç–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if (col, row) in self.filled_cells:
            self.filled_cells.remove((col, row))

def draw_grid(screen, grid):
    """
    –û—Ç—Ä–∏—Å–æ–≤–∞—Ç—å —Å–µ—Ç–∫—É –Ω–∞ —ç–∫—Ä–∞–Ω–µ

    Args:
        screen: –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å pygame –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è
        grid: –æ–±—ä–µ–∫—Ç Grid —Å –¥–∞–Ω–Ω—ã–º–∏ —Å–µ—Ç–∫–∏
    """
    # –®–∞–≥ 1: –û—á–∏—Å—Ç–∏—Ç—å —ç–∫—Ä–∞–Ω –±–µ–ª—ã–º —Ü–≤–µ—Ç–æ–º
    screen.fill(WHITE)

    # –®–∞–≥ 2: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≤—Å–µ –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã–µ –∫–ª–µ—Ç–∫–∏
    for col, row in grid.filled_cells:
        # –í—ã—á–∏—Å–ª—è–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –ª–µ–≤–æ–≥–æ –≤–µ—Ä—Ö–Ω–µ–≥–æ —É–≥–ª–∞ –∫–ª–µ—Ç–∫–∏
        x = col * grid.cell_width
        y = row * grid.cell_height

        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ–π –∫–ª–µ—Ç–∫–∏
        pygame.draw.rect(screen, DARK_GRAY, (x, y, grid.cell_width, grid.cell_height))

    # –®–∞–≥ 3: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏ —Å–µ—Ç–∫–∏
    for col in range(grid.cols + 1):
        x = col * grid.cell_width
        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—É—é –ª–∏–Ω–∏—é
        pygame.draw.line(screen, GRAY, (x, 0), (x, grid.height), 1)

    # –®–∞–≥ 4: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏ —Å–µ—Ç–∫–∏
    for row in range(grid.rows + 1):
        # –í—ã—á–∏—Å–ª–∏—Ç–µ y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É
        y = row * grid.cell_height
        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—É—é –ª–∏–Ω–∏—é
        pygame.draw.line(screen, GRAY, (0, y), (grid.width, y), 1)
# –°–æ–∑–¥–∞—ë–º –Ω–æ–≤—É—é —Å–µ—Ç–∫—É
grid = Grid(WIDTH, HEIGHT, CELL_WIDTH, CELL_HEIGHT)

# TODO: –§–ª–∞–≥–∏ —Å–æ—Å—Ç–æ—è–Ω–∏—è –º—ã—à–∏
left_mouse_pressed = False  # –ó–∞–∂–∞—Ç–∞ –ª–∏ –ª–µ–≤–∞—è –∫–Ω–æ–ø–∫–∞?
last_drawn_cell = None      # –ü–æ—Å–ª–µ–¥–Ω—è—è –Ω–∞—Ä–∏—Å–æ–≤–∞–Ω–Ω–∞—è –∫–ª–µ—Ç–∫–∞

# –ò–≥—Ä–æ–≤–æ–π —Ü–∏–∫–ª
running = True
while running:
    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–æ–±—ã—Ç–∏–π
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        # –ù–∞–∂–∞—Ç–∏–µ –∫–Ω–æ–ø–∫–∏ –º—ã—à–∏
        elif event.type == pygame.MOUSEBUTTONDOWN:
            if event.button == 1:  # –õ–µ–≤–∞—è –∫–Ω–æ–ø–∫–∞
                # TODO: –£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ —Ñ–ª–∞–≥, —á—Ç–æ –∫–Ω–æ–ø–∫–∞ –∑–∞–∂–∞—Ç–∞
                left_mouse_pressed = True
                # TODO: –°–±—Ä–æ—Å—å—Ç–µ last_drawn_cell
                last_drawn_cell = None

        # –û—Ç–ø—É—Å–∫–∞–Ω–∏–µ –∫–Ω–æ–ø–∫–∏ –º—ã—à–∏
        elif event.type == pygame.MOUSEBUTTONUP:
            if event.button == 1:  # –õ–µ–≤–∞—è –∫–Ω–æ–ø–∫–∞
                # TODO: –°–±—Ä–æ—Å—å—Ç–µ —Ñ–ª–∞–≥
                left_mouse_pressed = False
                last_drawn_cell = None

    # TODO: –†–∏—Å–æ–≤–∞–Ω–∏–µ –ø—Ä–∏ –¥–≤–∏–∂–µ–Ω–∏–∏ –º—ã—à–∏ (–≤–Ω–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Å–æ–±—ã—Ç–∏–π!)
    # –ü–æ–ª—É—á–∞–µ–º —Ç–µ–∫—É—â—É—é –ø–æ–∑–∏—Ü–∏—é –º—ã—à–∏
    mouse_x, mouse_y = pygame.mouse.get_pos()

    # –ï—Å–ª–∏ –ª–µ–≤–∞—è –∫–Ω–æ–ø–∫–∞ –∑–∞–∂–∞—Ç–∞
    if left_mouse_pressed:
        # –ü–æ–ª—É—á–∞–µ–º –∫–ª–µ—Ç–∫—É –ø–æ–¥ –∫—É—Ä—Å–æ—Ä–æ–º
        cell = grid.get_cell_at_position(mouse_x, mouse_y)

        if cell is not None:
            col, row = cell

            # TODO: –†–∏—Å—É–µ–º —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ —ç—Ç–æ –ù–û–í–ê–Ø –∫–ª–µ—Ç–∫–∞
            # –≠—Ç–æ –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—â–∞–µ—Ç –º–Ω–æ–≥–æ–∫—Ä–∞—Ç–Ω–æ–µ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –æ–¥–Ω–æ–π –∫–ª–µ—Ç–∫–∏
            if cell != last_drawn_cell:
                grid.fill_cell(col, row)
                # TODO: –ó–∞–ø–æ–º–∏–Ω–∞–µ–º –ø–æ—Å–ª–µ–¥–Ω—é—é –Ω–∞—Ä–∏—Å–æ–≤–∞–Ω–Ω—É—é –∫–ª–µ—Ç–∫—É
                last_drawn_cell = cell

    # –†–∏—Å—É–µ–º —Å–µ—Ç–∫—É
    draw_grid(screen, grid)

    # –û–±–Ω–æ–≤–ª—è–µ–º —ç–∫—Ä–∞–Ω
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```

```python
import pygame
import sys

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Pygame
pygame.init()

# –ö–æ–Ω—Å—Ç–∞–Ω—Ç—ã
WIDTH, HEIGHT = 800, 600
CELL_WIDTH, CELL_HEIGHT = 20, 20
WHITE = (255, 255, 255)
GRAY = (200, 200, 200)
DARK_GRAY = (100, 100, 100)

# –°–æ–∑–¥–∞–Ω–∏–µ –æ–∫–Ω–∞
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Grid Test")
clock = pygame.time.Clock()

# –ö–ª–∞—Å—Å Grid
class Grid:
    def __init__(self, width, height, cell_width, cell_height):
        self.width = width
        self.height = height
        self.cell_width = cell_width
        self.cell_height = cell_height
        self.cols = width // cell_width
        self.rows = height // cell_height
        self.filled_cells = set()
    
    def get_cell_at_position(self, x, y):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω–¥–µ–∫—Å—ã –∫–ª–µ—Ç–∫–∏ –ø–æ —ç–∫—Ä–∞–Ω–Ω—ã–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º"""
        if x < 0 or x >= self.width or y < 0 or y >= self.height:
            return None
        col = int(x // self.cell_width)
        row = int(y // self.cell_height)
        return (col, row)
    
    def fill_cell(self, col, row):
        """–ó–∞–ø–æ–ª–Ω–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if 0 <= col < self.cols and 0 <= row < self.rows:
            self.filled_cells.add((col, row))
    
    def clear_cell(self, col, row):
        """–û—á–∏—Å—Ç–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if (col, row) in self.filled_cells:
            self.filled_cells.remove((col, row))

def draw_grid(screen, grid):
    """
    –û—Ç—Ä–∏—Å–æ–≤–∞—Ç—å —Å–µ—Ç–∫—É –Ω–∞ —ç–∫—Ä–∞–Ω–µ

    Args:
        screen: –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å pygame –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è
        grid: –æ–±—ä–µ–∫—Ç Grid —Å –¥–∞–Ω–Ω—ã–º–∏ —Å–µ—Ç–∫–∏
    """
    # –®–∞–≥ 1: –û—á–∏—Å—Ç–∏—Ç—å —ç–∫—Ä–∞–Ω –±–µ–ª—ã–º —Ü–≤–µ—Ç–æ–º
    screen.fill(WHITE)

    # –®–∞–≥ 2: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≤—Å–µ –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã–µ –∫–ª–µ—Ç–∫–∏
    for col, row in grid.filled_cells:
        # –í—ã—á–∏—Å–ª—è–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –ª–µ–≤–æ–≥–æ –≤–µ—Ä—Ö–Ω–µ–≥–æ —É–≥–ª–∞ –∫–ª–µ—Ç–∫–∏
        x = col * grid.cell_width
        y = row * grid.cell_height

        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ–π –∫–ª–µ—Ç–∫–∏
        pygame.draw.rect(screen, DARK_GRAY, (x, y, grid.cell_width, grid.cell_height))

    # –®–∞–≥ 3: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏ —Å–µ—Ç–∫–∏
    for col in range(grid.cols + 1):
        x = col * grid.cell_width
        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—É—é –ª–∏–Ω–∏—é
        pygame.draw.line(screen, GRAY, (x, 0), (x, grid.height), 1)

    # –®–∞–≥ 4: –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏ —Å–µ—Ç–∫–∏
    for row in range(grid.rows + 1):
        # –í—ã—á–∏—Å–ª–∏—Ç–µ y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É
        y = row * grid.cell_height
        # –ù–∞—Ä–∏—Å—É–π—Ç–µ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—É—é –ª–∏–Ω–∏—é
        pygame.draw.line(screen, GRAY, (0, y), (grid.width, y), 1)

# –°–æ–∑–¥–∞—ë–º –Ω–æ–≤—É—é —Å–µ—Ç–∫—É
grid = Grid(WIDTH, HEIGHT, CELL_WIDTH, CELL_HEIGHT)

# –§–ª–∞–≥–∏ —Å–æ—Å—Ç–æ—è–Ω–∏—è –º—ã—à–∏
left_mouse_pressed = False
last_drawn_cell = None

# TODO: –î–æ–±–∞–≤—å—Ç–µ —Ñ–ª–∞–≥–∏ –¥–ª—è –ø—Ä–∞–≤–æ–π –∫–Ω–æ–ø–∫–∏
right_mouse_pressed = False  # False
last_erased_cell = None     # None

# –ò–≥—Ä–æ–≤–æ–π —Ü–∏–∫–ª
running = True
while running:
    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–æ–±—ã—Ç–∏–π
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        # –ù–∞–∂–∞—Ç–∏–µ –∫–Ω–æ–ø–∫–∏ –º—ã—à–∏
        elif event.type == pygame.MOUSEBUTTONDOWN:
            if event.button == 1:  # –õ–µ–≤–∞—è –∫–Ω–æ–ø–∫–∞
                left_mouse_pressed = True
                last_drawn_cell = None

            # TODO: –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–∞–≤–æ–π –∫–Ω–æ–ø–∫–∏ (button == 3)
            elif event.button == 3:
                right_mouse_pressed = True
                last_erased_cell = None

        # –û—Ç–ø—É—Å–∫–∞–Ω–∏–µ –∫–Ω–æ–ø–∫–∏ –º—ã—à–∏
        elif event.type == pygame.MOUSEBUTTONUP:
            if event.button == 1:  # –õ–µ–≤–∞—è –∫–Ω–æ–ø–∫–∞
                left_mouse_pressed = False
                last_drawn_cell = None

            # TODO: –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–∞–≤–æ–π –∫–Ω–æ–ø–∫–∏
            elif event.button == 3:
                right_mouse_pressed = False
                last_erased_cell = None

    # –ü–æ–ª—É—á–∞–µ–º —Ç–µ–∫—É—â—É—é –ø–æ–∑–∏—Ü–∏—é –º—ã—à–∏
    mouse_x, mouse_y = pygame.mouse.get_pos()

    # –†–∏—Å–æ–≤–∞–Ω–∏–µ –ø—Ä–∏ –∑–∞–∂–∞—Ç–æ–π –ª–µ–≤–æ–π –∫–Ω–æ–ø–∫–µ
    if left_mouse_pressed:
        cell = grid.get_cell_at_position(mouse_x, mouse_y)
        if cell is not None:
            col, row = cell
            if cell != last_drawn_cell:
                grid.fill_cell(col, row)
                last_drawn_cell = cell

    # TODO: –°—Ç–∏—Ä–∞–Ω–∏–µ –ø—Ä–∏ –∑–∞–∂–∞—Ç–æ–π –ø—Ä–∞–≤–æ–π –∫–Ω–æ–ø–∫–µ
    if right_mouse_pressed:  # right_mouse_pressed
        cell = grid.get_cell_at_position(mouse_x, mouse_y)
        if cell is not None:
            col, row = cell
            # –°—Ç–∏—Ä–∞–µ–º —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ —ç—Ç–æ –Ω–æ–≤–∞—è –∫–ª–µ—Ç–∫–∞
            if cell != last_erased_cell:  # last_erased_cell
                # TODO: –û—á–∏—Å—Ç–∏—Ç–µ –∫–ª–µ—Ç–∫—É
                grid.clear_cell(col, row)
                # TODO: –ó–∞–ø–æ–º–Ω–∏—Ç–µ –ø–æ—Å–ª–µ–¥–Ω—é—é —Å—Ç–µ—Ä—Ç—É—é –∫–ª–µ—Ç–∫—É
                last_erased_cell = cell

    # –†–∏—Å—É–µ–º —Å–µ—Ç–∫—É
    draw_grid(screen, grid)

    # –û–±–Ω–æ–≤–ª—è–µ–º —ç–∫—Ä–∞–Ω
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```

```python
# üéØ –ü–æ–ª–Ω—ã–π –∫–æ–¥ —Å –ø–æ–¥—Å–≤–µ—Ç–∫–æ–π –∫–ª–µ—Ç–∫–∏ –ø–æ–¥ –∫—É—Ä—Å–æ—Ä–æ–º
import pygame
import sys

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è Pygame
pygame.init()

# –ö–æ–Ω—Å—Ç–∞–Ω—Ç—ã
WIDTH, HEIGHT = 800, 600
CELL_WIDTH, CELL_HEIGHT = 20, 20
WHITE = (255, 255, 255)
GRAY = (200, 200, 200)
DARK_GRAY = (100, 100, 100)
BLUE = (150, 150, 255)  # –¶–≤–µ—Ç –ø–æ–¥—Å–≤–µ—Ç–∫–∏

# –°–æ–∑–¥–∞–Ω–∏–µ –æ–∫–Ω–∞
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Grid Test - Draw, Erase & Highlight")
clock = pygame.time.Clock()

# –ö–ª–∞—Å—Å Grid
class Grid:
    def __init__(self, width, height, cell_width, cell_height):
        self.width = width
        self.height = height
        self.cell_width = cell_width
        self.cell_height = cell_height
        self.cols = width // cell_width
        self.rows = height // cell_height
        self.filled_cells = set()
    
    def get_cell_at_position(self, x, y):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω–¥–µ–∫—Å—ã –∫–ª–µ—Ç–∫–∏ –ø–æ —ç–∫—Ä–∞–Ω–Ω—ã–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º"""
        if x < 0 or x >= self.width or y < 0 or y >= self.height:
            return None
        col = int(x // self.cell_width)
        row = int(y // self.cell_height)
        return (col, row)
    
    def fill_cell(self, col, row):
        """–ó–∞–ø–æ–ª–Ω–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if 0 <= col < self.cols and 0 <= row < self.rows:
            self.filled_cells.add((col, row))
    
    def clear_cell(self, col, row):
        """–û—á–∏—Å—Ç–∏—Ç—å –∫–ª–µ—Ç–∫—É –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º —Å–µ—Ç–∫–∏"""
        if (col, row) in self.filled_cells:
            self.filled_cells.remove((col, row))
    
    def is_cell_filled(self, col, row):
        """–ü—Ä–æ–≤–µ—Ä–∏—Ç—å, –∑–∞–ø–æ–ª–Ω–µ–Ω–∞ –ª–∏ –∫–ª–µ—Ç–∫–∞"""
        return (col, row) in self.filled_cells

def draw_grid(screen, grid, hover_cell=None):
    """
    –û—Ç—Ä–∏—Å–æ–≤–∞—Ç—å —Å–µ—Ç–∫—É –Ω–∞ —ç–∫—Ä–∞–Ω–µ

    Args:
        screen: –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å pygame –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è
        grid: –æ–±—ä–µ–∫—Ç Grid —Å –¥–∞–Ω–Ω—ã–º–∏ —Å–µ—Ç–∫–∏
        hover_cell: –∫–ª–µ—Ç–∫–∞ –ø–æ–¥ –∫—É—Ä—Å–æ—Ä–æ–º (col, row) –∏–ª–∏ None
    """
    # –û—á–∏—Å—Ç–∏—Ç—å —ç–∫—Ä–∞–Ω
    screen.fill(WHITE)

    # –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –≤—Å–µ –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã–µ –∫–ª–µ—Ç–∫–∏
    for col, row in grid.filled_cells:
        x = col * grid.cell_width
        y = row * grid.cell_height
        pygame.draw.rect(screen, DARK_GRAY, (x, y, grid.cell_width, grid.cell_height))

    # –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –ø–æ–¥—Å–≤–µ—Ç–∫—É –∫–ª–µ—Ç–∫–∏ –ø–æ–¥ –∫—É—Ä—Å–æ—Ä–æ–º
    if hover_cell is not None:
        col, row = hover_cell

        # –ü–æ–¥—Å–≤–µ—á–∏–≤–∞–µ–º —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ –∫–ª–µ—Ç–∫–∞ –ù–ï –∑–∞–ø–æ–ª–Ω–µ–Ω–∞
        if not grid.is_cell_filled(col, row):
            x = col * grid.cell_width
            y = row * grid.cell_height

            # –ù–∞—Ä–∏—Å—É–π—Ç–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≥–æ–ª—É–±–æ–≥–æ —Ü–≤–µ—Ç–∞
            pygame.draw.rect(screen, BLUE, (x, y, grid.cell_width, grid.cell_height))

    # –ù–∞—Ä–∏—Å–æ–≤–∞—Ç—å –ª–∏–Ω–∏–∏ —Å–µ—Ç–∫–∏
    for col in range(grid.cols + 1):
        x = col * grid.cell_width
        pygame.draw.line(screen, GRAY, (x, 0), (x, grid.height), 1)

    for row in range(grid.rows + 1):
        y = row * grid.cell_height
        pygame.draw.line(screen, GRAY, (0, y), (grid.width, y), 1)

# –°–æ–∑–¥–∞—ë–º –Ω–æ–≤—É—é —Å–µ—Ç–∫—É
grid = Grid(WIDTH, HEIGHT, CELL_WIDTH, CELL_HEIGHT)

# –§–ª–∞–≥–∏ —Å–æ—Å—Ç–æ—è–Ω–∏—è –º—ã—à–∏
left_mouse_pressed = False
right_mouse_pressed = False
last_drawn_cell = None
last_erased_cell = None

# –ò–≥—Ä–æ–≤–æ–π —Ü–∏–∫–ª
running = True
while running:
    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–æ–±—ã—Ç–∏–π
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.MOUSEBUTTONDOWN:
            if event.button == 1:
                left_mouse_pressed = True
                last_drawn_cell = None
            elif event.button == 3:
                right_mouse_pressed = True
                last_erased_cell = None

        elif event.type == pygame.MOUSEBUTTONUP:
            if event.button == 1:
                left_mouse_pressed = False
                last_drawn_cell = None
            elif event.button == 3:
                right_mouse_pressed = False
                last_erased_cell = None

    # –ü–æ–ª—É—á–∞–µ–º —Ç–µ–∫—É—â—É—é –ø–æ–∑–∏—Ü–∏—é –º—ã—à–∏
    mouse_x, mouse_y = pygame.mouse.get_pos()

    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –∫–ª–µ—Ç–∫—É –ø–æ–¥ –∫—É—Ä—Å–æ—Ä–æ–º –¥–ª—è –ø–æ–¥—Å–≤–µ—Ç–∫–∏
    hover_cell = grid.get_cell_at_position(mouse_x, mouse_y)

    # –†–∏—Å–æ–≤–∞–Ω–∏–µ
    if left_mouse_pressed:
        cell = grid.get_cell_at_position(mouse_x, mouse_y)
        if cell is not None:
            col, row = cell
            if cell != last_drawn_cell:
                grid.fill_cell(col, row)
                last_drawn_cell = cell

    # –°—Ç–∏—Ä–∞–Ω–∏–µ
    if right_mouse_pressed:
        cell = grid.get_cell_at_position(mouse_x, mouse_y)
        if cell is not None:
            col, row = cell
            if cell != last_erased_cell:
                grid.clear_cell(col, row)
                last_erased_cell = cell

    # –†–∏—Å—É–µ–º —Å–µ—Ç–∫—É —Å –ø–æ–¥—Å–≤–µ—Ç–∫–æ–π
    draw_grid(screen, grid, hover_cell=hover_cell)

    # –û–±–Ω–æ–≤–ª—è–µ–º —ç–∫—Ä–∞–Ω
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```
