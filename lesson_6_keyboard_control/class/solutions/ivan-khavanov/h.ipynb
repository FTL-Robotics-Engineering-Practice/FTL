```python
import pygame
import math

pygame.init()

linear_speed = 0
angular_speed = 0

#linear_speed = 50
#angular_speed = 50

# TODO: Создайте окно 800x600
WIDTH = 800
HEIGHT = 600
screen = pygame.display.set_mode(size=(WIDTH, HEIGHT))

# TODO: Создайте часы для контроля FPS
clock = pygame.time.Clock()

# TODO: Определите цвета (WHITE, BLACK, BLUE)
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
BLUE = (0, 0, 255)

# TODO: Создайте шрифт размером 24
font = pygame.font.Font(None, 24)

dt = clock.tick(60) / 1000.0

# TODO: Переменные робота (математические координаты)
robot_x = 0  # начальная позиция в математических координатах
robot_y = 0  # начальная позиция в математических координатах
robot_direction = 0  # градусы, 0 = вправо

def draw_robot(screen, x, y, direction, radius):
    """Рисует робота с направлением"""
    # TODO: Преобразуйте математические координаты в экранные
    
    screen_x = x + WIDTH // 2
    screen_y = HEIGHT // 2 - y
    draw_info(screen, x, y, direction, linear_speed, angular_speed)
    # TODO: Нарисуйте круг (тело робота)
    pygame.draw.circle(screen, color=BLUE, center=(screen_x, screen_y), radius=radius)
    
    # TODO: Вычислите координаты треугольника для направления
    angle_rad = math.radians(direction)
    
    # Вершина треугольника (впереди)
    tip_x = screen_x + radius * math.cos(angle_rad)  # используйте screen_x, radius, cos(angle_rad)
    tip_y = screen_y - radius * math.sin(angle_rad)  # используйте screen_y, radius, sin(angle_rad)
    
    # Задние углы треугольника (левый и правый)
    # Подсказка: используйте углы (angle_rad + 2.5) и (angle_rad - 2.5)
    left_x = screen_x + radius * math.cos(angle_rad + 2.5) * 0.5
    left_y = screen_y - radius * math.sin(angle_rad + 2.5) * 0.5
    
    right_x = screen_x + radius * math.cos(angle_rad - 2.5) * 0.5
    right_y = screen_y - radius * math.sin(angle_rad - 2.5) * 0.5
    
    # TODO: Нарисуйте треугольник
    pygame.draw.polygon(surface=screen, color=WHITE, points=[(tip_x, tip_y), (left_x, left_y), (right_x, right_y)])
    
    # TODO: Нарисуйте контур круга (толщина 2)
    pygame.draw.circle(surface=screen, color=BLACK, center=(screen_x, screen_y), radius=radius, width=2)

def draw_info(screen, x, y, direction, linear_speed, angular_speed):
    """Отображает информацию о роботе на экране"""
    # TODO: Создайте список строк с информацией
    lines = [
        # Подсказка: используйте f-строки
        # Позиция: округлите x и y до целых через int()
        # Направление: нормализуйте через % 360
        # Скорости: округлите до целых
        f"Позиция: ({int(x)}, {int(y)})",  # Позиция: (150, -50)        # Скорость: 100 пикс/с
        f"Угол: {round(angular_speed*dt % 360, 3):.3f}°",  
        f"Скорость: {linear_speed} пикс/с"
    ]
    
    # TODO: Отрисуйте каждую строку
    y_offset = 10
    for line in lines:
        print(line, type(line))
        text_surface = font.render(line, True, BLACK)
        screen.blit(source=text_surface, dest=(10, y_offset))
        y_offset += 25
        # Создайте текстовую поверхность через font.render()
        # Отобразите её через screen.blit()
        # Увеличьте y_offset на 25 для следующей строки
        pass
# Игровой цикл
running = True
while running:
    # TODO: Обработка событий
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            
     # TODO: Вычисляйте delta time
    dt = clock.tick(60) / 1000.0

    # TODO: Обнуляем скорости (если клавиши не нажаты, робот стоит)
    linear_speed = 0
    angular_speed = 0

    # TODO: Получаем состояние всех клавиш
    keys = pygame.key.get_pressed()

    # TODO: Управление линейной скоростью
    # Сначала проверьте конфликт: если W и S нажаты одновременно
    if keys[pygame.K_w] and keys[pygame.K_s]:
        linear_speed = 0
    # Затем проверьте W отдельно (используйте elif!)
    elif keys[pygame.K_w]:
        linear_speed = 150
    # Затем проверьте S отдельно (используйте elif!)
    elif keys[pygame.K_s]:
        linear_speed = -100
    # TODO: Управление угловой скоростью
    # Можете сделать проверку A+D для полноты, или просто две отдельные проверки
    if keys[pygame.K_a]:
        angular_speed = 120  # Влево (против часовой)
    if keys[pygame.K_d]:
        angular_speed = -120  # Вправо (по часовой)
    
    # TODO: Обновите направление робота
    robot_direction += angular_speed * dt
    
    # TODO: Обновите позицию робота
    # Сначала вычислите угол в радианах
    angle_rad = math.radians(robot_direction)
    
    # Затем вычислите изменение координат и обновите позицию
    robot_x += linear_speed * math.cos(angle_rad) * dt
    robot_y += linear_speed * math.sin(angle_rad) * dt
    
    # TODO: Залить экран белым
    screen.fill(WHITE)
    
    # TODO: Нарисовать робота
    draw_robot(screen=screen, x=robot_x, y=robot_y, direction=robot_direction, radius=20)
    
    # Обновление экрана
    pygame.display.flip()
    
    

pygame.quit()
```
