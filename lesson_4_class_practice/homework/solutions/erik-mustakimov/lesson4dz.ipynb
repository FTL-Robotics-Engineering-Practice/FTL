```python
import math
import matplotlib.pyplot as plt

class Robot:
    def __init__(self, name):
        self.name = name
        self.x = 0
        self.y = 0
        self.direction = 0
        self.speed = 1
        self.angular_speed = 0
        self.dt = 0.01
        self.history = [(self.x, self.y)]
    
    def move_forward(self, distance):
        steps = int(distance / (self.speed * self.dt))
        for _ in range(steps):
            self.update()
    
    def turn(self, angle):
        self.direction += angle
    
    def set_angular_speed(self, new_speed):
        self.angular_speed = new_speed
    
    def update(self):
        rad_dir = math.radians(self.direction)
        self.x += self.speed * math.cos(rad_dir) * self.dt
        self.y += self.speed * math.sin(rad_dir) * self.dt
        self.direction += self.angular_speed * self.dt
        self.history.append((self.x, self.y))
    
    def draw_polygon(self, n, side_length):
        angle = 360 / n
        for _ in range(n):
            self.move_forward(side_length)
            self.turn(angle)
    
    def draw_rounded_polygon(self, n, side_length, radius):
        gg = 360 / n
        hh = self.speed / radius if radius > 0 else 0
        
        for _ in range(n):
            self.set_angular_speed(0)
            self.move_forward(side_length)
            
            if radius > 0:
                self.set_angular_speed(hh)
                kk = gg / hh
                mm = int(kk / self.dt)
                for _ in range(mm):
                    self.update()

plt.figure(figsize=(12, 10))

robot1 = Robot("лбЛђлхЛЃл│лЙл╗ЛїлйлИл║")
plt.subplot(2, 2, 1)
robot1.draw_polygon(3, 2)
xx, yy = zip(*robot1.history)
plt.plot(xx, yy)
plt.title("лбЛђлхЛЃл│лЙл╗ЛїлйлИл║ (n=3)")
plt.axis('equal')

robot2 = Robot("лџл▓л░л┤Лђл░Лѓ")
plt.subplot(2, 2, 2)
robot2.draw_polygon(4, 2)
xx, yy = zip(*robot2.history)
plt.plot(xx, yy)
plt.title("лџл▓л░л┤Лђл░Лѓ (n=4)")
plt.axis('equal')

robot3 = Robot("лЪЛЈЛѓлИЛЃл│лЙл╗ЛїлйлИл║")
plt.subplot(2, 2, 3)
robot3.draw_polygon(5, 2)
xx, yy = zip(*robot3.history)
plt.plot(xx, yy)
plt.title("лЪЛЈЛѓлИЛЃл│лЙл╗ЛїлйлИл║ (n=5)")
plt.axis('equal')

robot4 = Robot("лњлЙЛЂЛїл╝лИЛЃл│лЙл╗ЛїлйлИл║")
plt.subplot(2, 2, 4)
robot4.draw_polygon(8, 1.5)
xx, yy = zip(*robot4.history)
plt.plot(xx, yy)
plt.title("лњлЙЛЂЛїл╝лИЛЃл│лЙл╗ЛїлйлИл║ (n=8)")
plt.axis('equal')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 10))

robot5 = Robot("лАл║ЛђЛЃл│л╗лхлйлйЛІл╣ ЛѓЛђлхЛЃл│лЙл╗ЛїлйлИл║")
plt.subplot(2, 2, 1)
robot5.draw_rounded_polygon(3, 1.5, 0.3)
xx, yy = zip(*robot5.history)
plt.plot(xx, yy)
plt.title("лАл║ЛђЛЃл│л╗лхлйлйЛІл╣ ЛѓЛђлхЛЃл│лЙл╗ЛїлйлИл║")
plt.axis('equal')

robot6 = Robot("лАл║ЛђЛЃл│л╗лхлйлйЛІл╣ л║л▓л░л┤Лђл░Лѓ")
plt.subplot(2, 2, 2)
robot6.draw_rounded_polygon(4, 1.5, 0.3)
xx, yy = zip(*robot6.history)
plt.plot(xx, yy)
plt.title("лАл║ЛђЛЃл│л╗лхлйлйЛІл╣ л║л▓л░л┤Лђл░Лѓ")
plt.axis('equal')

robot7 = Robot("лАл║ЛђЛЃл│л╗лхлйлйЛІл╣ л┐ЛЈЛѓлИЛЃл│лЙл╗ЛїлйлИл║")
plt.subplot(2, 2, 3)
robot7.draw_rounded_polygon(5, 1.2, 0.2)
xx, yy = zip(*robot7.history)
plt.plot(xx, yy)
plt.title("лАл║ЛђЛЃл│л╗лхлйлйЛІл╣ л┐ЛЈЛѓлИЛЃл│лЙл╗ЛїлйлИл║")
plt.axis('equal')

robot8 = Robot("лАл║ЛђЛЃл│л╗лхлйлйЛІл╣ ЛѕлхЛЂЛѓлИЛЃл│лЙл╗ЛїлйлИл║")
plt.subplot(2, 2, 4)
robot8.draw_rounded_polygon(6, 1.0, 0.15)
xx, yy = zip(*robot8.history)
plt.plot(xx, yy)
plt.title("лАл║ЛђЛЃл│л╗лхлйлйЛІл╣ ЛѕлхЛЂЛѓлИЛЃл│лЙл╗ЛїлйлИл║")
plt.axis('equal')

plt.tight_layout()
plt.show()
```
