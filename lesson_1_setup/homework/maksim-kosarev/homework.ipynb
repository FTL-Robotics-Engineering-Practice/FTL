```python
# –ë–ª–æ–∫ 2.1: –ü–µ—Ä–≤–æ–µ –∑–Ω–∞–∫–æ–º—Å—Ç–≤–æ —Å –∫–ª–∞—Å—Å–∞–º–∏

# –ó–∞–¥–∞–Ω–∏–µ 1
# 1) –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å —Å –ø–æ–¥—Ö–æ–¥—è—â–∏–º –∏–º–µ–Ω–µ–º (–ø—Ä–∏–¥—É–º–∞–π—Ç–µ —Å–∞–º–∏)
# 2) –í –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä –ø–µ—Ä–µ–¥–∞–π—Ç–µ –≤—Ö–æ–¥–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã (–Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–º—è –∏ —Å—Ç–∞—Ä—Ç–æ–≤—ã–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã)
# 3) –°–æ–∑–¥–∞–π—Ç–µ –æ–¥–∏–Ω –æ–±—ä–µ–∫—Ç –∏ –ø—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ –ø–æ–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏–ª–∏—Å—å
# 4) –ò–∑–º–µ–Ω–∏—Ç–µ –ø–æ–ª—è —É –æ–±—ä–µ–∫—Ç–∞ –∏ –ø—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –ø—Ä–∏–º–µ–Ω–∏–ª–∏—Å—å

class Point:
    def __init__(self, name, x, y):
        self.name = name
        self.x = x
        self.y = y

# —Å–æ–∑–¥–∞–π—Ç–µ –æ–±—ä–µ–∫—Ç
obj = Point("A", 10, 2)

# –≤—ã–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ–ª–µ–π
print("–î–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è:", obj.name, obj.x, obj.y)

# –∏–∑–º–µ–Ω–∏—Ç–µ –ø–æ–ª—è
obj.name = "b"
obj.x = 3
obj.y = 7

# –ø—Ä–æ–≤–µ—Ä—å—Ç–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è
print("–ü–æ—Å–ª–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è:", obj.name, obj.x, obj.y)

print("\n---\n")

# –ó–∞–¥–∞–Ω–∏–µ 2
# 1) –°–æ–∑–¥–∞–π—Ç–µ –¥–≤–∞ –æ–±—ä–µ–∫—Ç–∞ —Ç–æ–≥–æ –∂–µ –∫–ª–∞—Å—Å–∞
# 2) "–ü–µ—Ä–µ–¥–≤–∏–Ω—å—Ç–µ" –∏—Ö, –∏–∑–º–µ–Ω–∏–≤ —á–∏—Å–ª–æ–≤—ã–µ –ø–æ–ª—è (–Ω–∞–ø—Ä–∏–º–µ—Ä, –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã)
# 3) –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ –∏—Ç–æ–≥–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã

obj1 = Point("A", 4, 3)
obj2 = Point("N", 6, 9)

# –¥–≤–∏–∂–µ–Ω–∏–µ: –∑–∞–¥–∞–π—Ç–µ –ø—Ä–∞–≤–∏–ª–∞ –∏ –ø—Ä–∏–º–µ–Ω–∏—Ç–µ
# –ø—Ä–∏–º–µ—Ä: obj1: +2, +1; obj2: -3, +4
obj1.x = obj1.x + 3
obj1.y = obj1.y + 5

obj2.x = obj2.x - 4
obj2.y = obj2.y + 8

print("obj1:", obj1.x, obj1.y)
print("obj2:", obj2.x, obj2.y)
```

```python
class Robot:
    """–†–æ–±–æ—Ç, –∫–æ—Ç–æ—Ä—ã–π —É–º–µ–µ—Ç —Ö–æ–¥–∏—Ç—å –ø–æ –∫–≤–∞–¥—Ä–∞—Ç—É"""
    
    def __init__(self, name, start_x=0, start_y=0):
        """–°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ —Å –∏–º–µ–Ω–µ–º –∏ —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –ø–æ–∑–∏—Ü–∏–µ–π"""
        self.name = name
        self.x = start_x
        self.y = start_y
        print(f"ü§ñ {name} —Å–æ–∑–¥–∞–Ω –≤ –ø–æ–∑–∏—Ü–∏–∏ ({start_x}, {start_y})")
    
    def info(self):
        """–ü–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"ü§ñ {self.name} –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ ({self.x}, {self.y})")
    
    def walk_square(self, size=3):
        """–†–æ–±–æ—Ç —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞"""
        print(f"üö∂ {self.name} –Ω–∞—á–∏–Ω–∞–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}")
        
        # –î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ (size —à–∞–≥–æ–≤)
        print(f"  {self.name} –∏–¥–µ—Ç –≤–ø—Ä–∞–≤–æ:")
        for step in range(size):
            self.x += 1
            print(f"    –®–∞–≥ {step+1}: {self.name} –≤ ({self.x}, {self.y})")
        
        # –î–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö (size —à–∞–≥–æ–≤)
        print(f"  {self.name} –∏–¥–µ—Ç –≤–≤–µ—Ä—Ö:")
        for step in range(size):
            self.y += 1
            print(f"    –®–∞–≥ {step+1}: {self.name} –≤ ({self.x}, {self.y})")
        
        # –î–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ (size —à–∞–≥–æ–≤)
        print(f"  {self.name} –∏–¥–µ—Ç –≤–ª–µ–≤–æ:")
        for step in range(size):
            self.x -= 1
            print(f"    –®–∞–≥ {step+1}: {self.name} –≤ ({self.x}, {self.y})")
        
        # –î–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑ (size —à–∞–≥–æ–≤)
        print(f"  {self.name} –∏–¥–µ—Ç –≤–Ω–∏–∑:")
        for step in range(size):
            self.y -= 1
            print(f"    –®–∞–≥ {step+1}: {self.name} –≤ ({self.x}, {self.y})")
        
        print(f"‚úÖ {self.name} –∑–∞–≤–µ—Ä—à–∏–ª –∫–≤–∞–¥—Ä–∞—Ç! –§–∏–Ω–∞–ª—å–Ω–∞—è –ø–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y})")


# üöÄ –¢–ï–°–¢: –û–¥–∏–Ω —Ä–æ–±–æ—Ç —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç
robot = Robot("–¢–µ—Å—Ç–µ—Ä", 0, 0)

# –ü–æ–∫–∞–∑–∞—Ç—å –Ω–∞—á–∞–ª—å–Ω—É—é –ø–æ–∑–∏—Ü–∏—é
robot.info()

# –ó–∞—Å—Ç–∞–≤–∏—Ç—å —Ä–æ–±–æ—Ç–∞ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç —Ä–∞–∑–º–µ—Ä–æ–º 2x2
robot.walk_square(2)

# –ü–æ–∫–∞–∑–∞—Ç—å —Ñ–∏–Ω–∞–ª—å–Ω—É—é –ø–æ–∑–∏—Ü–∏—é
robot.info()

# –ü—Ä–æ–≤–µ—Ä—è–µ–º: —Ä–æ–±–æ—Ç –¥–æ–ª–∂–µ–Ω –≤–µ—Ä–Ω—É—Ç—å—Å—è –≤ —Å—Ç–∞—Ä—Ç–æ–≤—É—é –ø–æ–∑–∏—Ü–∏—é
start_x, start_y = 0, 0
if robot.x == start_x and robot.y == start_y:
    print("‚úÖ –û—Ç–ª–∏—á–Ω–æ! –†–æ–±–æ—Ç –≤–µ—Ä–Ω—É–ª—Å—è –¥–æ–º–æ–π!")
else:
    print(f"‚ùå –†–æ–±–æ—Ç –∑–∞–±–ª—É–¥–∏–ª—Å—è! –û–∂–∏–¥–∞–ª–æ—Å—å ({start_x}, {start_y}), –ø–æ–ª—É—á–∏–ª–∏ ({robot.x}, {robot.y})")

print("\nüéâ –ü–æ–∑–¥—Ä–∞–≤–ª—è–µ–º! –í—ã —Å–æ–∑–¥–∞–ª–∏ —Ä–æ–±–æ—Ç–∞ —Å –º–µ—Ç–æ–¥–æ–º!")
```

```python
# –ë–ª–æ–∫ 2.3.1: –ö—Ä–∞—Å–∏–≤—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ä–æ–±–æ—Ç–µ

class Robot:
    """–†–æ–±–æ—Ç —Å –∫—Ä–∞—Å–∏–≤—ã–º –≤—ã–≤–æ–¥–æ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏"""
    
    def __init__(self, name, start_x=0, start_y=0):
        """–°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ —Å –∏–º–µ–Ω–µ–º –∏ —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –ø–æ–∑–∏—Ü–∏–µ–π"""
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x  # –∑–∞–ø–æ–º–∏–Ω–∞–µ–º –¥–æ–º
        self.start_y = start_y
        self.steps_made = 0     # —Å—á–µ—Ç—á–∏–∫ —à–∞–≥–æ–≤
        print(f"ü§ñ {name} —Å–æ–∑–¥–∞–Ω –≤ –ø–æ–∑–∏—Ü–∏–∏ ({start_x}, {start_y})")
    
    def show_info(self):
        """–ö—Ä–∞—Å–∏–≤—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"‚îå{'‚îÄ' * 40}‚îê")
        print(f"‚îÇ ü§ñ –†–æ–±–æ—Ç: {self.name:<30} ‚îÇ")
        print(f"‚îÇ üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y}){' ' * (22 - len(str(self.x)) - len(str(self.y)))}‚îÇ")
        print(f"‚îÇ üè† –î–æ–º: ({self.start_x}, {self.start_y}){' ' * (26 - len(str(self.start_x)) - len(str(self.start_y)))}‚îÇ")
        print(f"‚îÇ üë£ –®–∞–≥–æ–≤ —Å–¥–µ–ª–∞–Ω–æ: {self.steps_made:<18} ‚îÇ")
        print(f"‚îî{'‚îÄ' * 40}‚îò")
    
    def move_right(self):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ"""
        self.x += 1
        self.steps_made += 1
        print(f"    üë£ {self.name} –≤ ({self.x}, {self.y})")

# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ –∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
print("üè≠ –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞!")
robot = Robot("–ê–ª—å—Ñ–∞", 5, 3)

print("\nüìä –ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –†–û–ë–û–¢–ï:")
# TODO: –í—ã–∑–æ–≤–∏—Ç–µ –º–µ—Ç–æ–¥ show_info() –¥–ª—è —Ä–æ–±–æ—Ç–∞
robot.show_info()

print("\nüö∂ –†–æ–±–æ—Ç –¥–µ–ª–∞–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ —à–∞–≥–æ–≤:")
robot.move_right()
robot.move_right()

print("\nüìä –û–ë–ù–û–í–õ–ï–ù–ù–ê–Ø –ò–ù–§–û–†–ú–ê–¶–ò–Ø:")
# TODO: –°–Ω–æ–≤–∞ –ø–æ–∫–∞–∂–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ä–æ–±–æ—Ç–µ
robot.show_info()

print("\n‚úÖ –û—Ç–ª–∏—á–Ω–æ! –¢–µ–ø–µ—Ä—å —É –Ω–∞—Å –µ—Å—Ç—å –∫—Ä–∞—Å–∏–≤—ã–π —Å–ø–æ—Å–æ–± –ø–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ä–æ–±–æ—Ç–µ!")
```

```python
# –ë–ª–æ–∫ 2.3.2: –û—Ç–¥–µ–ª—å–Ω—ã–µ –º–µ—Ç–æ–¥—ã –¥–≤–∏–∂–µ–Ω–∏—è

class Robot:
    """–†–æ–±–æ—Ç —Å –æ—Ç–¥–µ–ª—å–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏ –¥–≤–∏–∂–µ–Ω–∏—è"""
    
    def __init__(self, name, start_x=0, start_y=0):
        """–°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ —Å –∏–º–µ–Ω–µ–º –∏ —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –ø–æ–∑–∏—Ü–∏–µ–π"""
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.steps_made = 0
        print(f"ü§ñ {name} —Å–æ–∑–¥–∞–Ω –≤ –ø–æ–∑–∏—Ü–∏–∏ ({start_x}, {start_y})")
    
    def show_info(self):
        """–ö—Ä–∞—Å–∏–≤—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"‚îå{'‚îÄ' * 40}‚îê")
        print(f"‚îÇ ü§ñ –†–æ–±–æ—Ç: {self.name:<30} ‚îÇ")
        print(f"‚îÇ üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y}){' ' * (22 - len(str(self.x)) - len(str(self.y)))}‚îÇ")
        print(f"‚îÇ üè† –î–æ–º: ({self.start_x}, {self.start_y}){' ' * (26 - len(str(self.start_x)) - len(str(self.start_y)))}‚îÇ")
        print(f"‚îÇ üë£ –®–∞–≥–æ–≤ —Å–¥–µ–ª–∞–Ω–æ: {self.steps_made:<18} ‚îÇ")
        print(f"‚îî{'‚îÄ' * 40}‚îò")
    
    def move_right(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        # TODO: –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ steps > 0
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        # TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ
        for step in range(steps):
            self.x += 1
            self.steps_made += 1
            print(f"    ‚û°Ô∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_left(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        # TODO: –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ steps > 0
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        # TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ
        for step in range(steps):
            self.x -= 1
            self.steps_made += 1
            print(f"    ‚¨ÖÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_up(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        # TODO: –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ steps > 0
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        # TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö
        for step in range(steps):
            self.y += 1
            self.steps_made += 1
            print(f"    ‚¨ÜÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_down(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        # TODO: –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ steps > 0
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        # TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑
        for step in range(steps):
            self.y -= 1
            self.steps_made += 1
            print(f"    ‚¨áÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")

# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞
print("üè≠ –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞!")
robot = Robot("–ë–µ—Ç–∞", 0, 0)

print("\nüìä –ù–ê–ß–ê–õ–¨–ù–û–ï –°–û–°–¢–û–Ø–ù–ò–ï:")
robot.show_info()

print("\nüéÆ –£–ü–†–ê–í–õ–Ø–ï–ú –†–û–ë–û–¢–û–ú –û–¢–î–ï–õ–¨–ù–´–ú–ò –ú–ï–¢–û–î–ê–ú–ò:")
# TODO: –ó–∞—Å—Ç–∞–≤—å—Ç–µ —Ä–æ–±–æ—Ç–∞ –¥–≤–∏–≥–∞—Ç—å—Å—è —Ä–∞–∑–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏
robot.move_right(2)  # 2 —à–∞–≥–∞ –≤–ø—Ä–∞–≤–æ
robot.move_up(3)     # 3 —à–∞–≥–∞ –≤–≤–µ—Ä—Ö
robot.move_left(1)   # 1 —à–∞–≥ –≤–ª–µ–≤–æ
robot.move_down(2)   # 2 —à–∞–≥–∞ –≤–Ω–∏–∑

print("\nüß™ –¢–ï–°–¢–ò–†–£–ï–ú –û–®–ò–ë–ö–ò:")
# TODO: –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —à–∞–≥–∏ - —á—Ç–æ –ø—Ä–æ–∏–∑–æ–π–¥–µ—Ç?
robot.move_right(-3)

print("\nüìä –§–ò–ù–ê–õ–¨–ù–û–ï –°–û–°–¢–û–Ø–ù–ò–ï:")
robot.show_info()

print("\n‚úÖ –û—Ç–ª–∏—á–Ω–æ! –¢–µ–ø–µ—Ä—å –º—ã –º–æ–∂–µ–º —É–ø—Ä–∞–≤–ª—è—Ç—å —Ä–æ–±–æ—Ç–æ–º –æ—Ç–¥–µ–ª—å–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏!")
print("üí° –ù–µ –∑–∞–±—É–¥—å—Ç–µ –¥–æ–±–∞–≤–∏—Ç—å –ø—Ä–æ–≤–µ—Ä–∫—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —à–∞–≥–æ–≤!")
```

```python
# –ë–ª–æ–∫ 2.3.4: –¢—Ä–∏ —Ä–æ–±–æ—Ç–∞ —Ä–∏—Å—É—é—Ç –∫–≤–∞–¥—Ä–∞—Ç—ã –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ

class Robot:
    """–ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–π —Ä–æ–±–æ—Ç —Å –∫—Ä–∞—Å–∏–≤—ã–º –≤—ã–≤–æ–¥–æ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏"""
    
    def __init__(self, name, start_x=0, start_y=0):
        """–°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ —Å –∏–º–µ–Ω–µ–º –∏ —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –ø–æ–∑–∏—Ü–∏–µ–π"""
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x  # –∑–∞–ø–æ–º–∏–Ω–∞–µ–º –¥–æ–º
        self.start_y = start_y
        self.steps_made = 0     # —Å—á–µ—Ç—á–∏–∫ —à–∞–≥–æ–≤
        print(f"ü§ñ {name} —Å–æ–∑–¥–∞–Ω –≤ –ø–æ–∑–∏—Ü–∏–∏ ({start_x}, {start_y})")
    
    def show_info(self):
        """–ö—Ä–∞—Å–∏–≤—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"‚îå{'‚îÄ' * 40}‚îê")
        print(f"‚îÇ ü§ñ –†–æ–±–æ—Ç: {self.name:<30} ‚îÇ")
        print(f"‚îÇ üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y}){' ' * (22 - len(str(self.x)) - len(str(self.y)))}‚îÇ")
        print(f"‚îÇ üè† –î–æ–º: ({self.start_x}, {self.start_y}){' ' * (26 - len(str(self.start_x)) - len(str(self.start_y)))}‚îÇ")
        print(f"‚îÇ üë£ –®–∞–≥–æ–≤ —Å–¥–µ–ª–∞–Ω–æ: {self.steps_made:<18} ‚îÇ")
        print(f"‚îî{'‚îÄ' * 40}‚îò")
    
    def move_right(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.x += 1
            self.steps_made += 1
            print(f"    ‚û°Ô∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_left(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.x -= 1
            self.steps_made += 1
            print(f"    ‚¨ÖÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_up(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.y += 1
            self.steps_made += 1
            print(f"    ‚¨ÜÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_down(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.y -= 1
            self.steps_made += 1
            print(f"    ‚¨áÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def walk_square(self, size=3):
        """–†–æ–±–æ—Ç —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞"""
        print(f"\nüé® {self.name} –Ω–∞—á–∏–Ω–∞–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}")
        
        # –í–ø—Ä–∞–≤–æ
        print(f"  ‚û°Ô∏è –ò–¥–µ–º –≤–ø—Ä–∞–≤–æ ({size} —à–∞–≥–æ–≤):")
        self.move_right(3)
        
        # –í–≤–µ—Ä—Ö
        print(f"  ‚¨ÜÔ∏è –ò–¥–µ–º –≤–≤–µ—Ä—Ö ({size} —à–∞–≥–æ–≤):")
        self.move_up(3)
        
        # –í–ª–µ–≤–æ
        print(f"  ‚¨ÖÔ∏è –ò–¥–µ–º –≤–ª–µ–≤–æ ({size} —à–∞–≥–æ–≤):")
        self.move_left(3)
        
        # –í–Ω–∏–∑
        print(f"  ‚¨áÔ∏è –ò–¥–µ–º –≤–Ω–∏–∑ ({size} —à–∞–≥–æ–≤):")
        self.move_down(3)
        
        print(f"‚úÖ {self.name} –∑–∞–≤–µ—Ä—à–∏–ª –∫–≤–∞–¥—Ä–∞—Ç!")

# –°–æ–∑–¥–∞–µ–º —Ç—Ä–µ—Ö —Ä–æ–±–æ—Ç–æ–≤ –≤ —Ä–∞–∑–Ω—ã—Ö –ø–æ–∑–∏—Ü–∏—è—Ö
print("üè≠ –ó–∞–ø—É—Å–∫–∞–µ–º —Ñ–∞–±—Ä–∏–∫—É —Ä–æ–±–æ—Ç–æ–≤!")

# TODO: –°–æ–∑–¥–∞–π—Ç–µ —Ç—Ä–µ—Ö —Ä–æ–±–æ—Ç–æ–≤
# –†–æ–±–æ—Ç 1: "–ê–ª—å—Ñ–∞" –≤ –ø–æ–∑–∏—Ü–∏–∏ (0, 0)
robot1 = Robot("a", 0, 0)

# –†–æ–±–æ—Ç 2: "–ë–µ—Ç–∞" –≤ –ø–æ–∑–∏—Ü–∏–∏ (10, 5)  
robot2 = Robot("b", 10, 5)

# –†–æ–±–æ—Ç 3: "–ì–∞–º–º–∞" –≤ –ø–æ–∑–∏—Ü–∏–∏ (20, 10)
robot3 = Robot("c", 20, 10)

# –ü–æ–∫–∞–∑–∞—Ç—å –Ω–∞—á–∞–ª—å–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
print("\nüìä –ù–ê–ß–ê–õ–¨–ù–û–ï –°–û–°–¢–û–Ø–ù–ò–ï –†–û–ë–û–¢–û–í:")
# TODO: –í—ã–∑–≤–∞—Ç—å show_info() –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–æ–±–æ—Ç–∞
robot1.show_info()
robot2.show_info()
robot3.show_info()

# –ó–∞–ø—É—Å—Ç–∏—Ç—å –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤ —Ä–∏—Å–æ–≤–∞—Ç—å
print("\nüé® –í–°–ï –†–û–ë–û–¢–´ –ù–ê–ß–ò–ù–ê–Æ–¢ –†–ò–°–û–í–ê–¢–¨:")

# TODO: –ó–∞—Å—Ç–∞–≤–∏—Ç—å –∫–∞–∂–¥–æ–≥–æ —Ä–æ–±–æ—Ç–∞ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç —Ä–∞–∑–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
# –ê–ª—å—Ñ–∞ —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç 2x2
robot1.walk_square(2)

# –ë–µ—Ç–∞ —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç 3x3
robot2.walk_square(3)

# –ì–∞–º–º–∞ —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç 4x4  
robot3.walk_square(4)

# –ü–æ–∫–∞–∑–∞—Ç—å —Ñ–∏–Ω–∞–ª—å–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
print("\nüìä –§–ò–ù–ê–õ–¨–ù–û–ï –°–û–°–¢–û–Ø–ù–ò–ï –†–û–ë–û–¢–û–í:")
# TODO: –°–Ω–æ–≤–∞ –ø–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–∞—Ö
robot1.show_info()
robot2.show_info()
robot3.show_info()

# –ü—Ä–æ–≤–µ—Ä–∫–∞: –∫—Ç–æ –≤–µ—Ä–Ω—É–ª—Å—è –¥–æ–º–æ–π?
print("\nüè† –ü–†–û–í–ï–†–ö–ê: –ö–¢–û –í–ï–†–ù–£–õ–°–Ø –î–û–ú–û–ô?")
robots = [robot1, robot2, robot3]
for robot in robots:
    if robot.x == robot.start_x and robot.y == robot.start_y:
        print(f"‚úÖ {robot.name} –≤–µ—Ä–Ω—É–ª—Å—è –¥–æ–º–æ–π!")
    else:
        print(f"‚ùå {robot.name} –∑–∞–±–ª—É–¥–∏–ª—Å—è!")

print(f"\nüéâ –ó–∞–¥–∞–Ω–∏–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ! –ú—ã —É–ø—Ä–∞–≤–ª—è–ª–∏ {len(robots)} —Ä–æ–±–æ—Ç–∞–º–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ!")
```

```python
class Robot:
    """–†–æ–±–æ—Ç –¥–ª—è –º–∞—Å—Å–æ–≤–æ–≥–æ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        print(f"ü§ñ {name} ({color}) —Å–æ–∑–¥–∞–Ω –≤ ({start_x}, {start_y})")
    
    def show_info(self):
        """–ö–æ–º–ø–∞–∫—Ç–Ω—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏"""
        status = "üè†" if (self.x == self.start_x and self.y == self.start_y) else "üó∫Ô∏è"
        print(f"{status} {self.name:<10} | –ü–æ–∑–∏—Ü–∏—è: ({self.x:2}, {self.y:2}) | –®–∞–≥–æ–≤: {self.steps_made:2} | –¶–≤–µ—Ç: {self.color}")
    
    def walk_square(self, size=3):
        """–ë—ã—Å—Ç—Ä–æ–µ —Ä–∏—Å–æ–≤–∞–Ω–∏–µ –∫–≤–∞–¥—Ä–∞—Ç–∞ (–±–µ–∑ –¥–µ—Ç–∞–ª—å–Ω–æ–≥–æ –≤—ã–≤–æ–¥–∞)"""
        directions = ["right", "up", "left", "down"]
        
        for direction in directions:
            for step in range(size):
                if direction == "right":
                    self.x += 1
                elif direction == "up":
                    self.y += 1
                elif direction == "left":
                    self.x -= 1
                elif direction == "down":
                    self.y -= 1
                self.steps_made += 1
        
        print(f"‚úÖ {self.name} –∑–∞–≤–µ—Ä—à–∏–ª –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}")

# –î–∞–Ω–Ω—ã–µ –¥–ª—è —Ä–æ–±–æ—Ç–æ–≤
robots_data = [
    {"name": "A", "x": 0, "y": 0, "color": "blue"},
    {"name": "B", "x": 1, "y": 1, "color": "red"},
    {"name": "C", "x": 2, "y": 2, "color": "orange"},
]

print("üè≠ –§–ê–ë–†–ò–ö–ê –†–û–ë–û–¢–û–í –ó–ê–ü–£–©–ï–ù–ê!")
print(f"üìã –í –æ—á–µ—Ä–µ–¥–∏ –Ω–∞ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–æ: {len(robots_data)} —Ä–æ–±–æ—Ç–æ–≤")

# –°–æ–∑–¥–∞—ë–º –∞—Ä–º–∏—é —Ä–æ–±–æ—Ç–æ–≤
robot_army = []

print("\nüîß –ü–†–û–ò–ó–í–û–î–°–¢–í–û –†–û–ë–û–¢–û–í:")
for robot_data in robots_data:
    robot = Robot(
        robot_data["name"],
        robot_data["x"],
        robot_data["y"],
        robot_data["color"]
    )
    robot_army.append(robot)

print(f"\n‚úÖ –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–æ —Ä–æ–±–æ—Ç–æ–≤: {len(robot_army)}")

# –ü–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–∞—Ö
print("\nüìä –†–ï–ï–°–¢–† –†–û–ë–û–¢–û–í:")
print("   –°—Ç–∞—Ç—É—Å | –ò–º—è        | –ü–æ–∑–∏—Ü–∏—è    | –®–∞–≥–æ–≤ | –¶–≤–µ—Ç")
print("   " + "‚îÄ" * 50)
for robot in robot_army:
    robot.show_info()

# –ó–∞–ø—É—Å—Ç–∏—Ç—å –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤ —Ä–∞–±–æ—Ç–∞—Ç—å
print("\nüé® –ú–ê–°–°–û–í–û–ï –ü–†–û–ò–ó–í–û–î–°–¢–í–û –ö–í–ê–î–†–ê–¢–û–í:")
sizes = [4, 3, 2]

for i, robot in enumerate(robot_army):
    size = sizes[i]
    print(f"ü§ñ {robot.name} —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}...")
    robot.walk_square(size)

# –§–∏–Ω–∞–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
print("\nüìä –§–ò–ù–ê–õ–¨–ù–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê:")
print("   –°—Ç–∞—Ç—É—Å | –ò–º—è        | –ü–æ–∑–∏—Ü–∏—è    | –®–∞–≥–æ–≤ | –¶–≤–µ—Ç") 
print("   " + "‚îÄ" * 50)
for robot in robot_army:
    robot.show_info()

# –ê–Ω–∞–ª–∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤  
print(f"\nüìà –ê–ù–ê–õ–ò–ó –ü–†–û–ò–ó–í–û–î–°–¢–í–ê:")
total_steps = 0
robots_at_home = 0

for robot in robot_army:
    total_steps += robot.steps_made
    
    if robot.x == robot.start_x and robot.y == robot.start_y:
        robots_at_home += 1

print(f"   ü§ñ –í—Å–µ–≥–æ —Ä–æ–±–æ—Ç–æ–≤: {len(robot_army)}")
print(f"   üë£ –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤: {total_steps}")
print(f"   üìä –°—Ä–µ–¥–Ω–µ–µ —à–∞–≥–æ–≤ –Ω–∞ —Ä–æ–±–æ—Ç–∞: {total_steps / len(robot_army):.1f}")
print(f"   üè† –†–æ–±–æ—Ç–æ–≤ –≤–µ—Ä–Ω—É–ª–æ—Å—å –¥–æ–º–æ–π: {robots_at_home}")
print(f"   üìè –≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –≤–æ–∑–≤—Ä–∞—Ç–∞: {robots_at_home / len(robot_army) * 100:.1f}%")

most_active = max(robot_army, key=lambda r: r.steps_made)
print(f"   üèÜ –°–∞–º—ã–π –∞–∫—Ç–∏–≤–Ω—ã–π: {most_active.name} ({most_active.steps_made} —à–∞–≥–æ–≤)")

print(f"\nüéâ –§–ê–ë–†–ò–ö–ê –ó–ê–í–ï–†–®–ò–õ–ê –†–ê–ë–û–¢–£! –í—Å–µ —Ä–æ–±–æ—Ç—ã –≥–æ—Ç–æ–≤—ã –∫ —Å–ª—É–∂–±–µ!")
```

```python
import matplotlib.pyplot as plt
import numpy as np

class Robot:
    """–†–æ–±–æ—Ç, –∫–æ—Ç–æ—Ä—ã–π —É–º–µ–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å —Å–µ–±—è –Ω–∞ —ç–∫—Ä–∞–Ω–µ"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        print(f"ü§ñ {name} ({color}) —Å–æ–∑–¥–∞–Ω –≤ ({start_x}, {start_y})")
    
    def draw(self, size=100):
        """–†–∏—Å—É–µ—Ç —Ä–æ–±–æ—Ç–∞ –Ω–∞ —Ç–µ–∫—É—â–µ–º –≥—Ä–∞—Ñ–∏–∫–µ"""
        # –†–∏—Å—É–µ–º —Ä–æ–±–æ—Ç–∞ –∫–∞–∫ —Ç–æ—á–∫—É
        plt.scatter(self.x, self.y, 
                   color=self.color, 
                   s=size, 
                   alpha=0.8,
                   label=self.name)
        
        # –ü–æ–¥–ø–∏—Å—å —Å –∏–º–µ–Ω–µ–º –∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏
        plt.annotate(f'{self.name}\n({self.x},{self.y})', 
                    (self.x, self.y), 
                    xytext=(10, 10), 
                    textcoords='offset points',
                    fontsize=8,
                    bbox=dict(boxstyle="round,pad=0.3", 
                             facecolor=self.color, 
                             alpha=0.3))
    
    def move_right(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.x += 1
            self.steps_made += 1
            print(f"    ‚û°Ô∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_left(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.x -= 1
            self.steps_made += 1
            print(f"    ‚¨ÖÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_up(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.y += 1
            self.steps_made += 1
            print(f"    ‚¨ÜÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_down(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.y -= 1
            self.steps_made += 1
            print(f"    ‚¨áÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def walk_square(self, size=3):
        """–†–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç"""
        print(f"\nüé® {self.name} –Ω–∞—á–∏–Ω–∞–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}")
        
        print(f"  ‚û°Ô∏è –ò–¥–µ–º –≤–ø—Ä–∞–≤–æ ({size} —à–∞–≥–æ–≤):")
        self.move_right(size)
        
        print(f"  ‚¨ÜÔ∏è –ò–¥–µ–º –≤–≤–µ—Ä—Ö ({size} —à–∞–≥–æ–≤):")
        self.move_up(size)
        
        print(f"  ‚¨ÖÔ∏è –ò–¥–µ–º –≤–ª–µ–≤–æ ({size} —à–∞–≥–æ–≤):")
        self.move_left(size)
        
        print(f"  ‚¨áÔ∏è –ò–¥–µ–º –≤–Ω–∏–∑ ({size} —à–∞–≥–æ–≤):")
        self.move_down(size)
        
        print(f"‚úÖ {self.name} –∑–∞–≤–µ—Ä—à–∏–ª –∫–≤–∞–¥—Ä–∞—Ç!")

# üîπ –¢–µ—Å—Ç: —Ä–∏—Å—É–µ–º –æ–¥–Ω–æ–≥–æ —Ä–æ–±–æ—Ç–∞
print("üé® –¢–ï–°–¢: –†–∏—Å–æ–≤–∞–Ω–∏–µ –æ–¥–Ω–æ–≥–æ —Ä–æ–±–æ—Ç–∞")

test_robot = Robot("Alpha", 2, 3, "blue")

plt.figure(figsize=(8, 6))
test_robot.draw()

plt.grid(True, alpha=0.3)
plt.xlim(-2, 10)
plt.ylim(-2, 8)
plt.title("ü§ñ –ú–æ–π –ø–µ—Ä–≤—ã–π —Ä–æ–±–æ—Ç –Ω–∞ —ç–∫—Ä–∞–Ω–µ!")
plt.xlabel("X –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.ylabel("Y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.legend()
plt.show()

print("‚úÖ –¢–µ—Å—Ç –∑–∞–≤–µ—Ä—à–µ–Ω! –†–æ–±–æ—Ç –Ω–∞—Ä–∏—Å–æ–≤–∞–Ω!")

# üîπ –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ: –≤—Ç–æ—Ä–æ–π —Ä–æ–±–æ—Ç
print("\nüé® –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–û–ï –ó–ê–î–ê–ù–ò–ï:")

second_robot = Robot("Beta", 5, 6, "red")

plt.figure(figsize=(8, 6))
second_robot.draw(size=300)

plt.grid(True, alpha=0.3)
plt.xlim(-2, 10)
plt.ylim(-2, 8)
plt.title("ü§ñü§ñ –î–≤–∞ —Ä–æ–±–æ—Ç–∞ –Ω–∞ —ç–∫—Ä–∞–Ω–µ!")
plt.xlabel("X –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.ylabel("Y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.legend()
plt.show()

print("üéâ –û—Ç–ª–∏—á–Ω–æ! –¢–µ–ø–µ—Ä—å –≤—ã —É–º–µ–µ—Ç–µ —Ä–∏—Å–æ–≤–∞—Ç—å —Ä–æ–±–æ—Ç–æ–≤!")
```

```python
# –ë–ª–æ–∫ 3.2.3: –†–æ–±–æ—Ç—ã –¥–æ –∏ –ø–æ—Å–ª–µ —Ä–∞–±–æ—Ç—ã
# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º, –∫–∞–∫ —Ä–æ–±–æ—Ç—ã –∏–∑–º–µ–Ω—è—é—Ç—Å—è!

import matplotlib.pyplot as plt
import numpy as np

class Robot:
    """–†–æ–±–æ—Ç, –∫–æ—Ç–æ—Ä—ã–π —É–º–µ–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å —Å–µ–±—è –Ω–∞ —ç–∫—Ä–∞–Ω–µ"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        self.path = [(start_x, start_y)]  # —Å–æ—Ö—Ä–∞–Ω—è–µ–º —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏—é
        print(f"ü§ñ {name} ({color}) —Å–æ–∑–¥–∞–Ω –≤ ({start_x}, {start_y})")
    
    def draw(self, size=100):
        """–†–∏—Å—É–µ—Ç —Ä–æ–±–æ—Ç–∞ –Ω–∞ —Ç–µ–∫—É—â–µ–º –≥—Ä–∞—Ñ–∏–∫–µ"""
        plt.scatter(self.x, self.y, 
                   color=self.color, 
                   s=size, 
                   alpha=0.8,
                   label=self.name)
        
        plt.annotate(f'{self.name}\n({self.x},{self.y})', 
                    (self.x, self.y), 
                    xytext=(10, 10), 
                    textcoords='offset points',
                    fontsize=8,
                    bbox=dict(boxstyle="round,pad=0.3", 
                             facecolor=self.color, 
                             alpha=0.3))
    
    def move_right(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.x += 1
            self.steps_made += 1
            self.path.append((self.x, self.y))
            print(f"    ‚û°Ô∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")

# –¢–µ—Å—Ç: –†–æ–±–æ—Ç—ã –¥–æ –∏ –ø–æ—Å–ª–µ —Ä–∞–±–æ—Ç—ã
print("üé® –¢–ï–°–¢: –†–æ–±–æ—Ç—ã –¥–æ –∏ –ø–æ—Å–ª–µ —Ä–∞–±–æ—Ç—ã")

# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–æ–≤ –¥–ª—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞
work_robots = [
    Robot("–†–∞–±–æ—á–∏–π-1", 2, 2, "purple"),
    Robot("–†–∞–±–æ—á–∏–π-2", 8, 3, "cyan"),
    Robot("–†–∞–±–æ—á–∏–π-3", 5, 7, "magenta")
]

# –ì—Ä–∞—Ñ–∏–∫ –î–û —Ä–∞–±–æ—Ç—ã
plt.figure(figsize=(15, 5))

# –ü–æ–¥–≥—Ä–∞—Ñ–∏–∫ 1: –î–û
plt.subplot(1, 2, 1)
for robot in work_robots:
    robot.draw()

plt.grid(True, alpha=0.3)
plt.xlim(0, 12)
plt.ylim(0, 12)
plt.title("–î–û —Ä–∞–±–æ—Ç—ã")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# –ó–∞—Å—Ç–∞–≤–ª—è–µ–º —Ä–æ–±–æ—Ç–æ–≤ —Ä–∞–±–æ—Ç–∞—Ç—å
steps = [2, 3, 4]
for i, robot in enumerate(work_robots):
    robot.move_right(steps[i])
    print(f"‚úÖ {robot.name} –ø—Ä–æ—à–µ–ª {steps[i]} —à–∞–≥–æ–≤ –≤–ø—Ä–∞–≤–æ")

# –ü–æ–¥–≥—Ä–∞—Ñ–∏–∫ 2: –ü–û–°–õ–ï
plt.subplot(1, 2, 2)

# ‚úÖ –†–∏—Å—É–µ–º —Ä–æ–±–æ—Ç–æ–≤ –ø–æ—Å–ª–µ —Ä–∞–±–æ—Ç—ã
for robot in work_robots:
    robot.draw()

plt.grid(True, alpha=0.3)
plt.xlim(0, 12)
plt.ylim(0, 12)
plt.title("–ü–û–°–õ–ï —Ä–∞–±–æ—Ç—ã")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.tight_layout()
plt.show()

print("‚úÖ –¢–µ—Å—Ç –∑–∞–≤–µ—Ä—à–µ–Ω! –í–∏–¥–Ω–æ, –∫–∞–∫ —Ä–æ–±–æ—Ç—ã –ø–æ—Ä–∞–±–æ—Ç–∞–ª–∏!")

print("\nüéâ –ú–æ–¥—É–ª—å –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –æ—Å–≤–æ–µ–Ω! –†–æ–±–æ—Ç—ã —Ç–µ–ø–µ—Ä—å –≤–∏–¥–Ω—ã!")
```

```python
# –ë–ª–æ–∫ 3.3.1: –ü–µ—Ä–µ–≤–æ–¥—á–∏–∫ –∫–æ–º–∞–Ω–¥
# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–Ω–∏–º–∞–µ—Ç —Å–ª–æ–≤–∞!

class AdvancedRobot:
    """–ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–π —Ä–æ–±–æ—Ç —Å —É–º–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        self.path = [(start_x, start_y)]
        self.commands_executed = []  # –∏—Å—Ç–æ—Ä–∏—è –∫–æ–º–∞–Ω–¥
        print(f"ü§ñ {name} –≥–æ—Ç–æ–≤ –∫ –ø—Ä–∏–µ–º—É –∫–æ–º–∞–Ω–¥!")
    
    # –ß–µ—Ç—ã—Ä–µ –º–µ—Ç–æ–¥–∞ –¥–≤–∏–∂–µ–Ω–∏—è
    def move_up(self):
        """–®–∞–≥ –≤–≤–µ—Ä—Ö"""
        self.y += 1   # —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É Y
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨ÜÔ∏è {self.name} –∏–¥–µ—Ç –≤–≤–µ—Ä—Ö ‚Üí ({self.x}, {self.y})")
    
    def move_down(self):
        """–®–∞–≥ –≤–Ω–∏–∑"""
        self.y -= 1   # —É–º–µ–Ω—å—à–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É Y
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨áÔ∏è {self.name} –∏–¥–µ—Ç –≤–Ω–∏–∑ ‚Üí ({self.x}, {self.y})")
    
    def move_left(self):
        """–®–∞–≥ –≤–ª–µ–≤–æ"""
        self.x -= 1   # —É–º–µ–Ω—å—à–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É X
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨ÖÔ∏è {self.name} –∏–¥–µ—Ç –≤–ª–µ–≤–æ ‚Üí ({self.x}, {self.y})")
    
    def move_right(self):
        """–®–∞–≥ –≤–ø—Ä–∞–≤–æ"""
        self.x += 1   # —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É X
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚û°Ô∏è {self.name} –∏–¥–µ—Ç –≤–ø—Ä–∞–≤–æ ‚Üí ({self.x}, {self.y})")
    
    def execute_command(self, command):
        """–ü–µ—Ä–µ–≤–æ–¥—á–∏–∫: –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç —Å–ª–æ–≤–∞ –≤ –¥–µ–π—Å—Ç–≤–∏—è"""
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∫–æ–º–∞–Ω–¥—É –≤ –∏—Å—Ç–æ—Ä–∏—é
        self.commands_executed.append(command)
        
        if command == "up":
            self.move_up()
        elif command == "down":
            self.move_down()
        elif command == "left":
            self.move_left()
        elif command == "right":
            self.move_right()
        else:
            # –æ–±—Ä–∞–±–æ—Ç–∫–∞ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–π –∫–æ–º–∞–Ω–¥—ã
            print(f"‚ùì {self.name}: –ù–µ –∑–Ω–∞—é –∫–æ–º–∞–Ω–¥—É '{command}'!")
            print(f"   –î–æ—Å—Ç—É–ø–Ω—ã–µ –∫–æ–º–∞–Ω–¥—ã: up, down, left, right")
    
    def show_info(self):
        """–ü–æ–¥—Ä–æ–±–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"üìä {self.name}:")
        print(f"   üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y})")
        print(f"   üè† –î–æ–º: ({self.start_x}, {self.start_y})")
        print(f"   üë£ –®–∞–≥–æ–≤: {self.steps_made}")
        print(f"   üìú –ö–æ–º–∞–Ω–¥ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ: {len(self.commands_executed)}")
        if self.commands_executed:
            print(f"   üî§ –ü–æ—Å–ª–µ–¥–Ω–∏–µ –∫–æ–º–∞–Ω–¥—ã: {self.commands_executed[-5:]}")

# –¢–ï–°–¢ 1: –û—Å–Ω–æ–≤–Ω—ã–µ –¥–≤–∏–∂–µ–Ω–∏—è
print("üß™ –¢–ï–°–¢ 1: –ü—Ä–æ–≤–µ—Ä—è–µ–º –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–≤–∏–∂–µ–Ω–∏—è")

# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
test_robot = AdvancedRobot("–¢–µ—Å—Ç–µ—Ä", 0, 0, "red")

# –¢–µ—Å—Ç–∏—Ä—É–µ–º –∫–∞–∂–¥–æ–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ
print("\n–¢–µ—Å—Ç–∏—Ä—É–µ–º –≤—Å–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è:")
test_robot.move_up()
test_robot.move_right()
test_robot.move_down()
test_robot.move_left()

test_robot.show_info()

# –¢–ï–°–¢ 2: –ü–µ—Ä–µ–≤–æ–¥—á–∏–∫ –∫–æ–º–∞–Ω–¥
print("\nüß™ –¢–ï–°–¢ 2: –¢–µ—Å—Ç–∏—Ä—É–µ–º –ø–µ—Ä–µ–≤–æ–¥—á–∏–∫ –∫–æ–º–∞–Ω–¥")

# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞-–ø–µ—Ä–µ–≤–æ–¥—á–∏–∫–∞
translator_robot = AdvancedRobot("–ü–µ—Ä–µ–≤–æ–¥—á–∏–∫", 1, 1, "blue")

# –¢–µ—Å—Ç–∏—Ä—É–µ–º –∫–æ–º–∞–Ω–¥—ã —Å–ª–æ–≤–∞–º–∏
print("\n–¢–µ—Å—Ç–∏—Ä—É–µ–º –∫–æ–º–∞–Ω–¥—ã —Å–ª–æ–≤–∞–º–∏:")
translator_robot.execute_command("up")    
translator_robot.execute_command("right") 
translator_robot.execute_command("down")  
translator_robot.execute_command("left")  

# –¢–µ—Å—Ç–∏—Ä—É–µ–º –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—É—é –∫–æ–º–∞–Ω–¥—É
translator_robot.execute_command("–Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ")

translator_robot.show_info()

print(f"\nüéâ –ü–µ—Ä–µ–≤–æ–¥—á–∏–∫ –∫–æ–º–∞–Ω–¥ —Ä–∞–±–æ—Ç–∞–µ—Ç! –†–æ–±–æ—Ç –ø–æ–Ω–∏–º–∞–µ—Ç —Å–ª–æ–≤–∞!")
```

```python
# –ë–ª–æ–∫ 3.3.2: –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∫–æ–º–∞–Ω–¥
# –†–æ–±–æ—Ç –≤—ã–ø–æ–ª–Ω—è–µ—Ç —Å–ø–∏—Å–∫–∏ –∫–æ–º–∞–Ω–¥!

class AdvancedRobot:
    """–ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–π —Ä–æ–±–æ—Ç —Å —É–º–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        self.path = [(start_x, start_y)]
        self.commands_executed = []  # –∏—Å—Ç–æ—Ä–∏—è –∫–æ–º–∞–Ω–¥
        print(f"ü§ñ {name} –≥–æ—Ç–æ–≤ –∫ –ø—Ä–∏–µ–º—É –∫–æ–º–∞–Ω–¥!")
    
    def move_up(self):
        self.y += 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨ÜÔ∏è {self.name} –∏–¥–µ—Ç –≤–≤–µ—Ä—Ö ‚Üí ({self.x}, {self.y})")
    
    def move_down(self):
        self.y -= 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨áÔ∏è {self.name} –∏–¥–µ—Ç –≤–Ω–∏–∑ ‚Üí ({self.x}, {self.y})")
    
    def move_left(self):
        self.x -= 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨ÖÔ∏è {self.name} –∏–¥–µ—Ç –≤–ª–µ–≤–æ ‚Üí ({self.x}, {self.y})")
    
    def move_right(self):
        self.x += 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚û°Ô∏è {self.name} –∏–¥–µ—Ç –≤–ø—Ä–∞–≤–æ ‚Üí ({self.x}, {self.y})")
    
    def execute_command(self, command):
        """–ü–µ—Ä–µ–≤–æ–¥—á–∏–∫: –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç —Å–ª–æ–≤–∞ –≤ –¥–µ–π—Å—Ç–≤–∏—è"""
        self.commands_executed.append(command)
        if command == "up":
            self.move_up()
        elif command == "down":
            self.move_down()
        elif command == "left":
            self.move_left()
        elif command == "right":
            self.move_right()
        else:
            print(f"‚ùì {self.name}: –ù–µ –∑–Ω–∞—é –∫–æ–º–∞–Ω–¥—É '{command}'!")
            print(f"   –î–æ—Å—Ç—É–ø–Ω—ã–µ –∫–æ–º–∞–Ω–¥—ã: up, down, left, right")
    
    def execute_sequence(self, commands):
        """–í—ã–ø–æ–ª–Ω—è–µ—Ç —Å–ø–∏—Å–æ–∫ –∫–æ–º–∞–Ω–¥"""
        print(f"üìã {self.name} –ø–æ–ª—É—á–∏–ª {len(commands)} –∫–æ–º–∞–Ω–¥: {commands}")
        for i, command in enumerate(commands):
            print(f"  –ö–æ–º–∞–Ω–¥–∞ {i+1}/{len(commands)}: {command}")
            self.execute_command(command)  # –≤—ã–∑—ã–≤–∞–µ–º –º–µ—Ç–æ–¥ execute_command
        print(f"‚úÖ {self.name} –≤—ã–ø–æ–ª–Ω–∏–ª –≤—Å–µ –∫–æ–º–∞–Ω–¥—ã!")
    
    def show_info(self):
        """–ü–æ–¥—Ä–æ–±–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"üìä {self.name}:")
        print(f"   üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y})")
        print(f"   üè† –î–æ–º: ({self.start_x}, {self.start_y})")
        print(f"   üë£ –®–∞–≥–æ–≤: {self.steps_made}")
        print(f"   üìú –ö–æ–º–∞–Ω–¥ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ: {len(self.commands_executed)}")
        if self.commands_executed:
            print(f"   üî§ –ü–æ—Å–ª–µ–¥–Ω–∏–µ –∫–æ–º–∞–Ω–¥—ã: {self.commands_executed[-5:]}")

# –¢–ï–°–¢: –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∫–æ–º–∞–Ω–¥
print("üß™ –¢–ï–°–¢: –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∫–æ–º–∞–Ω–¥")

# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ –¥–ª—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π
sequence_robot = AdvancedRobot("–ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å", 0, 0, "green")

# –°–æ–∑–¥–∞–µ–º –∫–æ–º–∞–Ω–¥—ã –¥–ª—è –∫–≤–∞–¥—Ä–∞—Ç–∞
square_commands = ["right", "right", "up", "up", "left", "left", "down", "down"]

print(f"\n–†–∏—Å—É–µ–º –∫–≤–∞–¥—Ä–∞—Ç:")
sequence_robot.execute_sequence(square_commands)

print(f"\n–¢–µ–ª–µ–ø–æ—Ä—Ç–∏—Ä—É–µ–º—Å—è –∏ —Ä–∏—Å—É–µ–º –µ—â–µ –æ–¥–∏–Ω –∫–≤–∞–¥—Ä–∞—Ç:")
sequence_robot.x, sequence_robot.y = 10, 10  # —Ç–µ–ª–µ–ø–æ—Ä—Ç
sequence_robot.execute_sequence(square_commands)

sequence_robot.show_info()

print(f"\nüéâ –†–æ–±–æ—Ç —É–º–µ–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∫–æ–º–∞–Ω–¥!")
```

```python
import matplotlib.pyplot as plt

class AdvancedRobot:
    """–ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–π —Ä–æ–±–æ—Ç —Å —É–º–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        self.path = [(start_x, start_y)]
        self.commands_executed = []  # –∏—Å—Ç–æ—Ä–∏—è –∫–æ–º–∞–Ω–¥
        print(f"ü§ñ {name} –≥–æ—Ç–æ–≤ –∫ –ø—Ä–∏–µ–º—É –∫–æ–º–∞–Ω–¥!")
    
    def move_up(self):
        self.y += 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨ÜÔ∏è {self.name} –∏–¥–µ—Ç –≤–≤–µ—Ä—Ö ‚Üí ({self.x}, {self.y})")
    
    def move_down(self):
        self.y -= 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨áÔ∏è {self.name} –∏–¥–µ—Ç –≤–Ω–∏–∑ ‚Üí ({self.x}, {self.y})")
    
    def move_left(self):
        self.x -= 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨ÖÔ∏è {self.name} –∏–¥–µ—Ç –≤–ª–µ–≤–æ ‚Üí ({self.x}, {self.y})")
    
    def move_right(self):
        self.x += 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚û°Ô∏è {self.name} –∏–¥–µ—Ç –≤–ø—Ä–∞–≤–æ ‚Üí ({self.x}, {self.y})")
    
    def execute_command(self, command):
        self.commands_executed.append(command)
        if command == "up":
            self.move_up()
        elif command == "down":
            self.move_down()
        elif command == "left":
            self.move_left()
        elif command == "right":
            self.move_right()
        else:
            print(f"‚ùì {self.name}: –ù–µ –∑–Ω–∞—é –∫–æ–º–∞–Ω–¥—É '{command}'!")
            print(f"   –î–æ—Å—Ç—É–ø–Ω—ã–µ –∫–æ–º–∞–Ω–¥—ã: up, down, left, right")
    
    def execute_sequence(self, commands):
        print(f"üìã {self.name} –ø–æ–ª—É—á–∏–ª {len(commands)} –∫–æ–º–∞–Ω–¥: {commands}")
        for i, command in enumerate(commands):
            print(f"  –ö–æ–º–∞–Ω–¥–∞ {i+1}/{len(commands)}: {command}")
            self.execute_command(command)
        print(f"‚úÖ {self.name} –≤—ã–ø–æ–ª–Ω–∏–ª –≤—Å–µ –∫–æ–º–∞–Ω–¥—ã!")
    
    def draw(self):
        """–†–∏—Å—É–µ—Ç –ø—É—Ç—å –∏ —Ç–µ–∫—É—â—É—é –ø–æ–∑–∏—Ü–∏—é —Ä–æ–±–æ—Ç–∞"""
        # –†–∏—Å—É–µ–º –ø—É—Ç—å
        if len(self.path) > 1:
            x_coords = [p[0] for p in self.path]
            y_coords = [p[1] for p in self.path]
            plt.plot(x_coords, y_coords, 
                     color=self.color, 
                     linewidth=2, 
                     alpha=0.6, 
                     linestyle='--')
        
        # –†–∏—Å—É–µ–º —Ç–µ–∫—É—â—É—é –ø–æ–∑–∏—Ü–∏—é
        plt.scatter(self.x, self.y, 
                    color=self.color, 
                    s=200, 
                    alpha=0.9,
                    label=self.name)
        
        # –ü–æ–¥–ø–∏—Å—å —Å –∏–º–µ–Ω–µ–º –∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏
        plt.annotate(f'{self.name}\n({self.x},{self.y})', 
                     (self.x, self.y), 
                     xytext=(10, 10), 
                     textcoords='offset points',
                     fontsize=9)
    
    def show_info(self):
        print(f"üìä {self.name}:")
        print(f"   üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y})")
        print(f"   üè† –î–æ–º: ({self.start_x}, {self.start_y})")
        print(f"   üë£ –®–∞–≥–æ–≤: {self.steps_made}")
        print(f"   üìú –ö–æ–º–∞–Ω–¥ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ: {len(self.commands_executed)}")
        if self.commands_executed:
            print(f"   üî§ –ü–æ—Å–ª–µ–¥–Ω–∏–µ –∫–æ–º–∞–Ω–¥—ã: {self.commands_executed[-5:]}")

# –¢–ï–°–¢: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤
print("üß™ –¢–ï–°–¢: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤")

# –°–æ–∑–¥–∞–µ–º —Ä–∞–∑–Ω—ã—Ö —Ä–æ–±–æ—Ç–æ–≤
test_robot = AdvancedRobot("–¢–µ—Å—Ç–µ—Ä", 0, 0, "red")
translator_robot = AdvancedRobot("–ü–µ—Ä–µ–≤–æ–¥—á–∏–∫", 5, 5, "blue")
sequence_robot = AdvancedRobot("–ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å", 10, 10, "green")

# –î–∞–µ–º –∏–º —Ä–∞–∑–Ω—ã–µ –∑–∞–¥–∞–Ω–∏—è
test_robot.move_up()
test_robot.move_right()

translator_robot.execute_command("up")
translator_robot.execute_command("right")

square_commands = ["right", "right", "up", "up", "left", "left", "down", "down"]
sequence_robot.execute_sequence(square_commands)

# –†–∏—Å—É–µ–º –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤
robots = [test_robot, translator_robot, sequence_robot]

plt.figure(figsize=(12, 8))

for robot in robots:
    robot.draw()  # —Ä–∏—Å—É–µ–º –ø—É—Ç—å –∏ –ø–æ–∑–∏—Ü–∏—é

plt.grid(True, alpha=0.3)
plt.xlim(-2, 15)
plt.ylim(-2, 15)
plt.title("ü§ñ –ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–µ —Ä–æ–±–æ—Ç—ã –∏ –∏—Ö –ø—É—Ç–∏")
plt.xlabel("X –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.ylabel("Y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.legend()
plt.show()

# –ò—Ç–æ–≥–æ–≤–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
print(f"\nüìä –ò–¢–û–ì–û–í–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê:")
total_steps = sum(robot.steps_made for robot in robots)
total_commands = sum(len(robot.commands_executed) for robot in robots)

print(f"   ü§ñ –†–æ–±–æ—Ç–æ–≤ –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–æ: {len(robots)}")
print(f"   üë£ –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤: {total_steps}")
print(f"   üìú –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–º–∞–Ω–¥: {total_commands}")
print(f"   üìä –°—Ä–µ–¥–Ω–µ–µ —à–∞–≥–æ–≤ –Ω–∞ —Ä–æ–±–æ—Ç–∞: {total_steps/len(robots):.1f}")

print(f"\nüéâ –í—Å–µ —Ç–µ—Å—Ç—ã –ø—Ä–æ–π–¥–µ–Ω—ã! –†–æ–±–æ—Ç—ã —Å—Ç–∞–ª–∏ —É–º–Ω–µ–µ!")
```
