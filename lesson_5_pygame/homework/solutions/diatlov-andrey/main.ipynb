```python
import pygame
import math
from collections import deque

pygame.init()
pygame.mixer.init()

pygame.mixer.music.load("your_music_file.mp3") 
pygame.mixer.music.play(-1)
pygame.mixer.music.set_volume(0.5)


WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
GRAY = (200, 200, 200)

def draw_grid(screen):
    for x in range(0, 800, 50):
        color = GRAY if x != 400 else BLACK
        width = 1 if x != 400 else 2
        pygame.draw.line(surface=screen, color=color, start_pos=(x, 0), end_pos=(x, 600), width=width)

    for y in range(0, 600, 50):
        color = GRAY if y != 300 else BLACK
        width = 1 if y != 300 else 2
        pygame.draw.line(surface=screen, color=color, start_pos=(0, y), end_pos=(800, y), width=width)
    

def draw_info(screen, x, y, direction, trail,  font):
    info_texts = [
    f"Позиция: ({x:.1f}, {y:.1f})",
    f"Направление: {direction:.1f}°",
    f"След: {len(trail)} точек",
    "",
    "Управление:",
    "W A S D - движение",
    "Q E - поворот",
    "R - сброс в центр",
    "C - очистить след"
    ]

    y_offset = 10
    for text in info_texts:
        surface = font.render(text, True, BLACK)
        screen.blit(surface, (10, y_offset))
        y_offset += 25


def load_and_scale_robot_sprite(image_path, scale_factor=2.0):
    original_sprite = pygame.image.load(image_path).convert_alpha()
    original_size = original_sprite.get_size()
    new_size = (int(original_size[0] * scale_factor), int(original_size[1] * scale_factor))
    scaled_sprite = pygame.transform.scale(original_sprite, new_size)
    rotated_sprite = pygame.transform.rotate(scaled_sprite, -90)
    return rotated_sprite


def draw_robot_sprite(screen, x, y, direction, robot_sprite):
    screen_x = int(x + 800/2)
    screen_y = int(600/2 - y)
    
    rotated_sprite = pygame.transform.rotate(robot_sprite, direction)
    
    rotated_rect = rotated_sprite.get_rect()
    rotated_rect.center = (screen_x, screen_y)
    
    screen.blit(rotated_sprite, rotated_rect)


def draw_trail(screen, trail):
    if len(trail) < 2:
        return
    screen_points = []
    for x, y in trail:
        screen_x = int(x + 800/2)
        screen_y = int(600/2 - y)
        screen_points.append((screen_x, screen_y))
    if len(screen_points) >= 2:
        pygame.draw.lines(screen, RED, False, screen_points, 2)



screen = pygame.display.set_mode(size=(800, 600))

pygame.display.set_caption(title="My first screen")


clock = pygame.time.Clock()

robot_x = 0
robot_y = 0
robot_direction = 0
move_speed = 3
rotation_speed = 3
trail = deque(maxlen=150)

font = pygame.font.Font(None, 24)

try:
    robot_sprite = load_and_scale_robot_sprite("car_sprite.png", 1)
except:
    robot_sprite = pygame.Surface((32, 32), pygame.SRCALPHA)
    pygame.draw.circle(robot_sprite, BLUE, (16, 16), 16)
    pygame.draw.polygon(robot_sprite, RED, [(16, 4), (12, 28), (20, 28)])
    robot_sprite = pygame.transform.scale(robot_sprite, (64, 64))
    robot_sprite = pygame.transform.rotate(robot_sprite, -90)



running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_r:
                robot_x = 0
                robot_y = 0
                robot_direction = 0
        
            if event.key == pygame.K_c:
                trail.clear()

    screen.fill(color=WHITE)

    draw_grid(screen=screen)

    keys = pygame.key.get_pressed()
    if keys[pygame.K_w]:
        angle = math.radians(robot_direction)
        robot_x += math.cos(angle) * move_speed
        robot_y += math.sin(angle) * move_speed
    if keys[pygame.K_s]:
        angle = math.radians(robot_direction)
        robot_x -= math.cos(angle) * move_speed
        robot_y -= math.sin(angle) * move_speed
    if keys[pygame.K_a]:
        angle = math.radians(robot_direction + 90)
        robot_x += math.cos(angle) * move_speed
        robot_y += math.sin(angle) * move_speed
    if keys[pygame.K_d]:
        angle = math.radians(robot_direction - 90)
        robot_x += math.cos(angle) * move_speed
        robot_y += math.sin(angle) * move_speed
    if keys[pygame.K_q]:
        robot_direction += rotation_speed
    if keys[pygame.K_e]:
        robot_direction -= rotation_speed
    


    
    
    screen_x = robot_x + 400
    screen_y = -robot_y + 300
    robot_direction = robot_direction % 360
    if abs(robot_x) > 350:
        robot_x = 350 if robot_x > 0 else -350
    if abs(robot_y) > 250:
        robot_y = 250 if robot_y > 0 else -250


    draw_trail(screen, trail)
    draw_robot_sprite(screen, robot_x, robot_y, robot_direction, robot_sprite)
    trail.append((robot_x, robot_y))


    draw_info(screen, robot_x, robot_y, robot_direction, trail, font)

    pygame.display.flip()

    clock.tick(60)



pygame.quit()
```
