```python
import pygame
import math

pygame.init()

aa = 800
bb = 600

gg = pygame.display.set_mode((aa, bb))
hh = pygame.time.Clock()

ii = (255, 255, 255)
jj = (0, 0, 0)
kk = (0, 100, 255)

ll = pygame.font.Font(None, 24)

def mm(nn, oo, pp, qq, rr):
    ss = [
        f"Позиция: ({oo:.1f}, {pp:.1f})",
        f"Направление: {qq:.1f}°",
        "",
        "Управление:",
        "W A S D - движение",
        "R - сброс в центр"
    ]
    
    tt = 10
    
    for uu in ss:
        vv = rr.render(uu, True, jj)
        nn.blit(vv, (10, tt))
        tt += 25

def www(nn, oo, pp, qq, rr):
    sss = int(oo + aa / 2)
    ttt = int(bb / 2 - pp)
    
    pygame.draw.circle(nn, kk, (sss, ttt), rr)
    
    uuu = math.radians(qq)
    
    vvv = sss + math.cos(uuu) * rr * 1.5
    www = ttt - math.sin(uuu) * rr * 1.5
    
    xxx = uuu + math.radians(140)
    yyy = sss + math.cos(xxx) * rr * 0.8
    zzz = ttt - math.sin(xxx) * rr * 0.8
    
    aaaa = uuu - math.radians(140)
    bbbb = sss + math.cos(aaaa) * rr * 0.8
    cccc = ttt - math.sin(aaaa) * rr * 0.8
    
    pygame.draw.polygon(nn, ii, [
        (vvv, www),
        (yyy, zzz),
        (bbbb, cccc)
    ])
    
    pygame.draw.circle(nn, jj, (sss, ttt), rr, 2)

dddd = 0
eeee = 0
ffff = 0
gggg = 3

hhhh = True
while hhhh:
    for iiii in pygame.event.get():
        if iiii.type == pygame.QUIT:
            hhhh = False
        
        if iiii.type == pygame.KEYDOWN:
            if iiii.key == pygame.K_r:
                dddd = 0
                eeee = 0
                ffff = 0
    
    jjjj = pygame.key.get_pressed()
    
    if jjjj[pygame.K_w]:
        eeee += gggg
    if jjjj[pygame.K_s]:
        eeee -= gggg
    if jjjj[pygame.K_a]:
        dddd -= gggg
    if jjjj[pygame.K_d]:
        dddd += gggg
    
    gg.fill(ii)
    
    www(gg, dddd, eeee, ffff, 15)
    
    mm(gg, dddd, eeee, ffff, ll)
    
    pygame.display.flip()
    hh.tick(60)

pygame.quit()
```

```python
import pygame
import math

pygame.init()

yazakolebalsya = 800
yaerik = 600

gg = pygame.display.set_mode((yazakolebalsya, yaerik))
hh = pygame.time.Clock()

ii = (255, 255, 255)
jj = (0, 0, 0)
kk = (0, 100, 255)

ll = pygame.font.Font(None, 24)

def mm(nn, oo, pp, qq, rr):
    ss = [
        f"Позиция: ({oo:.1f}, {pp:.1f})",
        f"Направление: {qq:.1f}°",
        "",
        "Управление:",
        "W A S D - движение",
        "Q E - поворот",
        "R - сброс в центр"
    ]
    
    tt = 10
    
    for uu in ss:
        vv = rr.render(uu, True, jj)
        nn.blit(vv, (10, tt))
        tt += 25

def www(nn, oo, pp, qq, rr):
    sss = int(oo + yazakolebalsya / 2)
    ttt = int(yaerik / 2 - pp)
    
    pygame.draw.circle(nn, kk, (sss, ttt), rr)
    
    uuu = math.radians(qq)
    
    vvv = sss + math.cos(uuu) * rr * 1.5
    www = ttt - math.sin(uuu) * rr * 1.5
    
    xxx = uuu + math.radians(140)
    yyy = sss + math.cos(xxx) * rr * 0.8
    zzz = ttt - math.sin(xxx) * rr * 0.8
    
    aaaa = uuu - math.radians(140)
    bbbb = sss + math.cos(aaaa) * rr * 0.8
    cccc = ttt - math.sin(aaaa) * rr * 0.8
    
    pygame.draw.polygon(nn, ii, [
        (vvv, www),
        (yyy, zzz),
        (bbbb, cccc)
    ])
    
    pygame.draw.circle(nn, jj, (sss, ttt), rr, 2)

dddd = 0
eeee = 0
ffff = 0
gggg = 3
hhhh = 3

iiii = True
while iiii:
    for jjjj in pygame.event.get():
        if jjjj.type == pygame.QUIT:
            iiii = False
        
        if jjjj.type == pygame.KEYDOWN:
            if jjjj.key == pygame.K_r:
                dddd = 0
                eeee = 0
                ffff = 0
    
    kkkk = pygame.key.get_pressed()
    
    if kkkk[pygame.K_w]:
        llll = math.radians(ffff)
        dddd += math.cos(llll) * gggg
        eeee += math.sin(llll) * gggg
    
    if kkkk[pygame.K_s]:
        mmmm = math.radians(ffff)
        dddd -= math.cos(mmmm) * gggg
        eeee -= math.sin(mmmm) * gggg
    
    if kkkk[pygame.K_a]:
        nnnn = math.radians(ffff + 90)
        dddd += math.cos(nnnn) * gggg
        eeee += math.sin(nnnn) * gggg
    
    if kkkk[pygame.K_d]:
        oooo = math.radians(ffff - 90)
        dddd += math.cos(oooo) * gggg
        eeee += math.sin(oooo) * gggg
    
    if kkkk[pygame.K_q]:
        ffff += hhhh
        ffff %= 360
    
    if kkkk[pygame.K_e]:
        ffff -= hhhh
        ffff %= 360
    
    gg.fill(ii)
    www(gg, dddd, eeee, ffff, 15)
    mm(gg, dddd, eeee, ffff, ll)
    
    pygame.display.flip()
    hh.tick(60)

pygame.quit()
```

```python
import pygame
import math
from collections import deque

pygame.init()

yazakolebalsya = 800
yaerik = 600

gg = pygame.display.set_mode((yazakolebalsya, yaerik))
hh = pygame.time.Clock()

ii = (255, 255, 255)
jj = (0, 0, 0)
kk = (0, 100, 255)

ll = pygame.font.Font(None, 24)

def mm(nn, oo, pp, qq, rr, ss):
    tt = [
        f"Позиция: ({oo:.1f}, {pp:.1f})",
        f"Направление: {qq:.1f}°",
        f"След: {len(ss)} точек",
        "",
        "Управление:",
        "W A S D - движение",
        "Q E - поворот",
        "R - сброс в центр",
        "C - очистить след"
    ]
    
    uu = 10
    
    for vv in tt:
        ww = ll.render(vv, True, jj)
        nn.blit(ww, (10, uu))
        uu += 25

def xxx(nn, oo, pp, qq, rr):
    yyy = int(oo + yazakolebalsya / 2)
    zzz = int(yaerik / 2 - pp)
    
    pygame.draw.circle(nn, kk, (yyy, zzz), rr)
    
    aaaa = math.radians(qq)
    
    bbbb = yyy + math.cos(aaaa) * rr * 1.5
    cccc = zzz - math.sin(aaaa) * rr * 1.5
    
    dddd = aaaa + math.radians(140)
    eeee = yyy + math.cos(dddd) * rr * 0.8
    ffff = zzz - math.sin(dddd) * rr * 0.8
    
    gggg = aaaa - math.radians(140)
    hhhh = yyy + math.cos(gggg) * rr * 0.8
    iiii = zzz - math.sin(gggg) * rr * 0.8
    
    pygame.draw.polygon(nn, ii, [
        (bbbb, cccc),
        (eeee, ffff),
        (hhhh, iiii)
    ])
    
    pygame.draw.circle(nn, jj, (yyy, zzz), rr, 2)

def jjjj(nn, kkkk):
    if len(kkkk) < 2:
        return
    
    llll = []
    for mmmm, nnnn in kkkk:
        oooo = int(mmmm + yazakolebalsya / 2)
        pppp = int(yaerik / 2 - nnnn)
        llll.append((oooo, pppp))
    
    if len(llll) >= 2:
        pygame.draw.lines(nn, (100, 100, 255), False, llll, 2)

qqqq = 0
rrrr = 0
ssss = 0
tttt = 3
uuuu = 3
vvvv = deque(maxlen=150)

wwww = True
while wwww:
    for xxxx in pygame.event.get():
        if xxxx.type == pygame.QUIT:
            wwww = False
        
        if xxxx.type == pygame.KEYDOWN:
            if xxxx.key == pygame.K_r:
                qqqq = 0
                rrrr = 0
                ssss = 0
            
            if xxxx.key == pygame.K_c:
                vvvv.clear()
    
    yyyy = pygame.key.get_pressed()
    
    if yyyy[pygame.K_q]:
        ssss += uuuu
        ssss %= 360
    if yyyy[pygame.K_e]:
        ssss -= uuuu
        ssss %= 360
    
    if yyyy[pygame.K_w]:
        zzzz = math.radians(ssss)
        qqqq += math.cos(zzzz) * tttt
        rrrr += math.sin(zzzz) * tttt
    
    if yyyy[pygame.K_s]:
        aaaaa = math.radians(ssss)
        qqqq -= math.cos(aaaaa) * tttt
        rrrr -= math.sin(aaaaa) * tttt
    
    if yyyy[pygame.K_a]:
        bbbbb = math.radians(ssss + 90)
        qqqq += math.cos(bbbbb) * tttt
        rrrr += math.sin(bbbbb) * tttt
    
    if yyyy[pygame.K_d]:
        ccccc = math.radians(ssss - 90)
        qqqq += math.cos(ccccc) * tttt
        rrrr += math.sin(ccccc) * tttt
    
    vvvv.append((qqqq, rrrr))
    
    gg.fill(ii)
    jjjj(gg, vvvv)
    xxx(gg, qqqq, rrrr, ssss, 15)
    mm(gg, qqqq, rrrr, ssss, vvvv, ll)
    
    pygame.display.flip()
    hh.tick(60)

pygame.quit()
```
