–ò–∑ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π:

$$\frac{dx}{dt} \approx \frac{x(t + \Delta t) - x(t)}{\Delta t}$$

–û—Ç–∫—É–¥–∞:

$$x(t + \Delta t) \approx x(t) + \frac{dx}{dt} \cdot \Delta t$$

````python
def simulate_projectile_euler(v0, theta_rad, g, dt, max_time=None):
    """
    –°–∏–º—É–ª—è—Ü–∏—è –ø–æ–ª—ë—Ç–∞ —Å–Ω–∞—Ä—è–¥–∞ –º–µ—Ç–æ–¥–æ–º –≠–π–ª–µ—Ä–∞
    
    –û–±–Ω–æ–≤–ª–µ–Ω–∏–µ —É–≥–ª–∞ –∏ —É–≥–ª–æ–≤–æ–π —Å–∫–æ—Ä–æ—Å—Ç–∏
    """
    # –ù–∞—á–∞–ª—å–Ω—ã–µ —É—Å–ª–æ–≤–∏—è
    x = 0.0
    y = 0.0
    vx = v0 * np.cos(theta_rad)
    vy = v0 * np.sin(theta_rad)
    t = 0.0
    
    # –°–ø–∏—Å–∫–∏ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –∏—Å—Ç–æ—Ä–∏–∏
    history = {
        't': [t],
        'x': [x],
        'y': [y],
        'vx': [vx],
        'vy': [vy]
    }
    
    # –°–∏–º—É–ª—è—Ü–∏—è
    while True:
        # –ú–µ—Ç–æ–¥ –≠–π–ª–µ—Ä–∞:
        # x_new = x_old + vx * dt
        # y_new = y_old + vy * dt
        # vx_new = vx_old + 0 * dt  (–Ω–µ—Ç —É—Å–∫–æ—Ä–µ–Ω–∏—è –ø–æ x)
        # vy_new = vy_old + (-g) * dt
        
        x += ...
        y += ...
        vx += ...  # –º–æ–∂–Ω–æ –æ–ø—É—Å—Ç–∏—Ç—å, –Ω–æ –æ—Å—Ç–∞–≤–∏–º –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏
        vy += ...
        t += ...
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º
        history['t'].append(t)
        history['x'].append(x)
        history['y'].append(y)
        history['vx'].append(vx)
        history['vy'].append(vy)
        
        # –£—Å–ª–æ–≤–∏–µ –æ—Å—Ç–∞–Ω–æ–≤–∫–∏: —É–ø–∞–ª –Ω–∞ –∑–µ–º–ª—é (y < 0) –∏–ª–∏ –¥–æ—Å—Ç–∏–≥–Ω—É—Ç–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è
        if y < 0:
            break
        if max_time is not None and t >= max_time:
            break
    
    return pd.DataFrame(history)

# –ó–∞–ø—É—Å–∫–∞–µ–º —Å–∏–º—É–ª—è—Ü–∏—é —Å —à–∞–≥–æ–º dt = 0.01 —Å
dt = 0.01  # —à–∞–≥ –ø–æ –≤—Ä–µ–º–µ–Ω–∏
df_euler = simulate_projectile_euler(v0, theta_rad, g, dt)

print("="*60)
print("üíª –ß–ò–°–õ–ï–ù–ù–û–ï –†–ï–®–ï–ù–ò–ï (–º–µ—Ç–æ–¥ –≠–π–ª–µ—Ä–∞)")
print("="*60)
print(f"\n‚öôÔ∏è  –ü–∞—Ä–∞–º–µ—Ç—Ä—ã —Å–∏–º—É–ª—è—Ü–∏–∏:")
print(f"   –®–∞–≥ –ø–æ –≤—Ä–µ–º–µ–Ω–∏: dt = {dt} —Å")
print(f"   –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤: {len(df_euler)}")

# –†–µ–∑—É–ª—å—Ç–∞—Ç—ã
t_flight_euler = df_euler['t'].iloc[-1]
x_max_euler = df_euler['x'].iloc[-1]
y_max_euler = df_euler['y'].max()
t_max_height_euler = df_euler.loc[df_euler['y'].idxmax(), 't']

print(f"\nüìä –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–∏–º—É–ª—è—Ü–∏–∏:")
print(f"   –í—Ä–µ–º—è –ø–æ–ª—ë—Ç–∞: {t_flight_euler:.4f} —Å")
print(f"   –î–∞–ª—å–Ω–æ—Å—Ç—å: {x_max_euler:.4f} –º")
print(f"   –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –≤—ã—Å–æ—Ç–∞: {y_max_euler:.4f} –º")
print(f"   (–≤ –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ t = {t_max_height_euler:.4f} —Å)")

print("\n" + "="*60)

# –ü–µ—Ä–≤—ã–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫
print("\nüìù –ü–µ—Ä–≤—ã–µ 5 —à–∞–≥–æ–≤ —Å–∏–º—É–ª—è—Ü–∏–∏:")
print(df_euler.head())

print("\nüìù –ü–æ—Å–ª–µ–¥–Ω–∏–µ 5 —à–∞–≥–æ–≤ —Å–∏–º—É–ª—è—Ü–∏–∏:")
print(df_euler.tail())
````

---
## ‚öôÔ∏è –ê–Ω–∞–ª–∏–∑ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–µ—Ç–æ–¥–∞: –≤–ª–∏—è–Ω–∏–µ —à–∞–≥–∞ dt

–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ —Ç–æ—á–Ω–æ—Å—Ç—å –º–µ—Ç–æ–¥–∞ –≠–π–ª–µ—Ä–∞ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —à–∞–≥–∞ –ø–æ –≤—Ä–µ–º–µ–Ω–∏ $\Delta t$.

````python
# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è dt
dt_values = [0.001, 0.01, 0.05, 0.1, 0.2, 0.5]

results = []

for dt_test in dt_values:
    df_test = simulate_projectile_euler(v0, theta_rad, g, dt_test)
    
    # –û—à–∏–±–∫–∏
    t_err = abs(df_test['t'].iloc[-1] - t_flight)
    x_err = abs(df_test['x'].iloc[-1] - x_max)
    y_max_test = df_test['y'].max()
    y_err = abs(y_max_test - y_max)
    
    results.append({
        'dt': dt_test,
        'N_—à–∞–≥–æ–≤': len(df_test),
        '–û—à–∏–±–∫–∞ t (—Å)': t_err,
        '–û—à–∏–±–∫–∞ x (–º)': x_err,
        '–û—à–∏–±–∫–∞ y_max (–º)': y_err,
        '–ú–∞–∫—Å. –æ—à–∏–±–∫–∞ (%)': max(t_err/t_flight, x_err/x_max, y_err/y_max) * 100
    })

df_accuracy = pd.DataFrame(results)

print("="*80)
print("‚öôÔ∏è  –ê–ù–ê–õ–ò–ó –¢–û–ß–ù–û–°–¢–ò: –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –æ—Ç —à–∞–≥–∞ dt")
print("="*80)
print("\n")
print(df_accuracy.to_string(index=False))
print("\n" + "="*80)
print("\nüí° –í—ã–≤–æ–¥—ã:")
print("   ‚Ä¢ –ß–µ–º –º–µ–Ω—å—à–µ dt, —Ç–µ–º —Ç–æ—á–Ω–µ–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç")
print("   ‚Ä¢ –ù–æ –º–µ–Ω—å—à–µ dt ‚Üí –±–æ–ª—å—à–µ —à–∞–≥–æ–≤ ‚Üí –¥–æ–ª—å—à–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è")
print("   ‚Ä¢ –ù—É–∂–µ–Ω –∫–æ–º–ø—Ä–æ–º–∏—Å—Å –º–µ–∂–¥—É —Ç–æ—á–Ω–æ—Å—Ç—å—é –∏ —Å–∫–æ—Ä–æ—Å—Ç—å—é!")
print(f"\n   –î–ª—è dt = {dt} —Å –æ—à–∏–±–∫–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç ~ {df_accuracy[df_accuracy['dt']==dt]['–ú–∞–∫—Å. –æ—à–∏–±–∫–∞ (%)'].values[0]:.3f}%")
````
