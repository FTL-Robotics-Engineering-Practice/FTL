```python
# –ë–ª–æ–∫ 2.2: –†–æ–±–æ—Ç —É—á–∏—Ç—Å—è —Ö–æ–¥–∏—Ç—å –ø–æ –∫–≤–∞–¥—Ä–∞—Ç—É
# –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–æ–¥—ã –¥–≤–∏–∂–µ–Ω–∏—è –≤ –∫–ª–∞—Å—Å

class Robot:
    """–†–æ–±–æ—Ç, –∫–æ—Ç–æ—Ä—ã–π —É–º–µ–µ—Ç —Ö–æ–¥–∏—Ç—å –ø–æ –∫–≤–∞–¥—Ä–∞—Ç—É"""
    
    def __init__(self, name, start_x=0, start_y=0):
        """–°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ —Å –∏–º–µ–Ω–µ–º –∏ —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –ø–æ–∑–∏—Ü–∏–µ–π"""
        self.name = name
        self.x = start_x
        self.y = start_y
        print(f"ü§ñ {name} —Å–æ–∑–¥–∞–Ω –≤ –ø–æ–∑–∏—Ü–∏–∏ ({start_x}, {start_y})")
    
    def info(self):
        """–ü–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"ü§ñ {self.name} –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ ({self.x}, {self.y})")
    
    def walk_square(self, size=3):
        """–†–æ–±–æ—Ç —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞"""
        print(f"üö∂ {self.name} –Ω–∞—á–∏–Ω–∞–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}")
        
        # TODO: –ó–∞–ø–æ–ª–Ω–∏—Ç–µ –∫–æ–¥ –¥–≤–∏–∂–µ–Ω–∏—è —Ä–æ–±–æ—Ç–∞
        # –ü–æ–¥—Å–∫–∞–∑–∫–∞: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ self.x –∏ self.y
        
        # –î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ (size —à–∞–≥–æ–≤)
        print(f"  {self.name} –∏–¥–µ—Ç –≤–ø—Ä–∞–≤–æ:")
        for step in range(size):
            self.x += 1  # —à–∞–≥ –≤–ø—Ä–∞–≤–æ
            print(f"    –®–∞–≥ {step+1}: {self.name} –≤ ({self.x}, {self.y})")
        
        # TODO: –î–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö (size —à–∞–≥–æ–≤)
        print(f"  {self.name} –∏–¥–µ—Ç –≤–≤–µ—Ä—Ö:")
        for step in range(size):
            self.y += 1  # –í–ê–® –ö–û–î: –∫–∞–∫—É—é –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É –º–µ–Ω—è—Ç—å?
            print(f"    –®–∞–≥ {step+1}: {self.name} –≤ ({self.x}, {self.y})")
        
        # TODO: –î–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ (size —à–∞–≥–æ–≤)
        print(f"  {self.name} –∏–¥–µ—Ç –≤–ª–µ–≤–æ:")
        for step in range(size):  # –í–ê–® –ö–û–î: —Å–∫–æ–ª—å–∫–æ —à–∞–≥–æ–≤?
            self.x -= 1  # –í–ê–® –ö–û–î: –∫–∞–∫ –∏–¥—Ç–∏ –≤–ª–µ–≤–æ?
            print(f"    –®–∞–≥ {step+1}: {self.name} –≤ ({self.x}, {self.y})")
        
        # TODO: –î–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑ (size —à–∞–≥–æ–≤)
        print(f"  {self.name} –∏–¥–µ—Ç –≤–Ω–∏–∑:")
        # –í–ê–® –ö–û–î: –∑–∞–ø–æ–ª–Ω–∏—Ç–µ –≤–µ—Å—å —Ü–∏–∫–ª –¥–ª—è –¥–≤–∏–∂–µ–Ω–∏—è –≤–Ω–∏–∑
        for step in range(size):  # –í–ê–® –ö–û–î: —Å–∫–æ–ª—å–∫–æ —à–∞–≥–æ–≤?
            self.y -= 1  # –í–ê–® –ö–û–î: –∫–∞–∫ –∏–¥—Ç–∏ –≤–ª–µ–≤–æ?
            print(f"    –®–∞–≥ {step+1}: {self.name} –≤ ({self.x}, {self.y})")
        
        
        print(f"‚úÖ {self.name} –∑–∞–≤–µ—Ä—à–∏–ª –∫–≤–∞–¥—Ä–∞—Ç! –§–∏–Ω–∞–ª—å–Ω–∞—è –ø–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y})")

# –¢–µ—Å—Ç–∏—Ä—É–µ–º –Ω–∞—à–µ–≥–æ —Ä–æ–±–æ—Ç–∞
print("üöÄ –¢–ï–°–¢: –û–¥–∏–Ω —Ä–æ–±–æ—Ç —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç")

# TODO: –°–æ–∑–¥–∞–π—Ç–µ —Ä–æ–±–æ—Ç–∞ –ø–æ –∏–º–µ–Ω–∏ "–¢–µ—Å—Ç–µ—Ä" –≤ –ø–æ–∑–∏—Ü–∏–∏ (0, 0)
robot = Robot("–¢–µ—Å—Ç–µ—Ä", 0, 0)

# –ü–æ–∫–∞–∑–∞—Ç—å –Ω–∞—á–∞–ª—å–Ω—É—é –ø–æ–∑–∏—Ü–∏—é
robot.info()

# TODO: –ó–∞—Å—Ç–∞–≤–∏—Ç—å —Ä–æ–±–æ—Ç–∞ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç —Ä–∞–∑–º–µ—Ä–æ–º 2x2
robot.walk_square(2)

# –ü–æ–∫–∞–∑–∞—Ç—å —Ñ–∏–Ω–∞–ª—å–Ω—É—é –ø–æ–∑–∏—Ü–∏—é
robot.info()

# –ü—Ä–æ–≤–µ—Ä—è–µ–º: —Ä–æ–±–æ—Ç –¥–æ–ª–∂–µ–Ω –≤–µ—Ä–Ω—É—Ç—å—Å—è –≤ —Å—Ç–∞—Ä—Ç–æ–≤—É—é –ø–æ–∑–∏—Ü–∏—é
start_x, start_y = 0, 0
if robot.x == start_x and robot.y == start_y:
    print("‚úÖ –û—Ç–ª–∏—á–Ω–æ! –†–æ–±–æ—Ç –≤–µ—Ä–Ω—É–ª—Å—è –¥–æ–º–æ–π!")
else:
    print(f"‚ùå –†–æ–±–æ—Ç –∑–∞–±–ª—É–¥–∏–ª—Å—è! –û–∂–∏–¥–∞–ª–æ—Å—å ({start_x}, {start_y}), –ø–æ–ª—É—á–∏–ª–∏ ({robot.x}, {robot.y})")

print("\nüéâ –ü–æ–∑–¥—Ä–∞–≤–ª—è–µ–º! –í—ã —Å–æ–∑–¥–∞–ª–∏ —Ä–æ–±–æ—Ç–∞ —Å –º–µ—Ç–æ–¥–æ–º!")
```

```python
# –ë–ª–æ–∫ 2.3.1: –ö—Ä–∞—Å–∏–≤—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ä–æ–±–æ—Ç–µ

class Robot:
    """–†–æ–±–æ—Ç —Å –∫—Ä–∞—Å–∏–≤—ã–º –≤—ã–≤–æ–¥–æ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏"""
    
    def __init__(self, name, start_x=0, start_y=0):
        """–°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ —Å –∏–º–µ–Ω–µ–º –∏ —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –ø–æ–∑–∏—Ü–∏–µ–π"""
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x  # –∑–∞–ø–æ–º–∏–Ω–∞–µ–º –¥–æ–º
        self.start_y = start_y
        self.steps_made = 0     # —Å—á–µ—Ç—á–∏–∫ —à–∞–≥–æ–≤
        print(f"ü§ñ {name} —Å–æ–∑–¥–∞–Ω –≤ –ø–æ–∑–∏—Ü–∏–∏ ({start_x}, {start_y})")
    
    def show_info(self):
        """–ö—Ä–∞—Å–∏–≤—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"‚îå{'‚îÄ' * 40}‚îê")
        print(f"‚îÇ ü§ñ –†–æ–±–æ—Ç: {self.name:<28} ‚îÇ")
        print(f"‚îÇ üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y}){' ' * (23 - len(str(self.x)) - len(str(self.y)))}‚îÇ")
        print(f"‚îÇ üè† –î–æ–º: ({self.start_x}, {self.start_y}){' ' * (27 - len(str(self.start_x)) - len(str(self.start_y)))}‚îÇ")
        print(f"‚îÇ üë£ –®–∞–≥–æ–≤ —Å–¥–µ–ª–∞–Ω–æ: {self.steps_made:<20} ‚îÇ")
        print(f"‚îî{'‚îÄ' * 40}‚îò")
    
    def move_right(self):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ"""
        self.x += 1
        self.steps_made += 1
        print(f"    üë£ {self.name} –≤ ({self.x}, {self.y})")

# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ –∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
print("üè≠ –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞!")
robot = Robot("–ê–ª—å—Ñ–∞", 5, 3)

print("\nüìä –ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –†–û–ë–û–¢–ï:")
# TODO: –í—ã–∑–æ–≤–∏—Ç–µ –º–µ—Ç–æ–¥ show_info() –¥–ª—è —Ä–æ–±–æ—Ç–∞
robot.show_info()

print("\nüö∂ –†–æ–±–æ—Ç –¥–µ–ª–∞–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ —à–∞–≥–æ–≤:")
robot.move_right()
robot.move_right()

print("\nüìä –û–ë–ù–û–í–õ–ï–ù–ù–ê–Ø –ò–ù–§–û–†–ú–ê–¶–ò–Ø:")
# TODO: –°–Ω–æ–≤–∞ –ø–æ–∫–∞–∂–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ä–æ–±–æ—Ç–µ
robot.show_info()

print("\n‚úÖ –û—Ç–ª–∏—á–Ω–æ! –¢–µ–ø–µ—Ä—å —É –Ω–∞—Å –µ—Å—Ç—å –∫—Ä–∞—Å–∏–≤—ã–π —Å–ø–æ—Å–æ–± –ø–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ä–æ–±–æ—Ç–µ!")
```

```python
# –ë–ª–æ–∫ 2.3.2: –û—Ç–¥–µ–ª—å–Ω—ã–µ –º–µ—Ç–æ–¥—ã –¥–≤–∏–∂–µ–Ω–∏—è

class Robot:
    """–†–æ–±–æ—Ç —Å –æ—Ç–¥–µ–ª—å–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏ –¥–≤–∏–∂–µ–Ω–∏—è"""
    
    def __init__(self, name, start_x=0, start_y=0):
        """–°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ —Å –∏–º–µ–Ω–µ–º –∏ —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –ø–æ–∑–∏—Ü–∏–µ–π"""
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.steps_made = 0
        print(f"ü§ñ {name} —Å–æ–∑–¥–∞–Ω –≤ –ø–æ–∑–∏—Ü–∏–∏ ({start_x}, {start_y})")
    
    def show_info(self):
        """–ö—Ä–∞—Å–∏–≤—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"‚îå{'‚îÄ' * 40}‚îê")
        print(f"‚îÇ ü§ñ –†–æ–±–æ—Ç: {self.name:<30} ‚îÇ")
        print(f"‚îÇ üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y}){' ' * (22 - len(str(self.x)) - len(str(self.y)))}‚îÇ")
        print(f"‚îÇ üè† –î–æ–º: ({self.start_x}, {self.start_y}){' ' * (26 - len(str(self.start_x)) - len(str(self.start_y)))}‚îÇ")
        print(f"‚îÇ üë£ –®–∞–≥–æ–≤ —Å–¥–µ–ª–∞–Ω–æ: {self.steps_made:<18} ‚îÇ")
        print(f"‚îî{'‚îÄ' * 40}‚îò")
    
    def move_right(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        # TODO: –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ steps > 0
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        # TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ
        for step in range(steps):
            self.x += 1
            self.steps_made += 1
            print(f"    ‚û°Ô∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_left(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        # TODO: –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ steps > 0
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        # TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ
        for step in range(steps):
            self.x -= 1
            self.steps_made += 1
            print(f"    ‚¨ÖÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_up(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        # TODO: –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ steps > 0
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        # TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö
        for step in range(steps):
            self.y += 1
            self.steps_made += 1
            print(f"    ‚¨ÜÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_down(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        # TODO: –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ steps > 0
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        # TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑
        for step in range(steps):
            self.y -= 1
            self.steps_made += 1
            print(f"    ‚¨áÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")

# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞
print("üè≠ –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞!")
robot = Robot("–ë–µ—Ç–∞", 0, 0)

print("\nüìä –ù–ê–ß–ê–õ–¨–ù–û–ï –°–û–°–¢–û–Ø–ù–ò–ï:")
robot.show_info()

print("\nüéÆ –£–ü–†–ê–í–õ–Ø–ï–ú –†–û–ë–û–¢–û–ú –û–¢–î–ï–õ–¨–ù–´–ú–ò –ú–ï–¢–û–î–ê–ú–ò:")
# TODO: –ó–∞—Å—Ç–∞–≤—å—Ç–µ —Ä–æ–±–æ—Ç–∞ –¥–≤–∏–≥–∞—Ç—å—Å—è —Ä–∞–∑–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏
robot.move_right(2)  # 2 —à–∞–≥–∞ –≤–ø—Ä–∞–≤–æ
robot.move_up(3)     # 3 —à–∞–≥–∞ –≤–≤–µ—Ä—Ö
robot.move_left(1)   # 1 —à–∞–≥ –≤–ª–µ–≤–æ
robot.move_down(2)   # 2 —à–∞–≥–∞ –≤–Ω–∏–∑

print("\nüß™ –¢–ï–°–¢–ò–†–£–ï–ú –û–®–ò–ë–ö–ò:")
# TODO: –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —à–∞–≥–∏ - —á—Ç–æ –ø—Ä–æ–∏–∑–æ–π–¥–µ—Ç?
robot.move_right(-31)

print("\nüìä –§–ò–ù–ê–õ–¨–ù–û–ï –°–û–°–¢–û–Ø–ù–ò–ï:")
robot.show_info()

print("\n‚úÖ –û—Ç–ª–∏—á–Ω–æ! –¢–µ–ø–µ—Ä—å –º—ã –º–æ–∂–µ–º —É–ø—Ä–∞–≤–ª—è—Ç—å —Ä–æ–±–æ—Ç–æ–º –æ—Ç–¥–µ–ª—å–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏!")
print("üí° –ù–µ –∑–∞–±—É–¥—å—Ç–µ –¥–æ–±–∞–≤–∏—Ç—å –ø—Ä–æ–≤–µ—Ä–∫—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —à–∞–≥–æ–≤!")
```

```python
# –ë–ª–æ–∫ 2.3.4: –¢—Ä–∏ —Ä–æ–±–æ—Ç–∞ —Ä–∏—Å—É—é—Ç –∫–≤–∞–¥—Ä–∞—Ç—ã –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ

class Robot:
    """–ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–π —Ä–æ–±–æ—Ç —Å –∫—Ä–∞—Å–∏–≤—ã–º –≤—ã–≤–æ–¥–æ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏"""
    
    def __init__(self, name, start_x=0, start_y=0):
        """–°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ —Å –∏–º–µ–Ω–µ–º –∏ —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –ø–æ–∑–∏—Ü–∏–µ–π"""
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x  # –∑–∞–ø–æ–º–∏–Ω–∞–µ–º –¥–æ–º
        self.start_y = start_y
        self.steps_made = 0     # —Å—á–µ—Ç—á–∏–∫ —à–∞–≥–æ–≤
        print(f"ü§ñ {name} —Å–æ–∑–¥–∞–Ω –≤ –ø–æ–∑–∏—Ü–∏–∏ ({start_x}, {start_y})")
    
    def show_info(self):
        """–ö—Ä–∞—Å–∏–≤—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"‚îå{'‚îÄ' * 40}‚îê")
        print(f"‚îÇ ü§ñ –†–æ–±–æ—Ç: {self.name:<30} ‚îÇ")
        print(f"‚îÇ üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y}){' ' * (22 - len(str(self.x)) - len(str(self.y)))}‚îÇ")
        print(f"‚îÇ üè† –î–æ–º: ({self.start_x}, {self.start_y}){' ' * (26 - len(str(self.start_x)) - len(str(self.start_y)))}‚îÇ")
        print(f"‚îÇ üë£ –®–∞–≥–æ–≤ —Å–¥–µ–ª–∞–Ω–æ: {self.steps_made:<18} ‚îÇ")
        print(f"‚îî{'‚îÄ' * 40}‚îò")
    
    def move_right(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.x += 1
            self.steps_made += 1
            print(f"    ‚û°Ô∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_left(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.x -= 1
            self.steps_made += 1
            print(f"    ‚¨ÖÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_up(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.y += 1
            self.steps_made += 1
            print(f"    ‚¨ÜÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_down(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.y -= 1
            self.steps_made += 1
            print(f"    ‚¨áÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def walk_square(self, size=3):
        """–†–æ–±–æ—Ç —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞"""
        print(f"\nüé® {self.name} –Ω–∞—á–∏–Ω–∞–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}")
        
        # –í–ø—Ä–∞–≤–æ
        print(f"  ‚û°Ô∏è –ò–¥–µ–º –≤–ø—Ä–∞–≤–æ ({size} —à–∞–≥–æ–≤):")
        self.move_right(size)
        
        # –í–≤–µ—Ä—Ö
        print(f"  ‚¨ÜÔ∏è –ò–¥–µ–º –≤–≤–µ—Ä—Ö ({size} —à–∞–≥–æ–≤):")
        self.move_up(size)
        
        # –í–ª–µ–≤–æ
        print(f"  ‚¨ÖÔ∏è –ò–¥–µ–º –≤–ª–µ–≤–æ ({size} —à–∞–≥–æ–≤):")
        self.move_left(size)
        
        # –í–Ω–∏–∑
        print(f"  ‚¨áÔ∏è –ò–¥–µ–º –≤–Ω–∏–∑ ({size} —à–∞–≥–æ–≤):")
        self.move_down(size)
        
        print(f"‚úÖ {self.name} –∑–∞–≤–µ—Ä—à–∏–ª –∫–≤–∞–¥—Ä–∞—Ç!")

# –°–æ–∑–¥–∞–µ–º —Ç—Ä–µ—Ö —Ä–æ–±–æ—Ç–æ–≤ –≤ —Ä–∞–∑–Ω—ã—Ö –ø–æ–∑–∏—Ü–∏—è—Ö
print("üè≠ –ó–∞–ø—É—Å–∫–∞–µ–º —Ñ–∞–±—Ä–∏–∫—É —Ä–æ–±–æ—Ç–æ–≤!")

# TODO: –°–æ–∑–¥–∞–π—Ç–µ —Ç—Ä–µ—Ö —Ä–æ–±–æ—Ç–æ–≤
# –†–æ–±–æ—Ç 1: "–ê–ª—å—Ñ–∞" –≤ –ø–æ–∑–∏—Ü–∏–∏ (0, 0)
robot1 = Robot("–ê–ª—å—Ñ–∞", 0, 0)

# –†–æ–±–æ—Ç 2: "–ë–µ—Ç–∞" –≤ –ø–æ–∑–∏—Ü–∏–∏ (10, 5)  
robot2 = Robot("–ë–µ—Ç–∞", 10, 5)

# –†–æ–±–æ—Ç 3: "–ì–∞–º–º–∞" –≤ –ø–æ–∑–∏—Ü–∏–∏ (20, 10)
robot3 = Robot("–ì–∞–º–º–∞", 20, 10)

# –ü–æ–∫–∞–∑–∞—Ç—å –Ω–∞—á–∞–ª—å–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
print("\nüìä –ù–ê–ß–ê–õ–¨–ù–û–ï –°–û–°–¢–û–Ø–ù–ò–ï –†–û–ë–û–¢–û–í:")
# TODO: –í—ã–∑–≤–∞—Ç—å show_info() –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–æ–±–æ—Ç–∞
robot1.show_info()
robot2.show_info()
robot3.show_info()

# –ó–∞–ø—É—Å—Ç–∏—Ç—å –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤ —Ä–∏—Å–æ–≤–∞—Ç—å
print("\nüé® –í–°–ï –†–û–ë–û–¢–´ –ù–ê–ß–ò–ù–ê–Æ–¢ –†–ò–°–û–í–ê–¢–¨:")

# TODO: –ó–∞—Å—Ç–∞–≤–∏—Ç—å –∫–∞–∂–¥–æ–≥–æ —Ä–æ–±–æ—Ç–∞ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç —Ä–∞–∑–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
# –ê–ª—å—Ñ–∞ —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç 2x2
robot1.walk_square(2)

# –ë–µ—Ç–∞ —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç 3x3
robot2.walk_square(3)

# –ì–∞–º–º–∞ —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç 4x4  
robot3.walk_square(4)

# –ü–æ–∫–∞–∑–∞—Ç—å —Ñ–∏–Ω–∞–ª—å–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
print("\nüìä –§–ò–ù–ê–õ–¨–ù–û–ï –°–û–°–¢–û–Ø–ù–ò–ï –†–û–ë–û–¢–û–í:")
# TODO: –°–Ω–æ–≤–∞ –ø–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–∞—Ö
robot1.show_info()
robot2.show_info()
robot3.show_info()

# –ü—Ä–æ–≤–µ—Ä–∫–∞: –∫—Ç–æ –≤–µ—Ä–Ω—É–ª—Å—è –¥–æ–º–æ–π?
print("\nüè† –ü–†–û–í–ï–†–ö–ê: –ö–¢–û –í–ï–†–ù–£–õ–°–Ø –î–û–ú–û–ô?")
robots = [robot1, robot2, robot3]
for robot in robots:
    if robot.x == robot.start_x and robot.y == robot.start_y:
        print(f"‚úÖ {robot.name} –≤–µ—Ä–Ω—É–ª—Å—è –¥–æ–º–æ–π!")
    else:
        print(f"‚ùå {robot.name} –∑–∞–±–ª—É–¥–∏–ª—Å—è!")

print(f"\nüéâ –ó–∞–¥–∞–Ω–∏–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ! –ú—ã —É–ø—Ä–∞–≤–ª—è–ª–∏ {len(robots)} —Ä–æ–±–æ—Ç–∞–º–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ!")
```

```python
# –ë–ª–æ–∫ 3.1: –§–∞–±—Ä–∏–∫–∞ —Ä–æ–±–æ—Ç–æ–≤ - –º–∞—Å—Å–æ–≤–æ–µ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–æ!
# –°–æ–∑–¥–∞–µ–º –º–Ω–æ–≥–æ —Ä–æ–±–æ—Ç–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ –¥–∞–Ω–Ω—ã—Ö

class Robot:
    """–†–æ–±–æ—Ç –¥–ª—è –º–∞—Å—Å–æ–≤–æ–≥–æ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        print(f"ü§ñ {name} ({color}) —Å–æ–∑–¥–∞–Ω –≤ ({start_x}, {start_y})")
    
    def show_info(self):
        """–ö–æ–º–ø–∞–∫—Ç–Ω—ã–π –≤—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏"""
        status = "üè†" if (self.x == self.start_x and self.y == self.start_y) else "üó∫Ô∏è"
        print(f"{status} {self.name:<10} | –ü–æ–∑–∏—Ü–∏—è: ({self.x:2}, {self.y:2}) | –®–∞–≥–æ–≤: {self.steps_made:2} | –¶–≤–µ—Ç: {self.color}")
    
    def walk_square(self, size=3):
        """–ë—ã—Å—Ç—Ä–æ–µ —Ä–∏—Å–æ–≤–∞–Ω–∏–µ –∫–≤–∞–¥—Ä–∞—Ç–∞ (–±–µ–∑ –¥–µ—Ç–∞–ª—å–Ω–æ–≥–æ –≤—ã–≤–æ–¥–∞)"""
        directions = ["right", "up", "left", "down"]
        
        for direction in directions:
            for step in range(size):
                if direction == "right":
                    self.x += 1
                elif direction == "up":
                    self.y += 1
                elif direction == "left":
                    self.x -= 1
                elif direction == "down":
                    self.y -= 1
                self.steps_made += 1
        
        print(f"‚úÖ {self.name} –∑–∞–≤–µ—Ä—à–∏–ª –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}")

# –ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö —Ä–æ–±–æ—Ç–æ–≤ –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∏—á–µ—Å–∫–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
# –ö–∞–∂–¥—ã–π —Å–ª–æ–≤–∞—Ä—å –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å: "name", "x", "y", "color"
# –°—Ç–∞—Ä—Ç–æ–≤—ã–µ –ø–æ–∑–∏—Ü–∏–∏ –∏ —Ä–∞–∑–º–µ—Ä—ã –ø–æ–¥–æ–±—Ä–∞–Ω—ã —Ç–∞–∫, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏–ª–∏—Å—å –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∏—á–µ—Å–∫–∏–µ –∫–≤–∞–¥—Ä–∞—Ç—ã
robots_data = [
    # TODO: –ó–∞–ø–æ–ª–Ω–∏—Ç–µ —Å–ª–æ–≤–∞—Ä–∏ –¥–∞–Ω–Ω—ã–º–∏ –¥–ª—è —Ä–æ–±–æ—Ç–æ–≤
    # –°—Ç–∞—Ä—Ç–æ–≤—ã–µ –ø–æ–∑–∏—Ü–∏–∏: (0,0), (1,1), (2,2)
    # –¶–≤–µ—Ç–∞ –º–æ–∂–µ—Ç–µ –≤—ã–±—Ä–∞—Ç—å –ª—é–±—ã–µ
    {"name": "R2D2", "x": 0, "y": 0, "color": "red"},
    {"name": "Chopper", "x": 1, "y": 1, "color": "blue"},
    {"name": "Wall-e", "x": 2, "y": 2, "color": "green"}
]

print("üè≠ –§–ê–ë–†–ò–ö–ê –†–û–ë–û–¢–û–í –ó–ê–ü–£–©–ï–ù–ê!")
print(f"üìã –í –æ—á–µ—Ä–µ–¥–∏ –Ω–∞ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–æ: {len(robots_data)} —Ä–æ–±–æ—Ç–æ–≤")

# TODO: –°–æ–∑–¥–∞–π—Ç–µ –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ –¥–∞–Ω–Ω—ã—Ö
robot_army = []

print("\nüîß –ü–†–û–ò–ó–í–û–î–°–¢–í–û –†–û–ë–û–¢–û–í:")
for robot_data in robots_data:
    # TODO: –°–æ–∑–¥–∞–π—Ç–µ —Ä–æ–±–æ—Ç–∞ –∏–∑ –¥–∞–Ω–Ω—ã—Ö —Å–ª–æ–≤–∞—Ä—è
    # –ü–æ–¥—Å–∫–∞–∑–∫–∞: robot_data["name"], robot_data["x"], –∏ —Ç.–¥.
    robot = Robot(
        robot_data["name"],    # –∏–º—è
        robot_data["x"],    # x –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞  
        robot_data["y"],    # y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞
        robot_data["color"]     # —Ü–≤–µ—Ç
    )
    robot_army.append(robot)

print(f"\n‚úÖ –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–æ —Ä–æ–±–æ—Ç–æ–≤: {len(robot_army)}")

# –ü–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–∞—Ö
print("\nüìä –†–ï–ï–°–¢–† –†–û–ë–û–¢–û–í:")
print("   –°—Ç–∞—Ç—É—Å | –ò–º—è        | –ü–æ–∑–∏—Ü–∏—è    | –®–∞–≥–æ–≤ | –¶–≤–µ—Ç")
print("   " + "‚îÄ" * 50)

# TODO: –ü–æ–∫–∞–∂–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∫–∞–∂–¥–æ–º —Ä–æ–±–æ—Ç–µ
for robot in robot_army:
    robot.show_info()

# –ó–∞–ø—É—Å—Ç–∏—Ç—å –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤ —Ä–∞–±–æ—Ç–∞—Ç—å
print("\nüé® –ú–ê–°–°–û–í–û–ï –ü–†–û–ò–ó–í–û–î–°–¢–í–û –ö–í–ê–î–†–ê–¢–û–í:")

# TODO: –ó–∞—Å—Ç–∞–≤—å—Ç–µ –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤ –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç—ã —Ä–∞–∑–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
# –î–ª—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∏—á–µ—Å–∫–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ä–∞–∑–º–µ—Ä—ã: 4, 3, 2
sizes = [4, 3, 2]  # —Ä–∞–∑–º–µ—Ä—ã –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –¥–ª—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∏—á–µ—Å–∫–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤

for i, robot in enumerate(robot_army):
    size = sizes[i]
    print(f"ü§ñ {robot.name} —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}...")
    robot.walk_square(size)

# –§–∏–Ω–∞–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
print("\nüìä –§–ò–ù–ê–õ–¨–ù–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê:")
print("   –°—Ç–∞—Ç—É—Å | –ò–º—è        | –ü–æ–∑–∏—Ü–∏—è    | –®–∞–≥–æ–≤ | –¶–≤–µ—Ç") 
print("   " + "‚îÄ" * 50)

# TODO: –°–Ω–æ–≤–∞ –ø–æ–∫–∞–∂–∏—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–∞—Ö
for robot in robot_army:
    robot.show_info()

# –ê–Ω–∞–ª–∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤  
print(f"\nüìà –ê–ù–ê–õ–ò–ó –ü–†–û–ò–ó–í–û–î–°–¢–í–ê:")

# TODO: –ü–æ–¥—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
total_steps = 0
robots_at_home = 0

for robot in robot_army:
    total_steps += robot.steps_made  # –¥–æ–±–∞–≤—å—Ç–µ —à–∞–≥–∏ —Ä–æ–±–æ—Ç–∞
    
    # –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, –¥–æ–º–∞ –ª–∏ —Ä–æ–±–æ—Ç
    if robot.x == robot.start_x and robot.y == robot.start_y:
        robots_at_home += 1

print(f"   ü§ñ –í—Å–µ–≥–æ —Ä–æ–±–æ—Ç–æ–≤: {len(robot_army)}")
print(f"   üë£ –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤: {total_steps}")
print(f"   üìä –°—Ä–µ–¥–Ω–µ–µ —à–∞–≥–æ–≤ –Ω–∞ —Ä–æ–±–æ—Ç–∞: {total_steps / len(robot_army):.1f}")
print(f"   üè† –†–æ–±–æ—Ç–æ–≤ –≤–µ—Ä–Ω—É–ª–æ—Å—å –¥–æ–º–æ–π: {robots_at_home}")
print(f"   üìè –≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –≤–æ–∑–≤—Ä–∞—Ç–∞: {robots_at_home / len(robot_army) * 100:.1f}%")

# –ü–æ–∏—Å–∫ —á–µ–º–ø–∏–æ–Ω–æ–≤
most_active = max(robot_army, key=lambda r: r.steps_made)
print(f"   üèÜ –°–∞–º—ã–π –∞–∫—Ç–∏–≤–Ω—ã–π: {most_active.name} ({most_active.steps_made} —à–∞–≥–æ–≤)")

print(f"\nüéâ –§–ê–ë–†–ò–ö–ê –ó–ê–í–ï–†–®–ò–õ–ê –†–ê–ë–û–¢–£! –í—Å–µ —Ä–æ–±–æ—Ç—ã –≥–æ—Ç–æ–≤—ã –∫ —Å–ª—É–∂–±–µ!")
```

```python
# –ë–ª–æ–∫ 3.2.1: –†–∏—Å–æ–≤–∞–Ω–∏–µ –æ–¥–Ω–æ–≥–æ —Ä–æ–±–æ—Ç–∞ —Å –ø–æ–º–æ—â—å—é matplotlib
# –£—á–∏–º—Å—è —Ä–∏—Å–æ–≤–∞—Ç—å —Ä–æ–±–æ—Ç–∞ –Ω–∞ —ç–∫—Ä–∞–Ω–µ!

import matplotlib.pyplot as plt
import numpy as np

class Robot:
    """–†–æ–±–æ—Ç, –∫–æ—Ç–æ—Ä—ã–π —É–º–µ–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å —Å–µ–±—è –Ω–∞ —ç–∫—Ä–∞–Ω–µ"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        print(f"ü§ñ {name} ({color}) —Å–æ–∑–¥–∞–Ω –≤ ({start_x}, {start_y})")
    
    def draw(self, size=100):
        """–†–∏—Å—É–µ—Ç —Ä–æ–±–æ—Ç–∞ –Ω–∞ —Ç–µ–∫—É—â–µ–º –≥—Ä–∞—Ñ–∏–∫–µ"""
        # TODO: –ù–∞—Ä–∏—Å—É–π—Ç–µ —Ä–æ–±–æ—Ç–∞ –∫–∞–∫ —Ç–æ—á–∫—É
        # –ü–æ–¥—Å–∫–∞–∑–∫–∞: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ plt.scatter()
        plt.scatter(self.x, self.y, 
                   color=self.color, 
                   s=size, 
                   alpha=0.8,
                   label=self.name)
        
        # TODO: –î–æ–±–∞–≤—å—Ç–µ –ø–æ–¥–ø–∏—Å—å —Å –∏–º–µ–Ω–µ–º –∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏
        # –ü–æ–¥—Å–∫–∞–∑–∫–∞: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ plt.annotate()
        plt.annotate(f'{self.name}\n({self.x},{self.y})', 
                    (self.x, self.y), 
                    xytext=(10, 10), 
                    textcoords='offset points',
                    fontsize=8,
                    bbox=dict(boxstyle="round,pad=0.3", 
                             facecolor=self.color, 
                             alpha=0.3))
    
    def move_right(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ø—Ä–∞–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.x += 1
            self.steps_made += 1
            print(f"    ‚û°Ô∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_left(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–ª–µ–≤–æ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.x -= 1
            self.steps_made += 1
            print(f"    ‚¨ÖÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_up(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–≤–µ—Ä—Ö –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.y += 1
            self.steps_made += 1
            print(f"    ‚¨ÜÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def move_down(self, steps=1):
        """–î–≤–∏–∂–µ–Ω–∏–µ –≤–Ω–∏–∑ –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤"""
        if steps <= 0:
            print(f"‚ùå {self.name}: –Ω–µ–ª—å–∑—è –¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ {steps} —à–∞–≥–æ–≤!")
            return
        
        for step in range(steps):
            self.y -= 1
            self.steps_made += 1
            print(f"    ‚¨áÔ∏è {self.name} —à–∞–≥ {step+1} ‚Üí ({self.x}, {self.y})")
    
    def walk_square(self, size=3):
        """–†–æ–±–æ—Ç —Ä–∏—Å—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞"""
        print(f"\nüé® {self.name} –Ω–∞—á–∏–Ω–∞–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å –∫–≤–∞–¥—Ä–∞—Ç {size}x{size}")
        
        # –í–ø—Ä–∞–≤–æ
        print(f"  ‚û°Ô∏è –ò–¥–µ–º –≤–ø—Ä–∞–≤–æ ({size} —à–∞–≥–æ–≤):")
        self.move_right(size)
        
        # –í–≤–µ—Ä—Ö
        print(f"  ‚¨ÜÔ∏è –ò–¥–µ–º –≤–≤–µ—Ä—Ö ({size} —à–∞–≥–æ–≤):")
        self.move_up(size)
        
        # –í–ª–µ–≤–æ
        print(f"  ‚¨ÖÔ∏è –ò–¥–µ–º –≤–ª–µ–≤–æ ({size} —à–∞–≥–æ–≤):")
        self.move_left(size)
        
        # –í–Ω–∏–∑
        print(f"  ‚¨áÔ∏è –ò–¥–µ–º –≤–Ω–∏–∑ ({size} —à–∞–≥–æ–≤):")
        self.move_down(size)
        
        print(f"‚úÖ {self.name} –∑–∞–≤–µ—Ä—à–∏–ª –∫–≤–∞–¥—Ä–∞—Ç!")

# –¢–µ—Å—Ç: –†–∏—Å—É–µ–º –æ–¥–Ω–æ–≥–æ —Ä–æ–±–æ—Ç–∞
print("üé® –¢–ï–°–¢: –†–∏—Å–æ–≤–∞–Ω–∏–µ –æ–¥–Ω–æ–≥–æ —Ä–æ–±–æ—Ç–∞")

# TODO: –°–æ–∑–¥–∞–π—Ç–µ —Ä–æ–±–æ—Ç–∞ –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
test_robot = Robot("Voody", 3, 1, "red")

# –ù–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º –≥—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(8, 6))

# TODO: –ù–∞—Ä–∏—Å—É–π—Ç–µ —Ä–æ–±–æ—Ç–∞
test_robot.draw()

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –≥—Ä–∞—Ñ–∏–∫–∞
plt.grid(True, alpha=0.3)
plt.xlim(-2, 10)
plt.ylim(-2, 8)
plt.title("ü§ñ –ú–æ–π –ø–µ—Ä–≤—ã–π —Ä–æ–±–æ—Ç –Ω–∞ —ç–∫—Ä–∞–Ω–µ!")
plt.xlabel("X –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.ylabel("Y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.legend()
plt.show()

print("‚úÖ –¢–µ—Å—Ç –∑–∞–≤–µ—Ä—à–µ–Ω! –†–æ–±–æ—Ç –Ω–∞—Ä–∏—Å–æ–≤–∞–Ω!")

# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ: –ø–æ–ø—Ä–æ–±—É–π—Ç–µ —Ä–∞–∑–Ω—ã–µ —Ü–≤–µ—Ç–∞ –∏ —Ä–∞–∑–º–µ—Ä—ã
print("\nüé® –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–û–ï –ó–ê–î–ê–ù–ò–ï:")
print("–°–æ–∑–¥–∞–π—Ç–µ –µ—â–µ –æ–¥–Ω–æ–≥–æ —Ä–æ–±–æ—Ç–∞ —Å –¥—Ä—É–≥–∏–º —Ü–≤–µ—Ç–æ–º –∏ —Ä–∞–∑–º–µ—Ä–æ–º!")

# TODO: –°–æ–∑–¥–∞–π—Ç–µ –≤—Ç–æ—Ä–æ–≥–æ —Ä–æ–±–æ—Ç–∞
second_robot = Robot("Boomy", -1, 3, "green")

plt.figure(figsize=(8, 6))
# TODO: –ù–∞—Ä–∏—Å—É–π—Ç–µ –≤—Ç–æ—Ä–æ–≥–æ —Ä–æ–±–æ—Ç–∞ —Å –±–æ–ª—å—à–∏–º —Ä–∞–∑–º–µ—Ä–æ–º
second_robot.draw(size=600)

plt.grid(True, alpha=0.3)
plt.xlim(-2, 10)
plt.ylim(-2, 8)
plt.title("ü§ñü§ñ –î–≤–∞ —Ä–æ–±–æ—Ç–∞ –Ω–∞ —ç–∫—Ä–∞–Ω–µ!")
plt.xlabel("X –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.ylabel("Y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.legend()
plt.show()

print("üéâ –û—Ç–ª–∏—á–Ω–æ! –¢–µ–ø–µ—Ä—å –≤—ã —É–º–µ–µ—Ç–µ —Ä–∏—Å–æ–≤–∞—Ç—å —Ä–æ–±–æ—Ç–æ–≤!")
```

```python
# –ë–ª–æ–∫ 3.3.3: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–æ–¥–≤–∏–Ω—É—Ç—ã—Ö —Ä–æ–±–æ—Ç–æ–≤
# –†–∏—Å—É–µ–º –ø—É—Ç–∏ –∏ –ø–æ–∑–∏—Ü–∏–∏ —Ä–æ–±–æ—Ç–æ–≤!

import matplotlib.pyplot as plt

class AdvancedRobot:
    """–ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–π —Ä–æ–±–æ—Ç —Å —É–º–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        self.path = [(start_x, start_y)]
        self.commands_executed = []  # –∏—Å—Ç–æ—Ä–∏—è –∫–æ–º–∞–Ω–¥
        print(f"ü§ñ {name} –≥–æ—Ç–æ–≤ –∫ –ø—Ä–∏–µ–º—É –∫–æ–º–∞–Ω–¥!")
    
    def move_up(self):
        """–®–∞–≥ –≤–≤–µ—Ä—Ö"""
        self.y += 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨ÜÔ∏è {self.name} –∏–¥–µ—Ç –≤–≤–µ—Ä—Ö ‚Üí ({self.x}, {self.y})")
    
    def move_down(self):
        """–®–∞–≥ –≤–Ω–∏–∑"""
        self.y -= 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨áÔ∏è {self.name} –∏–¥–µ—Ç –≤–Ω–∏–∑ ‚Üí ({self.x}, {self.y})")
    
    def move_left(self):
        """–®–∞–≥ –≤–ª–µ–≤–æ"""
        self.x -= 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚¨ÖÔ∏è {self.name} –∏–¥–µ—Ç –≤–ª–µ–≤–æ ‚Üí ({self.x}, {self.y})")
    
    def move_right(self):
        """–®–∞–≥ –≤–ø—Ä–∞–≤–æ"""
        self.x += 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
        print(f"‚û°Ô∏è {self.name} –∏–¥–µ—Ç –≤–ø—Ä–∞–≤–æ ‚Üí ({self.x}, {self.y})")
    
    def execute_command(self, command):
        """–ü–µ—Ä–µ–≤–æ–¥—á–∏–∫: –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç —Å–ª–æ–≤–∞ –≤ –¥–µ–π—Å—Ç–≤–∏—è"""
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∫–æ–º–∞–Ω–¥—É –≤ –∏—Å—Ç–æ—Ä–∏–∏
        self.commands_executed.append(command)
        
        if command == "up":
            self.move_up()
        elif command == "down":
            self.move_down()
        elif command == "left":
            self.move_left()
        elif command == "right":
            self.move_right()
        else:
            print(f"‚ùì {self.name}: –ù–µ –∑–Ω–∞—é –∫–æ–º–∞–Ω–¥—É '{command}'!")
            print(f"   –î–æ—Å—Ç—É–ø–Ω—ã–µ –∫–æ–º–∞–Ω–¥—ã: up, down, left, right")
    
    def execute_sequence(self, commands):
        """–í—ã–ø–æ–ª–Ω—è–µ—Ç —Å–ø–∏—Å–æ–∫ –∫–æ–º–∞–Ω–¥"""
        print(f"üìã {self.name} –ø–æ–ª—É—á–∏–ª {len(commands)} –∫–æ–º–∞–Ω–¥: {commands}")
        
        # –í—ã–ø–æ–ª–Ω—è–µ–º –≤—Å–µ –∫–æ–º–∞–Ω–¥—ã –∏–∑ —Å–ø–∏—Å–∫–∞
        for i, command in enumerate(commands):
            print(f"  –ö–æ–º–∞–Ω–¥–∞ {i+1}/{len(commands)}: {command}")
            self.execute_command(command)
        
        print(f"‚úÖ {self.name} –≤—ã–ø–æ–ª–Ω–∏–ª –≤—Å–µ –∫–æ–º–∞–Ω–¥—ã!")
    
    def draw(self):
        """–†–∏—Å—É–µ—Ç —Ä–æ–±–æ—Ç–∞ –∏ –µ–≥–æ –ø—É—Ç—å"""
        # TODO: –ù–∞—Ä–∏—Å—É–π—Ç–µ –ø—É—Ç—å —Ä–æ–±–æ—Ç–∞
        if len(self.path) > 1:
            x_coords = [p[0] for p in self.path]
            y_coords = [p[1] for p in self.path]
            plt.plot(x_coords, y_coords, 
                    color=self.color, 
                    linewidth=2, 
                    alpha=0.6, 
                    linestyle='--')
        
        # TODO: –ù–∞—Ä–∏—Å—É–π—Ç–µ —Ç–µ–∫—É—â—É—é –ø–æ–∑–∏—Ü–∏—é
        plt.scatter(self.x, self.y, 
                   color=self.color, 
                   s=200, 
                   alpha=0.9,
                   label=self.name)
        
        # TODO: –î–æ–±–∞–≤—å—Ç–µ –ø–æ–¥–ø–∏—Å—å
        plt.annotate(f'{self.name}\n({self.x},{self.y})', 
                    (self.x, self.y), 
                    xytext=(10, 10), 
                    textcoords='offset points',
                    fontsize=9)
    
    def show_info(self):
        """–ü–æ–¥—Ä–æ–±–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ä–æ–±–æ—Ç–µ"""
        print(f"üìä {self.name}:")
        print(f"   üìç –ü–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y})")
        print(f"   üè† –î–æ–º: ({self.start_x}, {self.start_y})")
        print(f"   üë£ –®–∞–≥–æ–≤: {self.steps_made}")
        print(f"   üìú –ö–æ–º–∞–Ω–¥ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ: {len(self.commands_executed)}")
        if self.commands_executed:
            print(f"   üî§ –ü–æ—Å–ª–µ–¥–Ω–∏–µ –∫–æ–º–∞–Ω–¥—ã: {self.commands_executed[-5:]}")

# –¢–ï–°–¢: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤
print("üß™ –¢–ï–°–¢: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤")

# –°–æ–∑–¥–∞–µ–º —Ä–∞–∑–Ω—ã—Ö —Ä–æ–±–æ—Ç–æ–≤
test_robot = AdvancedRobot("–¢–µ—Å—Ç–µ—Ä", 0, 0, "red")
translator_robot = AdvancedRobot("–ü–µ—Ä–µ–≤–æ–¥—á–∏–∫", 5, 5, "blue")
sequence_robot = AdvancedRobot("–ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å", 10, 10, "green")

# –î–∞–µ–º –∏–º —Ä–∞–∑–Ω—ã–µ –∑–∞–¥–∞–Ω–∏—è
test_robot.move_up()
test_robot.move_right()

translator_robot.execute_command("up")
translator_robot.execute_command("right")

square_commands = ["right", "right", "up", "up", "left", "left", "down", "down"]
sequence_robot.execute_sequence(square_commands)

# –†–∏—Å—É–µ–º –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤
robots = [test_robot, translator_robot, sequence_robot]

plt.figure(figsize=(12, 8))

# TODO: –ù–∞—Ä–∏—Å—É–π—Ç–µ –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤ –≤ —Ü–∏–∫–ª–µ
for robot in robots:
    robot.draw()

plt.grid(True, alpha=0.3)
plt.xlim(-2, 15)
plt.ylim(-2, 15)
plt.title("ü§ñ –ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–µ —Ä–æ–±–æ—Ç—ã –∏ –∏—Ö –ø—É—Ç–∏")
plt.xlabel("X –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.ylabel("Y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.legend()
plt.show()

# –ò—Ç–æ–≥–æ–≤–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
print(f"\nüìä –ò–¢–û–ì–û–í–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê:")
total_steps = sum(robot.steps_made for robot in robots)
total_commands = sum(len(robot.commands_executed) for robot in robots)

print(f"   ü§ñ –†–æ–±–æ—Ç–æ–≤ –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–æ: {len(robots)}")
print(f"   üë£ –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤: {total_steps}")
print(f"   üìú –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–º–∞–Ω–¥: {total_commands}")
print(f"   üìä –°—Ä–µ–¥–Ω–µ–µ —à–∞–≥–æ–≤ –Ω–∞ —Ä–æ–±–æ—Ç–∞: {total_steps/len(robots):.1f}")

print(f"\nüéâ –í—Å–µ —Ç–µ—Å—Ç—ã –ø—Ä–æ–π–¥–µ–Ω—ã! –†–æ–±–æ—Ç—ã —Å—Ç–∞–ª–∏ —É–º–Ω–µ–µ!")
```

```python
# –ë–ª–æ–∫ 4.1: –ù–∞–≤–∏–≥–∞—Ü–∏—è –∫ —Ü–µ–ª–∏ - —Ä–æ–±–æ—Ç—ã –ø–ª–∞–Ω–∏—Ä—É—é—Ç –º–∞—Ä—à—Ä—É—Ç!
# –£—á–∏–º —Ä–æ–±–æ—Ç–æ–≤ –Ω–∞—Ö–æ–¥–∏—Ç—å –ø—É—Ç—å —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ

import matplotlib.pyplot as plt
import math

class NavigatorRobot:
    """–†–æ–±–æ—Ç-–Ω–∞–≤–∏–≥–∞—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π —É–º–µ–µ—Ç –ø–ª–∞–Ω–∏—Ä–æ–≤–∞—Ç—å –º–∞—Ä—à—Ä—É—Ç—ã"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        self.path = [(start_x, start_y)]
        self.targets_reached = 0
        print(f"üß≠ {name} –≥–æ—Ç–æ–≤ –∫ –Ω–∞–≤–∏–≥–∞—Ü–∏–∏ –∏–∑ ({start_x}, {start_y})")
    
    def move_up(self):
        self.y += 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
    
    def move_down(self):
        self.y -= 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
    
    def move_left(self):
        self.x -= 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
    
    def move_right(self):
        self.x += 1
        self.steps_made += 1
        self.path.append((self.x, self.y))
    
    def execute_command(self, command):
        """–í—ã–ø–æ–ª–Ω—è–µ—Ç –æ–¥–Ω—É –∫–æ–º–∞–Ω–¥—É"""
        if command == "up":
            self.move_up()
        elif command == "down":
            self.move_down()
        elif command == "left":
            self.move_left()
        elif command == "right":
            self.move_right()
    
    def calculate_distance(self, target_x, target_y):
        """–í—ã—á–∏—Å–ª—è–µ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–ª–∏"""
        # TODO: –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ–æ—Ä–º—É–ª—É —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è
        dx = target_x - self.x
        dy = target_y - self.y
        distance = math.sqrt(dx * dx + dy * dy)
        return distance
    
    def generate_path_commands(self, target_x, target_y):
        """–ì–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ –∫–æ–º–∞–Ω–¥ –¥–ª—è –¥–æ—Å—Ç–∏–∂–µ–Ω–∏—è —Ü–µ–ª–∏"""
        commands = []
        
        # TODO: –í—ã—á–∏—Å–ª–∏—Ç–µ —Å–∫–æ–ª—å–∫–æ —à–∞–≥–æ–≤ –Ω—É–∂–Ω–æ –ø–æ –∫–∞–∂–¥–æ–π –æ—Å–∏
        dx = target_x - self.x  # —Å–∫–æ–ª—å–∫–æ –ø–æ X
        dy = target_y - self.y  # —Å–∫–æ–ª—å–∫–æ –ø–æ Y
        
        print(f"üìê {self.name}: –Ω—É–∂–Ω–æ –ø—Ä–æ–π—Ç–∏ Œîx={dx}, Œîy={dy}")
        
        # TODO: –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∫–æ–º–∞–Ω–¥—ã –¥–ª—è –¥–≤–∏–∂–µ–Ω–∏—è –ø–æ X
        if dx > 0:  # –Ω—É–∂–Ω–æ –∏–¥—Ç–∏ –≤–ø—Ä–∞–≤–æ
            commands.extend(["right"] * dx)  # –¥–æ–±–∞–≤–ª—è–µ–º –∫–æ–º–∞–Ω–¥—ã "right"
        elif dx < 0:  # –Ω—É–∂–Ω–æ –∏–¥—Ç–∏ –≤–ª–µ–≤–æ
            commands.extend(["left"] * abs(dx))  # –¥–æ–±–∞–≤–ª—è–µ–º –∫–æ–º–∞–Ω–¥—ã "left"
        
        # TODO: –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∫–æ–º–∞–Ω–¥—ã –¥–ª—è –¥–≤–∏–∂–µ–Ω–∏—è –ø–æ Y
        if dy > 0:  # –Ω—É–∂–Ω–æ –∏–¥—Ç–∏ –≤–≤–µ—Ä—Ö
            commands.extend(["up"] * dy)
        elif dy < 0:  # –Ω—É–∂–Ω–æ –∏–¥—Ç–∏ –≤–Ω–∏–∑  
            commands.extend(["down"] * abs(dy))
        
        return commands
    
    def move_to_target(self, target_x, target_y):
        """–ì–ª–∞–≤–Ω—ã–π –º–µ—Ç–æ–¥: –ø–ª–∞–Ω–∏—Ä—É–µ—Ç –∏ –≤—ã–ø–æ–ª–Ω—è–µ—Ç –º–∞—Ä—à—Ä—É—Ç –∫ —Ü–µ–ª–∏"""
        print(f"\nüéØ {self.name} –ø–ª–∞–Ω–∏—Ä—É–µ—Ç –º–∞—Ä—à—Ä—É—Ç –∏–∑ ({self.x}, {self.y}) –≤ ({target_x}, {target_y})")
        
        # –í—ã—á–∏—Å–ª—è–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ
        distance = self.calculate_distance(target_x, target_y)
        print(f"üìè –†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–ª–∏: {distance:.2f}")
        
        # –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∫–æ–º–∞–Ω–¥—ã
        commands = self.generate_path_commands(target_x, target_y)
        print(f"üó∫Ô∏è –ü–ª–∞–Ω –º–∞—Ä—à—Ä—É—Ç–∞: {commands}")
        print(f"üìä –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤: {len(commands)}")
        
        # –í—ã–ø–æ–ª–Ω—è–µ–º –∫–æ–º–∞–Ω–¥—ã
        print(f"üö∂ –ù–∞—á–∏–Ω–∞–µ–º –¥–≤–∏–∂–µ–Ω–∏–µ...")
        for i, command in enumerate(commands):
            self.execute_command(command)
            print(f"   –®–∞–≥ {i+1}: {command} ‚Üí ({self.x}, {self.y})")
        
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —É—Å–ø–µ—Ö
        if self.x == target_x and self.y == target_y:
            self.targets_reached += 1
            print(f"‚úÖ {self.name} –¥–æ—Å—Ç–∏–≥ —Ü–µ–ª–∏ ({target_x}, {target_y})!")
        else:
            print(f"‚ùå {self.name} –ø—Ä–æ–º–∞—Ö–Ω—É–ª—Å—è! –¶–µ–ª—å: ({target_x}, {target_y}), —Ñ–∞–∫—Ç: ({self.x}, {self.y})")
        
        return self.x == target_x and self.y == target_y
    
    def show_info(self):
        """–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –Ω–∞–≤–∏–≥–∞—Ç–æ—Ä–µ"""
        print(f"üß≠ {self.name}:")
        print(f"   üìç –¢–µ–∫—É—â–∞—è –ø–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y})")
        print(f"   üè† –°—Ç–∞—Ä—Ç–æ–≤–∞—è –ø–æ–∑–∏—Ü–∏—è: ({self.start_x}, {self.start_y})")
        print(f"   üë£ –®–∞–≥–æ–≤ —Å–¥–µ–ª–∞–Ω–æ: {self.steps_made}")
        print(f"   üéØ –¶–µ–ª–µ–π –¥–æ—Å—Ç–∏–≥–Ω—É—Ç–æ: {self.targets_reached}")
    
    def draw(self, targets=None):
        """–†–∏—Å—É–µ—Ç —Ä–æ–±–æ—Ç–∞, –µ–≥–æ –ø—É—Ç—å –∏ —Ü–µ–ª–∏"""
        # –ü—É—Ç—å —Ä–æ–±–æ—Ç–∞
        if len(self.path) > 1:
            x_coords = [p[0] for p in self.path]
            y_coords = [p[1] for p in self.path]
            plt.plot(x_coords, y_coords, 
                    color=self.color, 
                    linewidth=2, 
                    alpha=0.7,
                    label=f'{self.name} –ø—É—Ç—å')
        
        # –°—Ç–∞—Ä—Ç–æ–≤–∞—è –ø–æ–∑–∏—Ü–∏—è
        plt.scatter(self.start_x, self.start_y, 
                   color=self.color, 
                   s=150, 
                   marker='^', 
                   alpha=0.8,
                   label=f'{self.name} —Å—Ç–∞—Ä—Ç')
        
        # –¢–µ–∫—É—â–∞—è –ø–æ–∑–∏—Ü–∏—è
        plt.scatter(self.x, self.y, 
                   color=self.color, 
                   s=200, 
                   marker='o',
                   alpha=0.9,
                   label=f'{self.name}')
        
        # –¶–µ–ª–∏ (–µ—Å–ª–∏ –ø–µ—Ä–µ–¥–∞–Ω—ã)
        if targets:
            for i, (tx, ty) in enumerate(targets):
                plt.scatter(tx, ty, 
                           color='red', 
                           s=150, 
                           marker='*',
                           alpha=0.8)
                plt.annotate(f'–¶–µ–ª—å {i+1}', (tx, ty), 
                           xytext=(5, 5), 
                           textcoords='offset points')

# –¢–ï–°–¢ 1: –ü—Ä–æ—Å—Ç–∞—è –Ω–∞–≤–∏–≥–∞—Ü–∏—è
print("üß™ –¢–ï–°–¢ 1: –ü—Ä–æ—Å—Ç–∞—è –Ω–∞–≤–∏–≥–∞—Ü–∏—è –∫ –æ–¥–Ω–æ–π —Ü–µ–ª–∏")

# TODO: –°–æ–∑–¥–∞–π—Ç–µ –Ω–∞–≤–∏–≥–∞—Ç–æ—Ä–∞
navigator = NavigatorRobot("Aang", 1, 6, "blue")

# TODO: –û—Ç–ø—Ä–∞–≤—å—Ç–µ –µ–≥–æ –∫ —Ü–µ–ª–∏ (5, 3)
target_x, target_y = 5, 3
success = navigator.move_to_target(target_x, target_y)

navigator.show_info()
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {'‚úÖ –£—Å–ø–µ—Ö' if success else '‚ùå –ù–µ—É–¥–∞—á–∞'}")

# –¢–ï–°–¢ 2: –ú–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ü–µ–ª–∏
print("\nüß™ –¢–ï–°–¢ 2: –ü—É—Ç–µ—à–µ—Å—Ç–≤–∏–µ –ø–æ –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω—ã–º —Ü–µ–ª—è–º")

# TODO: –°–æ–∑–¥–∞–π—Ç–µ –Ω–æ–≤–æ–≥–æ —Ä–æ–±–æ—Ç–∞-–ø—É—Ç–µ—à–µ—Å—Ç–≤–µ–Ω–Ω–∏–∫–∞
traveler = NavigatorRobot("Katara", 7, 4, "green")

# –°–ø–∏—Å–æ–∫ —Ü–µ–ª–µ–π –¥–ª—è –ø–æ—Å–µ—â–µ–Ω–∏—è
destinations = [(3, 4), (7, 2), (1, 6), (8, 8)]

print(f"üó∫Ô∏è {traveler.name} –¥–æ–ª–∂–µ–Ω –ø–æ—Å–µ—Ç–∏—Ç—å {len(destinations)} –º–µ—Å—Ç:")
for i, (x, y) in enumerate(destinations):
    print(f"   {i+1}. ({x}, {y})")

# TODO: –ü–æ—Å–µ—Ç–∏—Ç–µ –≤—Å–µ —Ü–µ–ª–∏ –ø–æ –æ—á–µ—Ä–µ–¥–∏
successful_trips = 0
for i, (target_x, target_y) in enumerate(destinations):
    print(f"\n--- –ü–û–ï–ó–î–ö–ê {i+1} –ö ({target_x}, {target_y}) ---")
    success = traveler.move_to_target(target_x, target_y)
    if success:
        successful_trips += 1

traveler.show_info()
print(f"\nüìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø—É—Ç–µ—à–µ—Å—Ç–≤–∏–π:")
print(f"   üéØ –¶–µ–ª–µ–π –Ω–∞–∑–Ω–∞—á–µ–Ω–æ: {len(destinations)}")
print(f"   ‚úÖ –¶–µ–ª–µ–π –¥–æ—Å—Ç–∏–≥–Ω—É—Ç–æ: {successful_trips}")
print(f"   üìà –ü—Ä–æ—Ü–µ–Ω—Ç —É—Å–ø–µ—Ö–∞: {successful_trips/len(destinations)*100:.1f}%")

# –¢–ï–°–¢ 3: –ù–∞–≤–∏–≥–∞—Ü–∏—è –≤ —Ä–∞–∑–Ω—ã–µ —Å—Ç–æ—Ä–æ–Ω—ã
print("\nüß™ –¢–ï–°–¢ 3: –ù–∞–≤–∏–≥–∞—Ü–∏—è –≤–æ –≤—Å–µ —Å—Ç–æ—Ä–æ–Ω—ã —Å–≤–µ—Ç–∞")

# TODO: –°–æ–∑–¥–∞–π—Ç–µ —Ä–æ–±–æ—Ç–∞-–∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—è
explorer = NavigatorRobot("Zukko", 3, 3, "red")

# –¶–µ–ª–∏ –≤–æ –≤—Å–µ—Ö –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è—Ö –æ—Ç —Ü–µ–Ω—Ç—Ä–∞
directions_targets = [
    (10, 5),  # –≤–æ—Å—Ç–æ–∫
    (5, 10),  # —Å–µ–≤–µ—Ä  
    (0, 5),   # –∑–∞–ø–∞–¥
    (5, 0),   # —é–≥
    (5, 5)    # –≤–æ–∑–≤—Ä–∞—Ç –≤ —Ü–µ–Ω—Ç—Ä
]

direction_names = ["–í–æ—Å—Ç–æ–∫", "–°–µ–≤–µ—Ä", "–ó–∞–ø–∞–¥", "–Æ–≥", "–¶–µ–Ω—Ç—Ä"]

for i, ((target_x, target_y), direction) in enumerate(zip(directions_targets, direction_names)):
    print(f"\n--- –ò–°–°–õ–ï–î–û–í–ê–ù–ò–ï: {direction.upper()} ({target_x}, {target_y}) ---")
    explorer.move_to_target(target_x, target_y)

explorer.show_info()

# –¢–ï–°–¢ 4: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤—Å–µ—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤
print("\nüß™ –¢–ï–°–¢ 4: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –Ω–∞–≤–∏–≥–∞—Ü–∏–∏")

robots = [navigator, traveler, explorer]

plt.figure(figsize=(12, 10))

# –†–∏—Å—É–µ–º –≤—Å–µ—Ö —Ä–æ–±–æ—Ç–æ–≤
for robot in robots:
    if robot == traveler:
        robot.draw(targets=destinations)
    elif robot == explorer:
        robot.draw(targets=directions_targets)
    else:
        robot.draw(targets=[(5, 3)])

plt.grid(True, alpha=0.3)
plt.xlim(-1, 12)
plt.ylim(-1, 12)
plt.title("üß≠ –ù–∞–≤–∏–≥–∞—Ü–∏–æ–Ω–Ω—ã–µ –º–∞—Ä—à—Ä—É—Ç—ã —Ä–æ–±–æ—Ç–æ–≤")
plt.xlabel("X –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.ylabel("Y –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# –ò—Ç–æ–≥–æ–≤–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
print(f"\nüìä –ò–¢–û–ì–û–í–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê –ù–ê–í–ò–ì–ê–¶–ò–ò:")
total_steps = sum(robot.steps_made for robot in robots)
total_targets = sum(robot.targets_reached for robot in robots)

print(f"   ü§ñ –†–æ–±–æ—Ç–æ–≤-–Ω–∞–≤–∏–≥–∞—Ç–æ—Ä–æ–≤: {len(robots)}")
print(f"   üë£ –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤: {total_steps}")
print(f"   üéØ –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–µ–ª–µ–π: {total_targets}")
print(f"   üìä –°—Ä–µ–¥–Ω–µ–µ —à–∞–≥–æ–≤ –Ω–∞ —Ü–µ–ª—å: {total_steps/total_targets:.1f}")

# –°–∞–º—ã–π —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–π –Ω–∞–≤–∏–≥–∞—Ç–æ—Ä
most_efficient = min(robots, key=lambda r: r.steps_made/max(r.targets_reached, 1))
print(f"   üèÜ –°–∞–º—ã–π —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–π: {most_efficient.name} ({most_efficient.steps_made/max(most_efficient.targets_reached, 1):.1f} —à–∞–≥–æ–≤/—Ü–µ–ª—å)")

print(f"\nüéâ –ù–∞–≤–∏–≥–∞—Ü–∏–æ–Ω–Ω–∞—è —Å–∏—Å—Ç–µ–º–∞ —É—Å–ø–µ—à–Ω–æ –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∞!")
```
