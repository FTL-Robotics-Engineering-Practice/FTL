```python
# ะะปะพะบ 3.1: ะคะฐะฑัะธะบะฐ ัะพะฑะพัะพะฒ - ะผะฐััะพะฒะพะต ะฟัะพะธะทะฒะพะดััะฒะพ!
# ะกะพะทะดะฐะตะผ ะผะฝะพะณะพ ัะพะฑะพัะพะฒ ะธะท ัะฟะธัะบะฐ ะดะฐะฝะฝัั

class Robot:
    """ะะพะฑะพั ะดะปั ะผะฐััะพะฒะพะณะพ ะฟัะพะธะทะฒะพะดััะฒะฐ"""
    
    def __init__(self, name, start_x=0, start_y=0, color="blue"):
        self.name = name
        self.x = start_x
        self.y = start_y
        self.start_x = start_x
        self.start_y = start_y
        self.color = color
        self.steps_made = 0
        print(f"๐ค {name} ({color}) ัะพะทะดะฐะฝ ะฒ ({start_x}, {start_y})")
    
    def show_info(self):
        """ะะพะผะฟะฐะบัะฝัะน ะฒัะฒะพะด ะธะฝัะพัะผะฐัะธะธ"""
        status = "๐" if (self.x == self.start_x and self.y == self.start_y) else "๐บ๏ธ"
        print(f"{status} {self.name:<10} | ะะพะทะธัะธั: ({self.x:2}, {self.y:2}) | ะจะฐะณะพะฒ: {self.steps_made:2} | ะฆะฒะตั: {self.color}")
    
    def walk_square(self, size=3):
        """ะััััะพะต ัะธัะพะฒะฐะฝะธะต ะบะฒะฐะดัะฐัะฐ (ะฑะตะท ะดะตัะฐะปัะฝะพะณะพ ะฒัะฒะพะดะฐ)"""
        directions = ["right", "up", "left", "down"]
        
        for direction in directions:
            for step in range(size):
                if direction == "right":
                    self.x += 1
                elif direction == "up":
                    self.y += 1
                elif direction == "left":
                    self.x -= 1
                elif direction == "down":
                    self.y -= 1
                self.steps_made += 1
        
        print(f"โ {self.name} ะทะฐะฒะตััะธะป ะบะฒะฐะดัะฐั {size}x{size}")

# ะะฐะทะฐ ะดะฐะฝะฝัั ัะพะฑะพัะพะฒ ะดะปั ัะพะทะดะฐะฝะธั ะบะพะฝัะตะฝััะธัะตัะบะธั ะบะฒะฐะดัะฐัะพะฒ
# ะะฐะถะดัะน ัะปะพะฒะฐัั ัะพะดะตัะถะธั: "name", "x", "y", "color"
# ะกัะฐััะพะฒัะต ะฟะพะทะธัะธะธ ะธ ัะฐะทะผะตัั ะฟะพะดะพะฑัะฐะฝั ัะฐะบ, ััะพะฑั ะฟะพะปััะธะปะธัั ะบะพะฝัะตะฝััะธัะตัะบะธะต ะบะฒะฐะดัะฐัั
robots_data = [
    {"name": "ะะปััะฐ", "x": 0, "y": 0, "color": "red"},
    {"name": "ะะตัะฐ", "x": 1, "y": 1, "color": "blue"},
    {"name": "ะะฐะผะผะฐ", "x": 2, "y": 2, "color": "green"},
    {"name": "ะะตะปััะฐ", "x": 3, "y": 3, "color": "orange"},
    {"name": "ะะผะตะณะฐ", "x": 4, "y": 4, "color": "purple"}
]

print("๐ญ ะคะะะะะะ ะะะะะขะะ ะะะะฃะฉะะะ!")
print(f"๐ ะ ะพัะตัะตะดะธ ะฝะฐ ะฟัะพะธะทะฒะพะดััะฒะพ: {len(robots_data)} ัะพะฑะพัะพะฒ")

# ะกะพะทะดะฐะตะผ ะฒัะตั ัะพะฑะพัะพะฒ ะธะท ัะฟะธัะบะฐ ะดะฐะฝะฝัั
robot_army = []  # ัะฟะธัะพะบ ะดะปั ััะฐะฝะตะฝะธั ัะพะฑะพัะพะฒ

print("\n๐ง ะะะะะะะะะกะขะะ ะะะะะขะะ:")
for robot_data in robots_data:
    # ะกะพะทะดะฐะตะผ ัะพะฑะพัะฐ ะธะท ะดะฐะฝะฝัั ัะปะพะฒะฐัั
    robot = Robot(
        robot_data["name"],    # ะธะผั
        robot_data["x"],       # x ะบะพะพัะดะธะฝะฐัะฐ  
        robot_data["y"],       # y ะบะพะพัะดะธะฝะฐัะฐ
        robot_data["color"]    # ัะฒะตั
    )
    robot_army.append(robot)

print(f"\nโ ะัะพะธะทะฒะตะดะตะฝะพ ัะพะฑะพัะพะฒ: {len(robot_army)}")

# ะะพะบะฐะทะฐัั ะธะฝัะพัะผะฐัะธั ะพ ะฒัะตั ัะพะฑะพัะฐั
print("\n๐ ะะะะกะขะ ะะะะะขะะ:")
print("   ะกัะฐััั | ะะผั        | ะะพะทะธัะธั    | ะจะฐะณะพะฒ | ะฆะฒะตั")
print("   " + "โ" * 50)

# ะะพะบะฐะทัะฒะฐะตะผ ะธะฝัะพัะผะฐัะธั ะพ ะบะฐะถะดะพะผ ัะพะฑะพัะต
for robot in robot_army:
    robot.show_info()

# ะะฐะฟัััะธัั ะฒัะตั ัะพะฑะพัะพะฒ ัะฐะฑะพัะฐัั
print("\n๐จ ะะะกะกะะะะ ะะะะะะะะะกะขะะ ะะะะะะะขะะ:")

# ะะฐััะฐะฒะปัะตะผ ะฒัะตั ัะพะฑะพัะพะฒ ะฝะฐัะธัะพะฒะฐัั ะบะฒะฐะดัะฐัั ัะฐะทะฝะพะณะพ ัะฐะทะผะตัะฐ
# ะะปั ะบะพะฝัะตะฝััะธัะตัะบะธั ะบะฒะฐะดัะฐัะพะฒ ะธัะฟะพะปัะทัะตะผ ัะฐะทะผะตัั: 5, 4, 3, 2, 1
sizes = [5, 4, 3, 2, 1]  # ัะฐะทะผะตัั ะบะฒะฐะดัะฐัะพะฒ ะดะปั ะบะพะฝัะตะฝััะธัะตัะบะธั ะบะฒะฐะดัะฐัะพะฒ

for i, robot in enumerate(robot_army):
    size = sizes[i]
    print(f"๐ค {robot.name} ัะธััะตั ะบะฒะฐะดัะฐั {size}x{size}...")
    robot.walk_square(size)

# ะคะธะฝะฐะปัะฝะฐั ััะฐัะธััะธะบะฐ
print("\n๐ ะคะะะะะฌะะะฏ ะกะขะะขะะกะขะะะ:")
print("   ะกัะฐััั | ะะผั        | ะะพะทะธัะธั    | ะจะฐะณะพะฒ | ะฆะฒะตั") 
print("   " + "โ" * 50)

# ะกะฝะพะฒะฐ ะฟะพะบะฐะทัะฒะฐะตะผ ะธะฝัะพัะผะฐัะธั ะพ ะฒัะตั ัะพะฑะพัะฐั
for robot in robot_army:
    robot.show_info()

# ะะฝะฐะปะธะท ัะตะทัะปััะฐัะพะฒ  
print(f"\n๐ ะะะะะะ ะะะะะะะะะกะขะะ:")

# ะะพะดััะธััะฒะฐะตะผ ััะฐัะธััะธะบั
total_steps = 0
robots_at_home = 0

for robot in robot_army:
    total_steps += robot.steps_made  # ะดะพะฑะฐะฒะปัะตะผ ัะฐะณะธ ัะพะฑะพัะฐ
    
    # ะัะพะฒะตััะตะผ, ะดะพะผะฐ ะปะธ ัะพะฑะพั
    if robot.x == robot.start_x and robot.y == robot.start_y:
        robots_at_home += 1

print(f"   ๐ค ะัะตะณะพ ัะพะฑะพัะพะฒ: {len(robot_army)}")
print(f"   ๐ฃ ะะฑัะตะต ะบะพะปะธัะตััะฒะพ ัะฐะณะพะฒ: {total_steps}")
print(f"   ๐ ะกัะตะดะฝะตะต ัะฐะณะพะฒ ะฝะฐ ัะพะฑะพัะฐ: {total_steps / len(robot_army):.1f}")
print(f"   ๐ ะะพะฑะพัะพะฒ ะฒะตัะฝัะปะพัั ะดะพะผะพะน: {robots_at_home}")
print(f"   ๐ ะญััะตะบัะธะฒะฝะพััั ะฒะพะทะฒัะฐัะฐ: {robots_at_home / len(robot_army) * 100:.1f}%")

# ะะพะธัะบ ัะตะผะฟะธะพะฝะพะฒ
most_active = max(robot_army, key=lambda r: r.steps_made)
print(f"   ๐ ะกะฐะผัะน ะฐะบัะธะฒะฝัะน: {most_active.name} ({most_active.steps_made} ัะฐะณะพะฒ)")

print(f"\n๐ ะคะะะะะะ ะะะะะะจะะะ ะะะะะขะฃ! ะัะต ัะพะฑะพัั ะณะพัะพะฒั ะบ ัะปัะถะฑะต!")

```
