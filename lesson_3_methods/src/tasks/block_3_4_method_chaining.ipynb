## üéØ –¢–µ–æ—Ä–∏—è

**–¶–µ–ø–æ—á–∫–∞ –≤—ã–∑–æ–≤–æ–≤ (Method Chaining):**
- –í–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –≤—ã–∑—ã–≤–∞—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–µ—Ç–æ–¥–æ–≤ –ø–æ–¥—Ä—è–¥ —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É
- –ö–∞–∂–¥—ã–π –º–µ—Ç–æ–¥ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç `self` (—Å—Å—ã–ª–∫—É –Ω–∞ —Å–∞–º –æ–±—ä–µ–∫—Ç)
- –ü–æ–∑–≤–æ–ª—è–µ—Ç –ø–∏—Å–∞—Ç—å –±–æ–ª–µ–µ —ç–ª–µ–≥–∞–Ω—Ç–Ω—ã–π –∏ —á–∏—Ç–∞–µ–º—ã–π –∫–æ–¥

**–ü—Ä–∏–º–µ—Ä—ã –∏–∑ –∂–∏–∑–Ω–∏:**
```python
# –û–±—ã—á–Ω—ã–π —Å–ø–æ—Å–æ–± (–¥–ª–∏–Ω–Ω–æ –∏ —Å–∫—É—á–Ω–æ)
robot.move_up()
robot.move_right()
robot.rest()
robot.show_status()

# –° —Ü–µ–ø–æ—á–∫–æ–π (–∫–æ—Ä–æ—Ç–∫–æ –∏ –∫—Ä–∞—Å–∏–≤–æ)
robot.move_up().move_right().rest().show_status()
```

**–ö–∞–∫ —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç:**
1. `robot.move_up()` –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç `self` (—Å–∞–º —Ä–æ–±–æ—Ç)
2. –ù–∞ —ç—Ç–æ–º —Ä–æ–±–æ—Ç–µ –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è `.move_right()`, –∫–æ—Ç–æ—Ä—ã–π —Ç–æ–∂–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç `self`
3. –ü—Ä–æ—Ü–µ—Å—Å –ø—Ä–æ–¥–æ–ª–∂–∞–µ—Ç—Å—è –¥–∞–ª—å—à–µ –ø–æ —Ü–µ–ø–æ—á–∫–µ

**–ö–ª—é—á–µ–≤–æ–µ –ø—Ä–∞–≤–∏–ª–æ:** –ú–µ—Ç–æ–¥ –¥–æ–ª–∂–µ–Ω –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å `return self` –≤ –∫–æ–Ω—Ü–µ

**Fluent Interface:**
- –ü–∞—Ç—Ç–µ—Ä–Ω –ø—Ä–æ–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è "—Ç–µ–∫—É—á–µ–≥–æ" –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å–∞
- –ö–æ–¥ —á–∏—Ç–∞–µ—Ç—Å—è –ø–æ—á—Ç–∏ –∫–∞–∫ –æ–±—ã—á–Ω—ã–π —è–∑—ã–∫
- `robot.move_to(5, 3).rest().learn_skill("jump").show_status()`

## üíª –ü—Ä–æ—Å—Ç–æ–π –ø—Ä–∏–º–µ—Ä –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏


````python
class Robot:
    def __init__(self, name, x=0, y=0):
        self.name = name
        self.x = x
        self.y = y
        self.energy = 100
        self.skills = []
    
    # –ú–µ—Ç–æ–¥—ã –ë–ï–ó —Ü–µ–ø–æ—á–∫–∏ (–æ–±—ã—á–Ω—ã–µ)
    def old_move_up(self):
        self.y += 1
        self.energy -= 10
        print(f"üîº {self.name} –¥–≤–∏–Ω—É–ª—Å—è –≤–≤–µ—Ä—Ö")
        # –ù–ï –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç self!
    
    # –ú–µ—Ç–æ–¥—ã –° —Ü–µ–ø–æ—á–∫–æ–π (chainable)
    def move_up(self):
        self.y += 1
        self.energy -= 10
        print(f"üîº {self.name} –¥–≤–∏–Ω—É–ª—Å—è –≤–≤–µ—Ä—Ö")
        return self  # –ö–ª—é—á–µ–≤–∞—è —Å—Ç—Ä–æ—á–∫–∞!
    
    def move_right(self):
        self.x += 1
        self.energy -= 10
        print(f"‚û°Ô∏è {self.name} –¥–≤–∏–Ω—É–ª—Å—è –≤–ø—Ä–∞–≤–æ")
        return self
    
    def move_down(self):
        self.y -= 1
        self.energy -= 10
        print(f"üîΩ {self.name} –¥–≤–∏–Ω—É–ª—Å—è –≤–Ω–∏–∑")
        return self
    
    def move_left(self):
        self.x -= 1
        self.energy -= 10
        print(f"‚¨ÖÔ∏è {self.name} –¥–≤–∏–Ω—É–ª—Å—è –≤–ª–µ–≤–æ")
        return self
    
    def rest(self):
        self.energy = 100
        print(f"üò¥ {self.name} –æ—Ç–¥–æ—Ö–Ω—É–ª")
        return self
    
    def learn_skill(self, skill):
        self.skills.append(skill)
        print(f"üéì {self.name} –∏–∑—É—á–∏–ª –Ω–∞–≤—ã–∫: {skill}")
        return self
    
    def show_status(self):
        print(f"üìä {self.name}: –ø–æ–∑–∏—Ü–∏—è ({self.x}, {self.y}), —ç–Ω–µ—Ä–≥–∏—è {self.energy}")
        print(f"üõ†Ô∏è –ù–∞–≤—ã–∫–∏: {', '.join(self.skills) if self.skills else '–Ω–µ—Ç'}")
        return self
    
    def say(self, message):
        print(f"üí¨ {self.name}: '{message}'")
        return self

# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è
robot = Robot("–¶–µ–ø–æ—á–µ—á–Ω–∏–∫")

print("=== –û–ë–´–ß–ù–´–ô –°–ü–û–°–û–ë ===")
robot.old_move_up()
# robot.old_move_up().move_right()  # ‚ùå –û–®–ò–ë–ö–ê! old_move_up –Ω–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç self

print("\n=== –° –¶–ï–ü–û–ß–ö–û–ô ===")
# –ü—Ä–æ—Å—Ç–∞—è —Ü–µ–ø–æ—á–∫–∞
robot.move_right().move_up()

# –°–ª–æ–∂–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞
robot.move_left().rest().learn_skill("navigation").show_status()

# –û—á–µ–Ω—å –¥–ª–∏–Ω–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞
(robot
 .say("–ù–∞—á–∏–Ω–∞—é –ø—É—Ç–µ—à–µ—Å—Ç–≤–∏–µ!")
 .move_up()
 .move_up()
 .move_right()
 .learn_skill("climbing")
 .rest()
 .say("–ü—É—Ç–µ—à–µ—Å—Ç–≤–∏–µ –∑–∞–≤–µ—Ä—à–µ–Ω–æ!")
 .show_status())

print("\n=== –ú–û–ñ–ù–û –†–ê–ó–ë–ò–í–ê–¢–¨ –ù–ê –°–¢–†–û–ö–ò ===")
result = (robot
          .move_left()
          .move_left()
          .rest()
          .learn_skill("exploration")
          .show_status())

print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç —Ü–µ–ø–æ—á–∫–∏: {type(result)} (—ç—Ç–æ —Å–∞–º —Ä–æ–±–æ—Ç!)")
````

````python
# –°–æ–∑–¥–∞–µ–º —Ä–æ–±–æ—Ç–∞ –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
test_robot = Robot("–¢–µ—Å—Ç–µ—Ä", 0, 0)

print("üß™ –¢–ï–°–¢–´ –¶–ï–ü–û–ß–ö–ò –í–´–ó–û–í–û–í –ú–ï–¢–û–î–û–í")
print("=" * 50)

# –¢–µ—Å—Ç 1: –ü—Ä–æ—Å—Ç–∞—è —Ü–µ–ø–æ—á–∫–∞
print("üìù –¢–µ—Å—Ç 1: –ü—Ä–æ—Å—Ç–∞—è —Ü–µ–ø–æ—á–∫–∞")
result = test_robot.move_to(3, 4).collect_item("–∫–ª—é—á").show_full_status()

if (
isinstance(result, Robot) and 
    test_robot.x == 3 and test_robot.y == 4 and 
    "–∫–ª—é—á" in test_robot.inventory):
    print("‚úÖ –¢–ï–°–¢ 1 –ü–†–û–ô–î–ï–ù: –ü—Ä–æ—Å—Ç–∞—è —Ü–µ–ø–æ—á–∫–∞ —Ä–∞–±–æ—Ç–∞–µ—Ç")
else:
    print("‚ùå –¢–ï–°–¢ 1 –ü–†–û–í–ê–õ–ï–ù: –ü—Ä–æ–±–ª–µ–º—ã —Å –±–∞–∑–æ–≤–æ–π —Ü–µ–ø–æ—á–∫–æ–π")
    print("üí° –ü–æ–¥—Å–∫–∞–∑–∫–∞: –≤—Å–µ –º–µ—Ç–æ–¥—ã –¥–æ–ª–∂–Ω—ã –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å return self")

# –¢–µ—Å—Ç 2: –¶–µ–ø–æ—á–∫–∞ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –ø—Ä–µ–¥–º–µ—Ç–æ–≤
print("\nüìù –¢–µ—Å—Ç 2: –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–µ–¥–º–µ—Ç–æ–≤")
initial_energy = test_robot.energy
result = (test_robot
          .collect_item("—ç–Ω–µ—Ä–≥–µ—Ç–∏–∫")
          .use_item("—ç–Ω–µ—Ä–≥–µ—Ç–∏–∫")
          .celebrate("–Ω–∞—à–µ–ª —ç–Ω–µ—Ä–≥–µ—Ç–∏–∫!"))

if (
isinstance(result, Robot) and 
    test_robot.energy > initial_energy and
    "—ç–Ω–µ—Ä–≥–µ—Ç–∏–∫" not in test_robot.inventory):
    print("‚úÖ –¢–ï–°–¢ 2 –ü–†–û–ô–î–ï–ù: –¶–µ–ø–æ—á–∫–∞ —Å –ø—Ä–µ–¥–º–µ—Ç–∞–º–∏ —Ä–∞–±–æ—Ç–∞–µ—Ç")
else:
    print("‚ùå –¢–ï–°–¢ 2 –ü–†–û–í–ê–õ–ï–ù: –ü—Ä–æ–±–ª–µ–º—ã —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –ø—Ä–µ–¥–º–µ—Ç–æ–≤")

# –¢–µ—Å—Ç 3: –¶–µ–ø–æ—á–∫–∞ —Å –ø–æ–≤—ã—à–µ–Ω–∏–µ–º —É—Ä–æ–≤–Ω—è
print("\nüìù –¢–µ—Å—Ç 3: –ü–æ–≤—ã—à–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è")
old_level = test_robot.level
result = test_robot.level_up().celebrate("–Ω–æ–≤—ã–π —É—Ä–æ–≤–µ–Ω—å!").show_full_status()

if (
isinstance(result, Robot) and 
    test_robot.level == old_level + 1 and
    test_robot.energy == 100):
    print("‚úÖ –¢–ï–°–¢ 3 –ü–†–û–ô–î–ï–ù: –¶–µ–ø–æ—á–∫–∞ —Å level_up —Ä–∞–±–æ—Ç–∞–µ—Ç")
else:
    print("‚ùå –¢–ï–°–¢ 3 –ü–†–û–í–ê–õ–ï–ù: –ü—Ä–æ–±–ª–µ–º—ã —Å –ø–æ–≤—ã—à–µ–Ω–∏–µ–º —É—Ä–æ–≤–Ω—è")

# –¢–µ—Å—Ç 4: –£—Å–ª–æ–≤–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ (–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —ç–Ω–µ—Ä–≥–∏–∏)
print("\nüìù –¢–µ—Å—Ç 4: –£—Å–ª–æ–≤–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ (—ç–Ω–µ—Ä–≥–∏—è –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–∞)")
test_robot.energy = 80
result = (test_robot
          .if_energy_above(50)
          .move_to(10, 10)
          .celebrate("—É—Å–ø–µ—à–Ω–æ–µ –ø–µ—Ä–µ–º–µ—â–µ–Ω–∏–µ!"))

if (
isinstance(result, Robot) and 
    test_robot.x == 10 and test_robot.y == 10):
    print("‚úÖ –¢–ï–°–¢ 4 –ü–†–û–ô–î–ï–ù: –£—Å–ª–æ–≤–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ —Ä–∞–±–æ—Ç–∞–µ—Ç –ø—Ä–∏ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–π —ç–Ω–µ—Ä–≥–∏–∏")
else:
    print("‚ùå –¢–ï–°–¢ 4 –ü–†–û–í–ê–õ–ï–ù: –ü—Ä–æ–±–ª–µ–º—ã —Å —É—Å–ª–æ–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–æ–π")
    print(f"   –ü–æ–∑–∏—Ü–∏—è: ({test_robot.x}, {test_robot.y})")

# –¢–µ—Å—Ç 5: –£—Å–ª–æ–≤–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ (–Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —ç–Ω–µ—Ä–≥–∏–∏)
print("\nüìù –¢–µ—Å—Ç 5: –£—Å–ª–æ–≤–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ (—ç–Ω–µ—Ä–≥–∏—è –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–∞)")
test_robot.energy = 20
old_x, old_y = test_robot.x, test_robot.y
result = (test_robot
          .if_energy_above(50)
          .move_to(20, 20)
          .celebrate("–Ω–µ –¥–æ–ª–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å—Å—è!"))

if (test_robot.x == old_x and test_robot.y == old_y):
    print("‚úÖ –¢–ï–°–¢ 5 –ü–†–û–ô–î–ï–ù: –£—Å–ª–æ–≤–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ –±–ª–æ–∫–∏—Ä—É–µ—Ç—Å—è –ø—Ä–∏ –Ω–µ–¥–æ—Å—Ç–∞—Ç–∫–µ —ç–Ω–µ—Ä–≥–∏–∏")
else:
    print("‚ùå –¢–ï–°–¢ 5 –ü–†–û–í–ê–õ–ï–ù: –¶–µ–ø–æ—á–∫–∞ –≤—ã–ø–æ–ª–Ω–∏–ª–∞—Å—å –ø—Ä–∏ –Ω–µ–¥–æ—Å—Ç–∞—Ç–∫–µ —ç–Ω–µ—Ä–≥–∏–∏")
    print(f"   –ü–æ–∑–∏—Ü–∏—è –∏–∑–º–µ–Ω–∏–ª–∞—Å—å: ({old_x}, {old_y}) ‚Üí ({test_robot.x}, {test_robot.y})")

# –¢–µ—Å—Ç 6: –î–ª–∏–Ω–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞
print("\nüìù –¢–µ—Å—Ç 6: –û—á–µ–Ω—å –¥–ª–∏–Ω–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞")
test_robot.energy = 100
result = (test_robot
          .move_to(0, 0)
          .collect_item("–∫–∞—Ä—Ç–∞")
          .collect_item("–∫–æ–º–ø–∞—Å")
          .level_up()
          .move_to(5, 5)
          .use_item("–∫–∞—Ä—Ç–∞")
          .celebrate("–¥–æ—Å—Ç–∏–≥ —Ü–µ–ª–∏!")
          .show_full_status())

if isinstance(result, Robot):
    print("‚úÖ –¢–ï–°–¢ 6 –ü–†–û–ô–î–ï–ù: –î–ª–∏–Ω–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –±–µ–∑ –æ—à–∏–±–æ–∫")
else:
    print("‚ùå –¢–ï–°–¢ 6 –ü–†–û–í–ê–õ–ï–ù: –ü—Ä–æ–±–ª–µ–º—ã —Å –¥–ª–∏–Ω–Ω–æ–π —Ü–µ–ø–æ—á–∫–æ–π")

print("\nüéØ –í—Å–µ —Ç–µ—Å—Ç—ã –∑–∞–≤–µ—Ä—à–µ–Ω—ã!")
print(f"\nü§ñ –§–∏–Ω–∞–ª—å–Ω–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ —Ä–æ–±–æ—Ç–∞:")
test_robot.show_full_status()
````
