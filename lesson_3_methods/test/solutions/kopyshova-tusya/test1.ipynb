```python
# test block 1.1
class Robot:
    def __init__(self, name, x=0, y=0):
        self.name = name
        self.x = x
        self.y = y
        self.history = [(self.x, self.y)]  # –ò—Å—Ç–æ—Ä–∏—è: [(x, y), (x, y), ...]
        # TODO: –°–æ—Ö—Ä–∞–Ω–∏—Ç—å –Ω–∞—á–∞–ª—å–Ω—É—é –ø–æ–∑–∏—Ü–∏—é –≤ –∏—Å—Ç–æ—Ä–∏—é
        
    def command(self, cmd):
        """–í—ã–ø–æ–ª–Ω—è–µ—Ç –∫–æ–º–∞–Ω–¥—É: 'up', 'down', 'left', 'right'"""
        if cmd == "up":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å y, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.y += 1
            self.history.append((self.x, self.y))
            pass
        elif cmd == "down":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å y, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.y -= 1
            self.history.append((self.x, self.y))
            pass
        elif cmd == "left":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å x, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.x -= 1
            self.history.append((self.x, self.y))
            pass
        elif cmd == "right":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å x, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.x += 1
            self.history.append((self.x, self.y))
            pass
    
    def show_history(self):
        """–ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤—Å—é –∏—Å—Ç–æ—Ä–∏—é –¥–≤–∏–∂–µ–Ω–∏—è"""
        # TODO: –≤—ã–≤–µ—Å—Ç–∏ –≤—Å–µ –ø–æ–∑–∏—Ü–∏–∏ –∏–∑ self.history
        print(self.history)

# –¢–µ—Å—Ç –≤ —è—á–µ–π–∫–µ:
print("üß™ –¢–ï–°–¢ 1.1: –ò—Å—Ç–æ—Ä–∏—è —Å–æ—Å—Ç–æ—è–Ω–∏–π")
robot = Robot("–¢–µ—Å—Ç", 0, 0)
robot.command("right")
robot.command("right") 
robot.command("up")
robot.show_history()

# –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:
expected_history = [(0,0), (1,0), (2,0), (2,1)]
if robot.history == expected_history:
    print("‚úÖ –¢–ï–°–¢ –ü–†–û–ô–î–ï–ù: –ò—Å—Ç–æ—Ä–∏—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–∞")
else:
    print(f"‚ùå –¢–ï–°–¢ –ù–ï –ü–†–û–ô–î–ï–ù: –û–∂–∏–¥–∞–ª–æ—Å—å {expected_history}, –ø–æ–ª—É—á–µ–Ω–æ {robot.history}")

# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —Ç–µ—Å—Ç - –≤–æ–∑–≤—Ä–∞—Ç –¥–æ–º–æ–π:
robot2 = Robot("–¢–µ—Å—Ç2", 5, 3)
robot2.command("left")
robot2.command("down") 
robot2.command("right")
robot2.command("up")

expected_history2 = [(5,3), (4,3), (4,2), (5,2), (5,3)]
if robot2.history == expected_history2:
    print("‚úÖ –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–´–ô –¢–ï–°–¢ –ü–†–û–ô–î–ï–ù: –í–æ–∑–≤—Ä–∞—Ç –≤ –∏—Å—Ö–æ–¥–Ω—É—é —Ç–æ—á–∫—É")
else:
    print(f"‚ùå –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–´–ô –¢–ï–°–¢ –ù–ï –ü–†–û–ô–î–ï–ù: –û–∂–∏–¥–∞–ª–æ—Å—å {expected_history2}, –ø–æ–ª—É—á–µ–Ω–æ {robot2.history}")
```

```python
#test block 2.1
def parse_combo_command(self, combo_cmd):
    """
    –†–∞–∑–±–∏—Ä–∞–µ—Ç –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—É—é –∫–æ–º–∞–Ω–¥—É –Ω–∞ –ø—Ä–æ—Å—Ç—ã–µ —Å –≤–∞–ª–∏–¥–∞—Ü–∏–µ–π
    'north+east' ‚Üí ['up', 'right']
    'south+west' ‚Üí ['down', 'left']
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ –∫–æ–º–∞–Ω–¥ –∏–ª–∏ None –ø—Ä–∏ –æ—à–∏–±–∫–µ
    """
    # –°–ª–æ–≤–∞—Ä—å –ø–µ—Ä–µ–≤–æ–¥–∞ –∫–æ–º–∞–Ω–¥
    direction_map = {
        'north': 'up',
        'south': 'down', 
        'west': 'left',
        'east': 'right'
    }
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è
    if "+" not in combo_cmd:
        # TODO: –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∫–æ–º–∞–Ω–¥–∞ –æ–¥–∏–Ω–æ—á–Ω–æ–π –≤–∞–ª–∏–¥–Ω–æ–π
        # –µ—Å–ª–∏ combo_cmd –≤ direction_map, –≤–µ—Ä–Ω—É—Ç—å [direction_map[combo_cmd]]
        # –∏–Ω–∞—á–µ –≤–µ—Ä–Ω—É—Ç—å None
        if direction_map.get(str(combo_cmd)):
            return direction_map.get(str(combo_cmd))
        else:
            return None
        pass
    
    # TODO: —Ä–∞–∑–±–∏—Ç—å –∫–æ–º–∞–Ω–¥—É —á–µ—Ä–µ–∑ split("+")
    parts = combo_cmd.split("+")
    # TODO: –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á—Ç–æ —á–∞—Å—Ç–µ–π —Ä–æ–≤–Ω–æ 2
    # TODO: –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á—Ç–æ –æ–±–µ —á–∞—Å—Ç–∏ –µ—Å—Ç—å –≤ direction_map
    # TODO: –≤–µ—Ä–Ω—É—Ç—å —Å–ø–∏—Å–æ–∫ –ø–µ—Ä–µ–≤–µ–¥–µ–Ω–Ω—ã—Ö –∫–æ–º–∞–Ω–¥ –∏–ª–∏ None –ø—Ä–∏ –æ—à–∏–±–∫–µ
    if len(parts) == 2 :
        if direction_map.get(str(parts[0])) and direction_map.get(str(parts[1])):
            return parts
        else:
            return None
    else:
        return None
    
# –î–æ–±–∞–≤–∏—Ç—å –º–µ—Ç–æ–¥ –≤ –∫–ª–∞—Å—Å Robot
class Robot:
    def __init__(self, name, x=0, y=0):
        self.name = name
        self.x = x
        self.y = y
        self.history = [(self.x, self.y)]  # –ò—Å—Ç–æ—Ä–∏—è: [(x, y), (x, y), ...]
        # TODO: –°–æ—Ö—Ä–∞–Ω–∏—Ç—å –Ω–∞—á–∞–ª—å–Ω—É—é –ø–æ–∑–∏—Ü–∏—é –≤ –∏—Å—Ç–æ—Ä–∏—é
        
    def command(self, cmd):
        """–í—ã–ø–æ–ª–Ω—è–µ—Ç –∫–æ–º–∞–Ω–¥—É: 'up', 'down', 'left', 'right'"""
        if cmd == "up":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å y, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.y += 1
            self.history.append((self.x, self.y))
            pass
        elif cmd == "down":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å y, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.y -= 1
            self.history.append((self.x, self.y))
            pass
        elif cmd == "left":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å x, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.x -= 1
            self.history.append((self.x, self.y))
            pass
        elif cmd == "right":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å x, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.x += 1
            self.history.append((self.x, self.y))
            pass
    
    def show_history(self):
        """–ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤—Å—é –∏—Å—Ç–æ—Ä–∏—é –¥–≤–∏–∂–µ–Ω–∏—è"""
        # TODO: –≤—ã–≤–µ—Å—Ç–∏ –≤—Å–µ –ø–æ–∑–∏—Ü–∏–∏ –∏–∑ self.history
        print(self.history)


    def parse_combo_command(self, combo_cmd):
        """
        –†–∞–∑–±–∏—Ä–∞–µ—Ç –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—É—é –∫–æ–º–∞–Ω–¥—É –Ω–∞ –ø—Ä–æ—Å—Ç—ã–µ —Å –≤–∞–ª–∏–¥–∞—Ü–∏–µ–π
        'north+east' ‚Üí ['up', 'right']
        'south+west' ‚Üí ['down', 'left']
        –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ –∫–æ–º–∞–Ω–¥ –∏–ª–∏ None –ø—Ä–∏ –æ—à–∏–±–∫–µ
        """
        # –°–ª–æ–≤–∞—Ä—å –ø–µ—Ä–µ–≤–æ–¥–∞ –∫–æ–º–∞–Ω–¥
        direction_map = {
            'north': 'up',
            'south': 'down', 
            'west': 'left',
            'east': 'right'
        }
        
        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è
        if "+" not in combo_cmd:
            # TODO: –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∫–æ–º–∞–Ω–¥–∞ –æ–¥–∏–Ω–æ—á–Ω–æ–π –≤–∞–ª–∏–¥–Ω–æ–π
            # –µ—Å–ª–∏ combo_cmd –≤ direction_map, –≤–µ—Ä–Ω—É—Ç—å [direction_map[combo_cmd]]
            # –∏–Ω–∞—á–µ –≤–µ—Ä–Ω—É—Ç—å None
            if direction_map.get(str(combo_cmd)):
                direct = [str(direction_map.get(str(combo_cmd)))]
                return direct
            else:
                return None
            pass
        
        # TODO: —Ä–∞–∑–±–∏—Ç—å –∫–æ–º–∞–Ω–¥—É —á–µ—Ä–µ–∑ split("+")
        parts = combo_cmd.split("+")
        # TODO: –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á—Ç–æ —á–∞—Å—Ç–µ–π —Ä–æ–≤–Ω–æ 2
        # TODO: –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á—Ç–æ –æ–±–µ —á–∞—Å—Ç–∏ –µ—Å—Ç—å –≤ direction_map
        # TODO: –≤–µ—Ä–Ω—É—Ç—å —Å–ø–∏—Å–æ–∫ –ø–µ—Ä–µ–≤–µ–¥–µ–Ω–Ω—ã—Ö –∫–æ–º–∞–Ω–¥ –∏–ª–∏ None –ø—Ä–∏ –æ—à–∏–±–∫–µ
        if len(parts) == 2 :
            if direction_map.get(str(parts[0])) and direction_map.get(str(parts[1])):
                direct = [direction_map.get(str(parts[0])), direction_map.get(str(parts[1]))]
                return direct
            else:
                return None
        else:
            return None

# –¢–µ—Å—Ç—ã:
print("\nüß™ –¢–ï–°–¢ 2.1: –†–∞–∑–±–æ—Ä –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∫–æ–º–∞–Ω–¥")

robot = Robot("–ü–∞—Ä—Å–µ—Ä")

# –¢–µ—Å—Ç 1: –í–∞–ª–∏–¥–Ω–∞—è –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –∫–æ–º–∞–Ω–¥–∞
result1 = robot.parse_combo_command("north+east")
expected1 = ['up', 'right']
if result1 == expected1:
    print("‚úÖ –¢–ï–°–¢ 1 –ü–†–û–ô–î–ï–ù: north+east ‚Üí ['up', 'right']")
else:
    print(f"‚ùå –¢–ï–°–¢ 1 –ù–ï –ü–†–û–ô–î–ï–ù: –û–∂–∏–¥–∞–ª–æ—Å—å {expected1}, –ø–æ–ª—É—á–µ–Ω–æ {result1}")

# –¢–µ—Å—Ç 2: –í–∞–ª–∏–¥–Ω–∞—è –æ–¥–∏–Ω–æ—á–Ω–∞—è –∫–æ–º–∞–Ω–¥–∞
result2 = robot.parse_combo_command("west")
expected2 = ['left']
if result2 == expected2:
    print("‚úÖ –¢–ï–°–¢ 2 –ü–†–û–ô–î–ï–ù: west ‚Üí ['left']")
else:
    print(f"‚ùå –¢–ï–°–¢ 2 –ù–ï –ü–†–û–ô–î–ï–ù: –û–∂–∏–¥–∞–ª–æ—Å—å {expected2}, –ø–æ–ª—É—á–µ–Ω–æ {result2}")

# –¢–µ—Å—Ç 3: –ù–µ–≤–∞–ª–∏–¥–Ω–∞—è –∫–æ–º–∞–Ω–¥–∞
result3 = robot.parse_combo_command("north+invalid")
expected3 = None
if result3 == expected3:
    print("‚úÖ –¢–ï–°–¢ 3 –ü–†–û–ô–î–ï–ù: –ù–µ–≤–∞–ª–∏–¥–Ω–∞—è –∫–æ–º–∞–Ω–¥–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç None")
else:
    print(f"‚ùå –¢–ï–°–¢ 3 –ù–ï –ü–†–û–ô–î–ï–ù: –û–∂–∏–¥–∞–ª–æ—Å—å {expected3}, –ø–æ–ª—É—á–µ–Ω–æ {result3}")

# –¢–µ—Å—Ç 4: –°–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ —á–∞—Å—Ç–µ–π
result4 = robot.parse_combo_command("north+east+south")
expected4 = None
if result4 == expected4:
    print("‚úÖ –¢–ï–°–¢ 4 –ü–†–û–ô–î–ï–ù: –ö–æ–º–∞–Ω–¥–∞ –∏–∑ 3 —á–∞—Å—Ç–µ–π –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç None")
else:
    print(f"‚ùå –¢–ï–°–¢ 4 –ù–ï –ü–†–û–ô–î–ï–ù: –û–∂–∏–¥–∞–ª–æ—Å—å {expected4}, –ø–æ–ª—É—á–µ–Ω–æ {result4}")

# –¢–µ—Å—Ç 5: –ö–æ–º–∞–Ω–¥–∞ –±–µ–∑ —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è, –Ω–æ –Ω–µ–≤–∞–ª–∏–¥–Ω–∞—è
result5 = robot.parse_combo_command("invalid")
expected5 = None
if result5 == expected5:
    print("‚úÖ –¢–ï–°–¢ 5 –ü–†–û–ô–î–ï–ù: –ù–µ–≤–∞–ª–∏–¥–Ω–∞—è –æ–¥–∏–Ω–æ—á–Ω–∞—è –∫–æ–º–∞–Ω–¥–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç None")
else:
    print(f"‚ùå –¢–ï–°–¢ 5 –ù–ï –ü–†–û–ô–î–ï–ù: –û–∂–∏–¥–∞–ª–æ—Å—å {expected5}, –ø–æ–ª—É—á–µ–Ω–æ {result5}")
```

```python
#test block 4
class Robot:
    def __init__(self, name, x=0, y=0):
        self.name = name
        self.x = x
        self.y = y
        self.history = [(self.x, self.y)]  # –ò—Å—Ç–æ—Ä–∏—è: [(x, y), (x, y), ...]
        # TODO: –°–æ—Ö—Ä–∞–Ω–∏—Ç—å –Ω–∞—á–∞–ª—å–Ω—É—é –ø–æ–∑–∏—Ü–∏—é –≤ –∏—Å—Ç–æ—Ä–∏—é
        
    def command(self, cmd):
        """–í—ã–ø–æ–ª–Ω—è–µ—Ç –∫–æ–º–∞–Ω–¥—É: 'up', 'down', 'left', 'right'"""
        if cmd == "up":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å y, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.y += 1
            self.history.append((self.x, self.y))
            pass
        elif cmd == "down":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å y, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.y -= 1
            self.history.append((self.x, self.y))
            pass
        elif cmd == "left":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å x, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.x -= 1
            self.history.append((self.x, self.y))
            pass
        elif cmd == "right":
            # TODO: –∏–∑–º–µ–Ω–∏—Ç—å x, —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∏—Å—Ç–æ—Ä–∏—é
            self.x += 1
            self.history.append((self.x, self.y))
            pass
    
    def show_history(self):
        """–ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤—Å—é –∏—Å—Ç–æ—Ä–∏—é –¥–≤–∏–∂–µ–Ω–∏—è"""
        # TODO: –≤—ã–≤–µ—Å—Ç–∏ –≤—Å–µ –ø–æ–∑–∏—Ü–∏–∏ –∏–∑ self.history
        print(self.history)


    def parse_combo_command(self, combo_cmd):
        """
        –†–∞–∑–±–∏—Ä–∞–µ—Ç –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—É—é –∫–æ–º–∞–Ω–¥—É –Ω–∞ –ø—Ä–æ—Å—Ç—ã–µ —Å –≤–∞–ª–∏–¥–∞—Ü–∏–µ–π
        'north+east' ‚Üí ['up', 'right']
        'south+west' ‚Üí ['down', 'left']
        –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ –∫–æ–º–∞–Ω–¥ –∏–ª–∏ None –ø—Ä–∏ –æ—à–∏–±–∫–µ
        """
        # –°–ª–æ–≤–∞—Ä—å –ø–µ—Ä–µ–≤–æ–¥–∞ –∫–æ–º–∞–Ω–¥
        direction_map = {
            'north': 'up',
            'south': 'down', 
            'west': 'left',
            'east': 'right'
        }
        
        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è
        if "+" not in combo_cmd:
            # TODO: –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∫–æ–º–∞–Ω–¥–∞ –æ–¥–∏–Ω–æ—á–Ω–æ–π –≤–∞–ª–∏–¥–Ω–æ–π
            # –µ—Å–ª–∏ combo_cmd –≤ direction_map, –≤–µ—Ä–Ω—É—Ç—å [direction_map[combo_cmd]]
            # –∏–Ω–∞—á–µ –≤–µ—Ä–Ω—É—Ç—å None
            if direction_map.get(str(combo_cmd)):
                direct = [str(direction_map.get(str(combo_cmd)))]
                return direct
            else:
                return None
            pass
        
        # TODO: —Ä–∞–∑–±–∏—Ç—å –∫–æ–º–∞–Ω–¥—É —á–µ—Ä–µ–∑ split("+")
        parts = combo_cmd.split("+")
        # TODO: –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á—Ç–æ —á–∞—Å—Ç–µ–π —Ä–æ–≤–Ω–æ 2
        # TODO: –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á—Ç–æ –æ–±–µ —á–∞—Å—Ç–∏ –µ—Å—Ç—å –≤ direction_map
        # TODO: –≤–µ—Ä–Ω—É—Ç—å —Å–ø–∏—Å–æ–∫ –ø–µ—Ä–µ–≤–µ–¥–µ–Ω–Ω—ã—Ö –∫–æ–º–∞–Ω–¥ –∏–ª–∏ None –ø—Ä–∏ –æ—à–∏–±–∫–µ
        if len(parts) == 2 :
            if direction_map.get(str(parts[0])) and direction_map.get(str(parts[1])):
                direct = [direction_map.get(str(parts[0])), direction_map.get(str(parts[1]))]
                return direct
            else:
                return None
        else:
            return None

    def execute_commands(self, commands_list):
        """–í—ã–ø–æ–ª–Ω—è–µ—Ç —Å–ø–∏—Å–æ–∫ –∫–æ–º–∞–Ω–¥ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ"""
        if commands_list is None:
            print("‚ùå –ù–µ–≤–æ–∑–º–æ–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å: –∫–æ–º–∞–Ω–¥—ã –Ω–µ —Ä–∞—Å–ø–æ–∑–Ω–∞–Ω—ã")
            return False
            
        for cmd in commands_list:
            # TODO: –≤—ã–ø–æ–ª–Ω–∏—Ç—å –∫–æ–º–∞–Ω–¥—É —á–µ—Ä–µ–∑ self.command(cmd)
            self.command(cmd)
            pass
        return True
        
    def combo_move(self, combo_cmd):
        """–û–¥–Ω–æ –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –¥–≤–∏–∂–µ–Ω–∏–µ: 'north+east' –¥–µ–ª–∞–µ—Ç —à–∞–≥ –≤–≤–µ—Ä—Ö –∏ –≤–ø—Ä–∞–≤–æ"""
        # TODO: —Ä–∞–∑–æ–±—Ä–∞—Ç—å –∫–æ–º–∞–Ω–¥—É —á–µ—Ä–µ–∑ parse_combo_command
        # TODO: –≤—ã–ø–æ–ª–Ω–∏—Ç—å –∫–æ–º–∞–Ω–¥—ã —á–µ—Ä–µ–∑ execute_commands    
        cmd_list = self.parse_combo_command(combo_cmd)
        self.execute_commands(cmd_list)
        pass

    def repeat_combo(self, combo_cmd, times):
        """–ü–æ–≤—Ç–æ—Ä—è–µ—Ç –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –¥–≤–∏–∂–µ–Ω–∏–µ –∑–∞–¥–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑"""
        print(f"üîÑ –ü–æ–≤—Ç–æ—Ä—è–µ–º '{combo_cmd}' {times} —Ä–∞–∑")
        for i in range(times):
            # TODO: –≤—ã–ø–æ–ª–Ω–∏—Ç—å combo_move
            self.combo_move(combo_cmd)
            print(f"  –ü–æ–≤—Ç–æ—Ä {i+1}: —Ä–æ–±–æ—Ç –≤ ({self.x}, {self.y})")

    def draw_diamond(self, size):
        """–†–∏—Å—É–µ—Ç —Ä–æ–º–± –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞"""
        print(f"üî∑ {self.name} —Ä–∏—Å—É–µ—Ç —Ä–æ–º–± —Ä–∞–∑–º–µ—Ä–∞ {size}")
        
        # –§–∞–∑–∞ 1: –°–µ–≤–µ—Ä–æ-–≤–æ—Å—Ç–æ–∫ (size —Ä–∞–∑)
        print("–§–∞–∑–∞ 1: –°–µ–≤–µ—Ä–æ-–≤–æ—Å—Ç–æ–∫")
        # TODO: repeat_combo("north+east", size)
        self.repeat_combo("north+east", size)
        
        # –§–∞–∑–∞ 2: –Æ–≥–æ-–≤–æ—Å—Ç–æ–∫ (size —Ä–∞–∑)
        print("–§–∞–∑–∞ 2: –Æ–≥–æ-–≤–æ—Å—Ç–æ–∫") 
        # TODO: repeat_combo("south+east", size)
        self.repeat_combo("south+east", size-1)
        self.repeat_combo("south", 1)
        
        # –§–∞–∑–∞ 3: –Æ–≥–æ-–∑–∞–ø–∞–¥ (size —Ä–∞–∑)
        print("–§–∞–∑–∞ 3: –Æ–≥–æ-–∑–∞–ø–∞–¥")
        # TODO: repeat_combo("south+west", size)
        self.repeat_combo("west", 1)
        self.repeat_combo("south+west", size-1)
        
        # –§–∞–∑–∞ 4: –°–µ–≤–µ—Ä–æ-–∑–∞–ø–∞–¥ (size —Ä–∞–∑)
        print("–§–∞–∑–∞ 4: –°–µ–≤–µ—Ä–æ-–∑–∞–ø–∞–¥")
        # TODO: repeat_combo("north+west", size)
        self.repeat_combo("north+west", size-1)
        
        print(f"‚úÖ –†–æ–º–± –≥–æ—Ç–æ–≤! –§–∏–Ω–∞–ª—å–Ω–∞—è –ø–æ–∑–∏—Ü–∏—è: ({self.x}, {self.y})")


# –§–∏–Ω–∞–ª—å–Ω—ã–π —Ç–µ—Å—Ç:
print("\nüß™ –§–ò–ù–ê–õ–¨–ù–´–ô –¢–ï–°–¢: –†–∏—Å–æ–≤–∞–Ω–∏–µ —Ä–æ–º–±–∞")

robot = Robot("–•—É–¥–æ–∂–Ω–∏–∫", 0, 0)
start_pos = (robot.x, robot.y)
robot.draw_diamond(3)

# –ü—Ä–æ–≤–µ—Ä–∫–∞ 1: –†–æ–±–æ—Ç –¥–æ–ª–∂–µ–Ω –≤–µ—Ä–Ω—É—Ç—å—Å—è –≤ –∏—Å—Ö–æ–¥–Ω—É—é —Ç–æ—á–∫—É
if (robot.x, robot.y) == start_pos:
    print("üéØ ‚úÖ –¢–ï–°–¢ 1 –ü–†–û–ô–î–ï–ù: –†–æ–±–æ—Ç –≤–µ—Ä–Ω—É–ª—Å—è –≤ –∏—Å—Ö–æ–¥–Ω—É—é —Ç–æ—á–∫—É!")
else:
    print(f"‚ùå –¢–ï–°–¢ 1 –ù–ï –ü–†–û–ô–î–ï–ù: –û–∂–∏–¥–∞–ª–æ—Å—å {start_pos}, –ø–æ–ª—É—á–µ–Ω–æ ({robot.x}, {robot.y})")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ 2: –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤
expected_total_steps = 25  # 1 –Ω–∞—á–∞–ª—å–Ω–∞—è + 24 —à–∞–≥–∞ (4 —Ñ–∞–∑—ã √ó 3 —Ä–∞–∑–º–µ—Ä √ó 2 –∫–æ–º–∞–Ω–¥—ã –∑–∞ —Ñ–∞–∑—É)
if len(robot.history) == expected_total_steps:
    print(f"‚úÖ –¢–ï–°–¢ 2 –ü–†–û–ô–î–ï–ù: –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤ {expected_total_steps}")
else:
    print(f"‚ùå –¢–ï–°–¢ 2 –ù–ï –ü–†–û–ô–î–ï–ù: –û–∂–∏–¥–∞–ª–æ—Å—å {expected_total_steps} —à–∞–≥–æ–≤, –ø–æ–ª—É—á–µ–Ω–æ {len(robot.history)}")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ 3: –ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã (–¥–æ–ª–∂–Ω—ã –¥–æ—Å—Ç–∏—á—å (3,3) –∏ (-3,-3))
x_coords = [pos[0] for pos in robot.history]
y_coords = [pos[1] for pos in robot.history]
max_x, min_x = max(x_coords), min(x_coords)
max_y, min_y = max(y_coords), min(y_coords)

expected_bounds = 3
if max_x == expected_bounds and min_x == -expected_bounds and max_y == expected_bounds and min_y == -expected_bounds:
    print(f"‚úÖ –¢–ï–°–¢ 3 –ü–†–û–ô–î–ï–ù: –†–æ–º–± –∏–º–µ–µ—Ç –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –≥—Ä–∞–Ω–∏—Ü—ã ¬±{expected_bounds}")
else:
    print(f"‚ùå –¢–ï–°–¢ 3 –ù–ï –ü–†–û–ô–î–ï–ù: –ì—Ä–∞–Ω–∏—Ü—ã –Ω–µ–≤–µ—Ä–Ω—ã. X: [{min_x}, {max_x}], Y: [{min_y}, {max_y}]")

# –§–∏–Ω–∞–ª—å–Ω–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è:
plot_robot_path(robot)

print(f"\nüèÜ –§–ò–ù–ê–õ–¨–ù–ê–Ø –û–¶–ï–ù–ö–ê: –ï—Å–ª–∏ –≤—Å–µ —Ç–µ—Å—Ç—ã –ø—Ä–æ–π–¥–µ–Ω—ã - 1.0 –±–∞–ª–ª!")
```
