[–ü–æ–¥—Ä–æ–±–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ —Ç–æ–≥–æ, —á—Ç–æ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤ –¥–æ–º–∞—à–Ω–µ–π —Ä–∞–±–æ—Ç–µ: —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥—ã —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –∏ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Ä–µ—à–µ–Ω–∏–µ.]


## üè† –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ

**–ó–∞–¥–∞—á–∞:** –°–æ–∑–¥–∞—Ç—å —Ä–æ–±–æ—Ç–∞-—Å—Ç—Ä–æ–∏—Ç–µ–ª—è —Å fluent interface

–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å `BuilderRobot` –∫–æ—Ç–æ—Ä—ã–π –º–æ–∂–µ—Ç —Å—Ç—Ä–æ–∏—Ç—å –∑–¥–∞–Ω–∏—è —á–µ—Ä–µ–∑ —Ü–µ–ø–æ—á–∫—É –∫–æ–º–∞–Ω–¥:

1.  **–ú–µ—Ç–æ–¥—ã –ø–ª–∞–Ω–∏—Ä–æ–≤–∞–Ω–∏—è:**
    -   `start_building(building_name)` - –Ω–∞—á–∞—Ç—å –Ω–æ–≤–æ–µ –∑–¥–∞–Ω–∏–µ
    -   `add_floor(floor_type="–æ–±—ã—á–Ω—ã–π")` - –¥–æ–±–∞–≤–∏—Ç—å —ç—Ç–∞–∂
    -   `add_room(room_name, floor_number=1)` - –¥–æ–±–∞–≤–∏—Ç—å –∫–æ–º–Ω–∞—Ç—É –Ω–∞ —ç—Ç–∞–∂
    -   `set_material(material)` - —É—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –º–∞—Ç–µ—Ä–∏–∞–ª –¥–ª—è —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–∞

2.  **–ú–µ—Ç–æ–¥—ã —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–∞:**
    -   `lay_foundation()` - –∑–∞–ª–æ–∂–∏—Ç—å —Ñ—É–Ω–¥–∞–º–µ–Ω—Ç
    -   `build_walls()` - –ø–æ—Å—Ç—Ä–æ–∏—Ç—å —Å—Ç–µ–Ω—ã
    -   `install_roof()` - —É—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –∫—Ä—ã—à—É
    -   `add_windows(count)` - –¥–æ–±–∞–≤–∏—Ç—å –æ–∫–Ω–∞

3.  **–ú–µ—Ç–æ–¥—ã –∫–æ–Ω—Ç—Ä–æ–ª—è:**
    -   `inspect_quality()` - –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ (—Å–ª—É—á–∞–π–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç)
    -   `if_quality_good()` - –ø—Ä–æ–¥–æ–ª–∂–∏—Ç—å —Ü–µ–ø–æ—á–∫—É —Ç–æ–ª—å–∫–æ –ø—Ä–∏ —Ö–æ—Ä–æ—à–µ–º –∫–∞—á–µ—Å—Ç–≤–µ
    -   `calculate_cost()` - –ø–æ–¥—Å—á–∏—Ç–∞—Ç—å —Å—Ç–æ–∏–º–æ—Å—Ç—å
    -   `finish_building()` - –∑–∞–≤–µ—Ä—à–∏—Ç—å –∑–¥–∞–Ω–∏–µ

4.  **–ê—Ç—Ä–∏–±—É—Ç—ã –¥–ª—è –æ—Ç—Å–ª–µ–∂–∏–≤–∞–Ω–∏—è:**
    -   `current_building` - —Å–ª–æ–≤–∞—Ä—å —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ –∑–¥–∞–Ω–∏–∏
    -   `buildings_completed` - —Å–ø–∏—Å–æ–∫ –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—ã—Ö –∑–¥–∞–Ω–∏–π
    -   `total_cost` - –æ–±—â–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ—Ö –∑–¥–∞–Ω–∏–π

**–ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:**
```python
builder = BuilderRobot("–°—Ç—Ä–æ–∏—Ç–µ–ª—å")

# –°—Ç—Ä–æ–∏–º –¥–æ–º
builder.start_building("–î–æ–º").set_material("–∫–∏—Ä–ø–∏—á").lay_foundation().build_walls().install_roof().add_windows(4).inspect_quality().if_quality_good().finish_building()

# –°—Ç—Ä–æ–∏–º –≥–∞—Ä–∞–∂  
builder.start_building("–ì–∞—Ä–∞–∂").set_material("–±–µ—Ç–æ–Ω").lay_foundation().build_walls().install_roof().add_windows(1).finish_building()

print(f"–ü–æ—Å—Ç—Ä–æ–µ–Ω–æ –∑–¥–∞–Ω–∏–π: {len(builder.buildings_completed)}")
print(f"–û–±—â–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å: {builder.total_cost}")
```

**–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ:** –î–æ–±–∞–≤—å—Ç–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å —Å—Ç—Ä–æ–∏—Ç—å –º–Ω–æ–≥–æ—ç—Ç–∞–∂–Ω—ã–µ –∑–¥–∞–Ω–∏—è:
```python
builder.start_building("–ú–Ω–æ–≥–æ—ç—Ç–∞–∂–∫–∞").add_floor("–ø–æ–¥–≤–∞–ª").add_floor("–ø–µ—Ä–≤—ã–π").add_floor("–≤—Ç–æ—Ä–æ–π").add_room("–∫—É—Ö–Ω—è", 1).add_room("—Å–ø–∞–ª—å–Ω—è", 2).finish_building()
```

**–°–¥–∞—á–∞:** –ü–æ–∫–∞–∂–∏—Ç–µ –∫–æ–¥ –∏ –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—é —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–∞ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –∑–¥–∞–Ω–∏–π –Ω–∞ —Å–ª–µ–¥—É—é—â–µ–º —É—Ä–æ–∫–µ.


````python
import random

class NullBuilderRobot:
    def __getattr__(self, name):
        return lambda *args, **kwargs: self

_null_builder_robot = NullBuilderRobot()

class BuilderRobot:
    def __init__(self, name):
        self.name = name
        self.current_building = None  # –°–ª–æ–≤–∞—Ä—å —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ —Ç–µ–∫—É—â–µ–º –∑–¥–∞–Ω–∏–∏
        self.buildings_completed = []  # –°–ø–∏—Å–æ–∫ –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—ã—Ö –∑–¥–∞–Ω–∏–π
        self.total_cost = 0  # –û–±—â–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ—Ö –∑–¥–∞–Ω–∏–π
    
    def start_building(self, building_name):
        print(f"üèóÔ∏è {self.name} –Ω–∞—á–∏–Ω–∞–µ—Ç —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ: {building_name}")
        self.current_building = {
            "name": building_name,
            "material": "–Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ",
            "floors": {"–æ–±—ã—á–Ω—ã–π": []},  # –°–ª–æ–≤–∞—Ä—å –¥–ª—è —ç—Ç–∞–∂–µ–π –∏ –∫–æ–º–Ω–∞—Ç
            "windows": 0,
            "foundation_laid": False,
            "walls_built": False,
            "roof_installed": False,
            "quality": "–Ω–µ –ø—Ä–æ–≤–µ—Ä–µ–Ω–æ"
        }
        return self
    
    def add_floor(self, floor_type="–æ–±—ã—á–Ω—ã–π"):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        if floor_type not in self.current_building["floors"]:
            self.current_building["floors"][floor_type] = []
        print(f"‚ûï –î–æ–±–∞–≤–ª–µ–Ω —ç—Ç–∞–∂: {floor_type}")
        return self
    
    def add_room(self, room_name, floor_number=1):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        # –ü—Ä–æ—Å—Ç–∞—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è: –¥–æ–±–∞–≤–ª—è–µ–º –∫–æ–º–Ω–∞—Ç—É –∫ –ø–µ—Ä–≤–æ–º—É —ç—Ç–∞–∂—É '–æ–±—ã—á–Ω—ã–π' –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
        # –î–ª—è –±–æ–ª–µ–µ —Å–ª–æ–∂–Ω–æ–π –ª–æ–≥–∏–∫–∏ –Ω—É–∂–Ω–æ –±—É–¥–µ—Ç –æ—Ç—Å–ª–µ–∂–∏–≤–∞—Ç—å —ç—Ç–∞–∂–∏ –ø–æ –Ω–æ–º–µ—Ä–∞–º –∏–ª–∏ —Ç–∏–ø–∞–º
        floor_key = "–æ–±—ã—á–Ω—ã–π" # –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º, —á—Ç–æ —ç—Ç–∞–∂–∏ —Ö—Ä–∞–Ω—è—Ç—Å—è –ø–æ —Ç–∏–ø—É
        floor_types = list(self.current_building["floors"].keys())
        if floor_number > 0 and floor_number <= len(floor_types):
            floor_key = floor_types[floor_number - 1]
            
        if floor_key in self.current_building["floors"]:
            self.current_building["floors"][floor_key].append(room_name)
            print(f"üö™ –î–æ–±–∞–≤–ª–µ–Ω–∞ –∫–æ–º–Ω–∞—Ç–∞ '{room_name}' –Ω–∞ —ç—Ç–∞–∂ '{floor_key}'")
        else:
            print(f"‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –¥–æ–±–∞–≤–∏—Ç—å –∫–æ–º–Ω–∞—Ç—É: —ç—Ç–∞–∂ '{floor_key}' –Ω–µ –Ω–∞–π–¥–µ–Ω.")
        return self
    
    def set_material(self, material):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        self.current_building["material"] = material
        print(f"üß± –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω –º–∞—Ç–µ—Ä–∏–∞–ª: {material}")
        return self
    
    def lay_foundation(self):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        self.current_building["foundation_laid"] = True
        print(f"üèóÔ∏è –ó–∞–ª–æ–∂–µ–Ω —Ñ—É–Ω–¥–∞–º–µ–Ω—Ç –¥–ª—è {self.current_building["name"]}")
        return self
    
    def build_walls(self):
        if not self.current_building or not self.current_building["foundation_laid"]:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –∑–∞–ª–æ–∂–∏—Ç–µ —Ñ—É–Ω–¥–∞–º–µ–Ω—Ç!")
            return self
        self.current_building["walls_built"] = True
        print(f"üèóÔ∏è –ü–æ—Å—Ç—Ä–æ–µ–Ω—ã —Å—Ç–µ–Ω—ã –¥–ª—è {self.current_building["name"]}")
        return self
    
    def install_roof(self):
        if not self.current_building or not self.current_building["walls_built"]:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —Å—Ç–µ–Ω—ã!")
            return self
        self.current_building["roof_installed"] = True
        print(f"üèóÔ∏è –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∞ –∫—Ä—ã—à–∞ –¥–ª—è {self.current_building["name"]}")
        return self
    
    def add_windows(self, count):
        if not self.current_building or not self.current_building["walls_built"]:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —Å—Ç–µ–Ω—ã, —á—Ç–æ–±—ã –¥–æ–±–∞–≤–∏—Ç—å –æ–∫–Ω–∞!")
            return self
        self.current_building["windows"] = count
        print(f"üñºÔ∏è –î–æ–±–∞–≤–ª–µ–Ω–æ {count} –æ–∫–æ–Ω")
        return self
    
    def inspect_quality(self):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        quality_good = random.choice([True, False])
        self.current_building["quality"] = "—Ö–æ—Ä–æ—à–µ–µ" if quality_good else "–ø–ª–æ—Ö–æ–µ"
        print(f"üîç –ü—Ä–æ–≤–µ—Ä–µ–Ω–æ –∫–∞—á–µ—Å—Ç–≤–æ: {self.current_building["quality"]}")
        return self
    
    def if_quality_good(self):
        if not self.current_building:
            return _null_builder_robot # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –∑–∞–≥–ª—É—à–∫—É, –µ—Å–ª–∏ –Ω–µ—Ç —Ç–µ–∫—É—â–µ–≥–æ –∑–¥–∞–Ω–∏—è
        
        if self.current_building["quality"] == "—Ö–æ—Ä–æ—à–µ–µ":
            return self
        else:
            print(f"‚ùå –ö–∞—á–µ—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è {self.current_building["name"]} –Ω–µ—É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ–µ. –ü—Ä–æ–ø—É—Å–∫–∞–µ–º –¥–∞–ª—å–Ω–µ–π—à–∏–µ –¥–µ–π—Å—Ç–≤–∏—è.")
            return _null_builder_robot
            
    def calculate_cost(self):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        # –ü—Ä–æ—Å—Ç–æ–π —Ä–∞—Å—á–µ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç–∏
        cost = 0
        cost += 1000 if self.current_building["foundation_laid"] else 0
        cost += 2000 if self.current_building["walls_built"] else 0
        cost += 1500 if self.current_building["roof_installed"] else 0
        cost += self.current_building["windows"] * 100
        
        for floor_type, rooms in self.current_building["floors"].items():
            cost += 500  # –ë–∞–∑–æ–≤–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å —ç—Ç–∞–∂–∞
            cost += len(rooms) * 200 # –°—Ç–æ–∏–º–æ—Å—Ç—å –∫–æ–º–Ω–∞—Ç
            
        material_cost = {"–∫–∏—Ä–ø–∏—á": 500, "–±–µ—Ç–æ–Ω": 400, "–¥–µ—Ä–µ–≤–æ": 300}.get(self.current_building["material"], 0)
        cost += material_cost
        
        self.total_cost += cost
        print(f"üí∞ –†–∞—Å—Å—á–∏—Ç–∞–Ω–∞ —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è {self.current_building["name"]}: {cost} (–æ–±—â–∞—è: {self.total_cost})")
        return self
    
    def finish_building(self):
        if not self.current_building:
            print("‚ùå –ù–µ—Ç —Ç–µ–∫—É—â–µ–≥–æ –∑–¥–∞–Ω–∏—è –¥–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è!")
            return self
        
        # –£–±–µ–¥–∏–º—Å—è, —á—Ç–æ –≤—Å–µ –æ—Å–Ω–æ–≤–Ω—ã–µ —ç—Ç–∞–ø—ã –ø—Ä–æ–π–¥–µ–Ω—ã
        if not all([
            self.current_building["foundation_laid"],
            self.current_building["walls_built"],
            self.current_building["roof_installed"]
        ]):
            print(f"‚ö†Ô∏è –ó–¥–∞–Ω–∏–µ {self.current_building["name"]} –Ω–µ –ø–æ–ª–Ω–æ—Å—Ç—å—é –ø–æ—Å—Ç—Ä–æ–µ–Ω–æ. –í—Å–µ —Ä–∞–≤–Ω–æ –∑–∞–≤–µ—Ä—à–∞–µ–º.")
            
        self.buildings_completed.append(self.current_building)
        print(f"‚úÖ –ó–¥–∞–Ω–∏–µ {self.current_building["name"]} –∑–∞–≤–µ—Ä—à–µ–Ω–æ!")
        self.current_building = None  # –°–±—Ä–æ—Å —Ç–µ–∫—É—â–µ–≥–æ –∑–¥–∞–Ω–∏—è
        return self
````
